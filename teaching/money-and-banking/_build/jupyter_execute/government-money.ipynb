# Government Spending and Monetary Policy

In this section, we extend the overlapping generations (OLG) model by introducing a government sector. We first assume that the government finances its spending entirely through money creation—known as seigniorage—which introduces a fundamental trade-off between generating revenue and inducing inflation. If government purchases are valued by households, there may exist an optimal level of government spending. We then consider alternative sources of government revenue, such as taxation and bond issuance, which open new channels of interaction between fiscal and monetary policy. Finally, we revisit these issues through the lens of <a href="https://www.minneapolisfed.org/research/quarterly-review/some-unpleasant-monetarist-arithmetic" target="_blank">Some *Unpleasant Monetarist Arithmetic* by Sargent and Wallace (1979).</a>

## Introducing a Government into the OLG model

We begin with the simplest case in which the only source of government revenue is newly printed money, called **seigniorage**.  Government expenditures are assumed not to affect individuals' consumption choices--spending on national defense or foreign aid, for example, does not alter household's desired consumption plan.  This assumption allows us to isolate the effects of seigniorage from those of government spending itself.  Later we relax this assumption and allow government spending to directly affect agents optimal consumption plan. 

Let total real government spending be $G_t$, so that the government budget constraint is given by

$$
\begin{align*}
G_t &= M_t-M_{t-1} \\ 
&=\left(1-\frac{1}{z}\right)\nu_t M_t
\end{align*}
$$

where the right hand side is the real value of newly printed money when nominal money supply grows at constant rate $z$.  Expressed in per capita terms, where $g_t=G_t/N_t$ is government spending per person, this becomes,

$$
g_t = \left(1-\frac{1}{z}\right)q_t
$$

where $q_t \equiv \nu_t M_t/N_t$ represents real money balances.  The key insight is that government spending is limited by the demand for real money balances. Each newly printed bill has an endogenous real value $\nu_t$ which depends on household demand $q_t$.  If money ceases to be valued $q_t=0$, then newly printed bills are worthless to the government.  

### The Consumer's Problem

Market clearing requires that the stationary rate of return on money is $\nu_{t+1}/\nu_t=n/z$.  The lifetime budget constraint facing a consumer is therefore

$$
c_{1}+\frac{z}{n}c_2 \leq y
$$

Since the government uses new money for its own spending rather than redistributing it to households (notice the absence of the term $a$), seigniorage effectively shrinks the household’s budget set.

How much spending can the government finance through seigniorage?  Are there limits or is it an infinite source of revenue? Although the government can print as many bills as it likes, their value declines as the money supply expands.  The real value of newly printed bills is $\left(1-\frac{1}{z}\right)q_t$.  Although the government can print more and more bills with higher values of $z$, this inflation induces households to reduce their real money balances causing $q$ to fall.  For small inflation rates, raising $z$ can indeed increase seigniorage, but beyond some point, the decline in real balances $q_t$ dominates.   This non-monotone relationship between money creation and seigniorage revenue generates a type of <a href="https://link.springer.com/rwe/10.1007/978-1-4614-7883-6_280-2" target="_blank">Laffer curve</a>--a theoretical relationship between tax rates and government revenue, where government revenue rises for low levels of taxation, but eventually declines when taxes become too high.   

We consider a more general version of the decentralized equilibrium with seigniorage in <a href="olg-model-details.html">this example of an OLG economy with a government sector.</a>.  In this example, we do not restrict ourselves to a constant money growth rate, but instead allow the path of money to simply satisfy the governments budgetary requirements.  Furthermore, we consider both stationary and non-stationary equilibria.

### Optimal Policy under Seigniorage Financing

What is the optimal fiscal and monetary policy in this case?  When seigniorage is the sole revenue source, fiscal and monetary policy coincide—-they are determined jointly by the growth rate of money supply $z$.  A central planner consider's the following problem:

$$
\begin{align*}
\max_{c_1,c_2}&u(c_1)+\beta u(c_2) \\
&\text{s.t.} \quad c_1+c_2/n+g \leq y
\end{align*}
$$

The feasibility constraint simply requires that total expenditure is no larger than total output: $N_t c_{1,t}+N_{t-1}c_{2,t} + G_t \leq N_t y$ or if we use alternative notation to denote aggregate quantities $C_t+G_t \leq Y_t$.  Dividing this equation by $N_t$ reduces it to a per capita feasibility constraint which is precisely what is presented in the planner's problem above.

The solution the planner's problem is straightforward.  Since government spending does not enhance welfare (it is not in the utility function), the optimal government spending is $g=0$.  This implies that $z=1$ from the government budget constraint so that there is no distortion to agents consumption/savings choice coming from inflation--no inflation, no seigniorage, and no government distortion.

### Introducing Taxes

Consider an alternative source of government revenue: taxes.  Suppose that the government levies a lump-sum tax on the young, so that total real tax revenue is $T=N_t \tau_t$.  In this case, the government budget constraint bcomes

$$
G_t = T_t+\left(1-\frac{1}{z}\right)\nu_t M_t
$$

or in per capita terms

$$
g_t = \tau_t +\left(1-\frac{1}{z}\right) \nu_t m_t
$$

and the households' lifetime budget constraint is

$$
c_{1}+\frac{z}{n}c_2=y-\tau
$$

Notice something crucial: the intertemporal trade-off between present and future consumption is unaffected by a *lump-sum* tax.  It is as if a young person's income has been reduced, which has no effect on the household's Euler equation.  Said differently, the slope of the agents budget line is unaffected by a lump-sum tax.  Only inflation $z$ is relevant for distorting the agent's intertemporal trade-off.  In this case, the government can still finance its desired level of spending by using only tax revenue $g=\tau$ and no seigniorage $z=1$.  Setting $z=1$ guarantees that the lifetime budget constraint of the household coincides with that of planner, so that for a fixed level of $g$ the optimal allocation is achieved.  Of course, if government spending is still not valued by households (not in the utility function), then it is optimal to have $g=0$.  Finally, let's consider the case where government spending is valued by households.

### Valuable Government Spending

Suppose that preferences are given by

$$
u(c_{1,t},c_{2,t+1},g_t)=u(c_{1,t})+\beta u(c_{2,t+1})+\theta u(g_t)
$$

where $g_t = G_t/N_t$ is government spending per young person.  Government spending may be on education, infrastructure, law enforcement, street lighting, etc.--any goods which are valued by households.  The parameter $\theta$ controls how valuable households find these public goods relative to their private consumption.  We will allow the government to finance its expenditure by seigniorage or taxation.

First, what is the optimal allocation in this economy?  The planner's problem is

$$
\begin{align*}
\max &u(c_{1,t},c_{2,t+1},g_t) \\
&\text{s.t.} \quad c_{1,t}+\frac{c_{2,t}}{n}+g_t \leq y
\end{align*}
$$

The solution is given by equating the marginal utility across private consumption and public consumption

$$
\frac{u'(c_{1,t})}{n}=\beta u'(c_{2,t+1})=\theta u'(g_t)
$$

and that now resources are wasted so that the feasibility constraint binds.  Plugging the first-order condition into the feasibility constraint gives an implicit equation for optimal government spending

$$
u'^{-1}[\theta u'(g_t^{pp})]+u'^{-1}[\theta/\beta u'(g_t^{pp})]+u'(g_t^{pp})=y
$$

There will exist a positive and unique value of $g^{pp}$ so long as the utility function satisfies the Inada condition $\lim_{x\rightarrow 0} u'(x) = \infty$ so that there should always be some positive amount of both private and public consumption.

Now consider the households' problem

$$
\begin{align*}
\max &u(c_{1,t},c_{2,t+1}) \\
&\text{s.t.} \quad c_{1,t}+\frac{n}{z}c_{2,t} \leq y - \tau_t
\end{align*}
$$

whose Euler equation is

$$
u'(c_{1,t})\frac{z}{n}=\beta u'(c_{2,t+1})
$$

An equilibrium in the decentralized economy is given by $(c_1,c_2,g)$ computed from the Euler equation, the lifetime budget constraint, and the government budget constraint given policy variables $(\tau, z)$.  Notice that for any $z \neq 1$ the trade-off between present and future consumption is distorted relative to the planner.  This is clearly demonstrated by comparing the Euler equations.  Therefore, the only chance for $(c_1,c_2)=(c_1^{pp},c_2^{pp})$ is to set $z=1$.  Of course, this precludes of seigniorage revenue implying that all government spending is financed by taxes $g=\tau$.  But notice that with $g=\tau$ and $z=1$ the lifetime budget constraint of the household exactly coincides with that of the planner. Therefore, the decentralized economy can achieve the optimal allocation!  

The main lesson is that using the printing press to finance government budgets introduces a welfare cost of inflation.  If instead lump-sum taxes are used, then these intertemporal distortion does not appear.  Why not use lump-sum taxes for everything?  There are at least two problems. The first concerns equity.  In practice we care about economic inequality, and the burden of a fixed tax is heavier on the poor than the rich.  Second concerns enforcement. There needs to be appropriate bureaucratic apparatus to collect the taxes.  It is extremely easy (and essentially costless) to fire up the printing presses.

## Some unpleasant monetarist arithmetic

This section reviews an important paper on the relationship between fiscal and monetary policy.  <a href="https://www.minneapolisfed.org/research/quarterly-review/some-unpleasant-monetarist-arithmetic" target="_blank">Sargent and Wallace's (1981)</a> paper *Some Unpleasant Monetarist Arithmetic* describe a situation where monetary policy has limited control over long-run inflation.  This occurs because the government, committed to a particular path for fiscal deficits, must finance its spending through either money creation or bond issuance.  Tightening monetary policy increases teh growth of interest-bearing government debt, which eventually necessitates higher money supply growth to services the accumulated debt and interest, thereby increasing inflation.  

This is a startling conclusion given conventional wisdom at the time of the paper's publication was that monetary policy, if it could exert any influence over the economy, could control long-run inflation.  This position is most eloquently presented in 1968 by <a href="https://www.aeaweb.org/aer/top20/58.1.1-17.pdf" target="_blank">Milton Friedman in his presidential address to the AEA</a>.
Friedman argued that monetary policy cannot permanently influence real output, unemployment, or real interest rates; but can exert substantial control over long-run inflation.  Sargent and Wallace show that that Friedman's dictum needs certain qualifications.  In particular, economists must explicitly describe the interaction between fiscal and monetary policies.  

The paper defines two different regimes: (i) **dominant monetary policy** or (ii) **dominant fiscal policy**.  Dominant monetary policy means that the monetary authority decides the path of base money so that the amount of seigniorage revenue supplied to the fiscal authority is pre-determined.  The fiscal authority therefore faces spending constraints imposed by the public's demand for government bonds.  In this regime, monetary policy retains control over inflation.  Dominant fiscal policy instead supposes that the fiscal authority decides its path of budget deficits, thus pre-determining the amount of revenue that must be raised through bond sales and seignorage.  The monetary authority ability to finance government spending is constrained by the demand for government bonds, and must finance the remainder with seignorage.  In this regime, it is possible that monetary policy looses control over inflation: tight monetary policy today may imply high future inflation.

Relative the government budget constraint described in the OLG model, we add the possibility that the government can raise revenue by issuing government bonds.  We have,

$$
D_t = \frac{M_t-M_{t-1}}{p_t}+B_t-r_{t-1}B_{t-1}
$$

where $D_t$ is the real primary government deficit (the difference between spending and taxes $G_t-T_t$), $(M_t-M_{t-1})/\nu_t$ is seigniorage revenue, $B_t$ is the real value of newly issues bonds, and $r_{t-1}$ is the real interest rate on previously issued bonds.

The environment Sargent and Wallace consider is consistent with the assumption underpinning our OLG model.  In particular, there is a common constant growth rate $n$ for real income and population; a quantity theory demand for base money; and a constant real return on government bonds that exceeds the growth rate of real income ($r > n$).

We describe **fiscal policy** by sequence of deficits $\{ D_t \}_{t=0}^{\infty}$ and **monetary policy** by a sequence of base money $\{ M_t \}_{t=0}^{\infty}$. We assume that policies are announced at $t=1$ and are perceived to be credible.  **Fiscal dominance** is where $\{ D_t \}_{t=0}^{\infty}$ is announced first and $\{ M_t \}_{t=0}^{\infty}$ reacts.  **Monetary dominance** is where $\{ M_t \}_{t=0}^{\infty}$ is announced first and $\{ D_t \}_{t=0}^{\infty}$ reacts.

Here is how we specify alternative time paths for monetary policy:
- take $M_1$ as given
- let alternative monetary policies be different choices for growth rate $z$ of $M_t$ for $t=2,3,...,T$
- for $t > T$ the path of $M_t$ is determined by the condition that the stock of real government debt per capita is held constant at whatever level it attains at $t=T$
- we say that $z_A < z_B$ means policy A is tighter than policy B

Our assumptions about a monetarist economy imply that the price level at any time is proportional to the stock of base money per capita,

$$
p_t = (1/h)\frac{M_t}{N_t}
$$

Why?  Recall that money market clearing says $M_t/p_t = q_t(p_{t+1}/p_t) \cdot N_t$, where $q(\cdot)$ is the real money demand function. 


Dividing both sides of the government budget constraint by $N_t$ and using the quantity theory relationship gives us the per capita form of the consolidated budget constraint,

$$
d_t = h(1-1/z)+b_t-\frac{r_{t-1}}{n}b_{t-1}
$$

The deficit must be financed by some combination of currency creation and interest-bearing debt.  The amount that can raised through currency depends on the monetary authority's choice of $z$.

**Proposition 1**: If fiscal policy in the form of the $D_t$ sequence is taken as given, then tighter current monetary policy implies higher future inflation.
- Step 1: show that inflation after $T$ increases in $b_T$
- Step 2: show that $b_t$ decreases in $z$



Tighter monetary policy now implies higher future inflation:

First we use the consolidated budget constraint (3) for any date $t > T$, substituting into it $b_t=b_{t+1}=b_T$, and recognize that (2) implies inflation is $\pi_t \equiv p_t/p_{t-1}=z/n$.
$$
\Rightarrow \quad 1-\frac{1}{n \cdot \pi} =\frac{1}{h}\left[d_t+b_T\left(\frac{r_{t-1}-n}{n}\right)\right]
$$

which shows a positive relationship between $b_T$ and $\pi$.


Second we compute the entire path of $b_t$ for $t = 1,2,...T$ using (3).


Recall that our budget constraint is a first order difference equation in $b_t$,

$$
b_t = d_t+\frac{r_{t-1}}{n}b_{t-1}-h\left(\frac{z-1}{z}\right)
$$

and notice that it has a recursive form,

$$
b_t = d_t+\frac{r_{t-1}}{n} \left[ d_{t-1}+\frac{r_{t-2}}{n}b_{t-2}-h\left(\frac{z-1}{z}\right) \right] -h\left(\frac{z-1}{z}\right)
$$

Iterated substitution leads to 

$$
b_t = \phi_{t,1} b_1 + \sum_{s=2}^t \phi_{t,s}d_s - h\left(\frac{z-1}{z}\right)\sum_{s=2}^{t}\phi_{t,s}
$$

where $\phi_{t,s}=\prod_{j=s}^{t-1} \left( \frac{r_j}{n}\right)^{t-s}$ and $\phi_{t,t}=1$.


It follows that $b_T$ is larger the smaller $z$ is.


Why? Tighter monetary policy (small $z$) generates less seignorage.  For a given $d_t$ sequence, this increases bond issuances causing $b_T$ to be large.


A technical note: 
We begin the sequence $b_t$ at time $t=1$ as follows

$$
b_1 = \frac{\tilde{B}_0}{N_1 \cdot p_1} + \frac{D_1}{N_1} - \frac{M_1-M_0}{N_1 \cdot p_1}
$$

Here, in place of $B_0[1+R_0]$, we have used $\tilde{B}_0/p_1$ where $\tilde{B}_0$ is the nomianl par value of debt issued at $t=0$.  Since our economic reasoning begins at period $t=1$, we use this substitution to avoid assuming anything about the relationship between acutal and expected inflation from time $t=0$ to time $t=1$.

**Proposition 2**: If fiscal policy in the form of the $D_t$ sequence is taken as given, then tighter current monetary policy can imply higher inflation now.
- We allow money demand to depend on expected future inflation,

$$
q_t = q(\pi_t)
$$

- High rates of money creation in the future tend to raise the current rate of inflation
- We already showed that tighter money now implies high expected inflation after $T$ (see proposition 1).
- The net effect of tighter money now (lower inflation now) and higher expected future inflation (higher inflation now) is ambiguous.

We maintain of the features as before except one: replace equation (2) by

$$
\frac{M_t}{N_t p_t} = \frac{\gamma_1}{2} - \frac{\gamma_2}{2} \frac{p_{t+1}}{p_t}, \quad \gamma_1 < \gamma_2 < 0
$$

Conclusions:
- Even in a monetarist economy, dominant fiscal policy produces spillover effects on monetary policy, and can ultimately undermine long-run inflation objectives of the central bank
- If the monetary authority is tied to the fiscal authority (modeled using a consolidated budget constraint), then long-run inflation may not align with central bank objectives.
- Central bank independence is crucial in allowing the monetary authority control over long-run inflation.
