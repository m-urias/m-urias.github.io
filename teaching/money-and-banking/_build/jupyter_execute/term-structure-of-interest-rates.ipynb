# Term Structure of Interest Rates

Often in economic models we assume that there exists a single interest rate that measures the real return to capital investment.  In reality, however, there are *many* different interest rates across assets that reflect differences in **risk, liquidity, and maturity** among heterogeneous capital projects.  We restrict our attention to the class of assets called **bonds**, and begin by understanding why bonds with the same term to maturity can have different interest rates.  We then examine how the *term to maturity itself* influences a bond's interest rate--this relationship is called the **term structure of interest rates**.

In addition to being important barometers of the rate of return on capital projects, interest rates are also central to monetary policy.  One key summary statistic which shows how short-term policy rates transmit to longer-term interest rates is the **yield curve**.  The yield curve gives interest rates on the same asset (mostly bonds or other debt instruments) at different maturities. Through open market operations and other instruments (e.g. reserve requirements, discount lending, repo markets), the central bank can influence the shape of the yield curve, especially its short end.  In later sections, we will also describe how central banks implement these policies under the **corridor** and **floor systems** of monetary policy.

## What is the Risk Structure of Interest Rates?

Interest rates on long-term bonds are crucial for many economic decisions-—such as firms’ investment in plants and equipment or households’ purchases of homes and automobiles.  For this reason, macroeconomists devote substantial attention to understanding how long-term rates are determined and how they interact with short-term rates. For monetary policy analysis this is a crucial topic, as it concerns the link between short-term policy rates and long-term rates.

It is immediately apparent that different categories of bonds yield systematically different yields.  Coporate bonds with lower credit ratings (e.g., Baa) pay a higher interest rates compared to higher rated  (Aaa) bonds, and both typically yield more than U.S. government bonds.  These **yield spreads** reflect compensation for differences in default **risk** and **liquidity**.  In general, investors demand higher yields to hold bonds that are riskier or less liquid. 

There are three main factors that influence the interest rate on debt instruments:

1. **Risk**:  Some types of debt--such as US Treasury bonds--have very little risk of default.  Other types--like credit cards, corporate bonds, mortgages--differ in the probability of default and in the potential loss to lenders if a default occurs.  **Default** occurs when the issuer of a bond is unable or unwilling to make interest payments or pay off the face value when the bond matures.  The spread between interest rates on bonds with default risk and interest rate on default-free bonds (both of same maturity) is called the **risk premium**.  This spread varies over time with perceptions of creditworthiness and overall market risk.

2. **Liquidity**:  Some types of debt can be easily traded (e.g. US Treasury bonds) and/or can be quickly and cheaply converted into cash (or reserves).  Corporate bonds, in contrast, may be thinly traded and costly to liquidate, especially during periods of financial stress.  In addition, some debt instruments are readily accepted as **collateral** in financial contracts.  The degree to which an asset is collateralizable largely depends on how quickly one party can liquidate it in the event that the counter-party breaks the contract.  All else equal, a bond that becomes less liquid will experience a fall in demand and thus a lower equilibrium price-—implying a higher yield.

3. **Maturity**: Different debt instruments have different maturities---how long the borrower has to pay back. Some instruments (e.g., repos or interbank loans) mature overnight, while others (e.g., government or corporate bonds) may extend decades into the future. The relationship between yields on bonds with identical risk and liquidity but differing maturities is called the **term structure of interest rates**.  When plotted, this relationship forms the **yield curve**.

## What is the Term Structure of Interest Rates?

A **yield curve** plots the yields on bonds with different maturities but the same risk and liquidity characteristics.  Yield curves convey crucial information about market expectations of future interest rates, inflation, and economic activity.  Empirically, several stylized facts hold over most period of time:
- Interest rates on bonds of different maturities move together over time.
- Yield curves can slope upward, downward, or be flat (but most often are upward sloping)
- When short-term interest rates are low (high), the yield curve tends to slope upward (downward)

## Explaining the yield curve

Three main theories attempt to explain the shape of the yield curve:
1. Expectations theory
2. Segmented markets
3. Liquidity premium theory

### Expectations theory

**Expectations theory** says that the interest rate on a long-term bond will equal the *average* of the short-term interest rates that people expect to occur over the bond's lifetime. This theory assumes bonds of different maturities are *perfect substitutes*, so that they must offer the same expected rate of return in equilibrium. This is a simple application of the rate of return equality principle we discussed earlier.

Consider investing \$1 of wealth for a two-period return in one of the following strategies:
- Buy a two-period bond today with yield $i_{2t}$
- Buy a one-period bond today with yield $i_t$, then reinvest the proceeds in another one-period bond next period expected to yield $i^e_{t+1}$

For both the one-period bond and two-period bond to exist in equilibrium, investors must be indifferent between the two investment strategies.  This is a consequence of the bonds having identical risk and liquidity characteristics so that they are perfect subsitutes.  If one type of bond yielded a higher expected rate of the return, then there would be zero demand for the alternative bond.

The expected return to holding a two-period bond is 

$$
(1+i_{2t})(1+i_{2t})-1 = 2 i_{2t}+(i_{2t})^2
$$

The expected return to holding two one-period bonds is

$$
(1+i_t)(1+i^e_{t+1})-1 = i_t + i^e_{t+1}+i_t(i^e_{t+1})
$$

Both bonds will be held only if these expected returns are equal.  Approximately we have that

$$
2 i_{2t} = i_t + i^e_{t+1} \quad \text{or} \quad i_{2t} = \frac{i_t + i^e_{t+1}}{2}
$$

We have shown that the two-period rate must equal the average of the two one-period rates.  Of course we could generalize this to an $n$-period bond and say that

$$
i_{nt} = \frac{i_t+i^e_{t+1}+i^e_{t+2}+...+i^e_{t+(n-1)}}{n}
$$

The shape of the yield curve contains valuable information about how investors expect short-term interest rates to evolve. When the yield curve slopes upward, it suggests that markets anticipate short-term interest rates will rise in the future. Conversely, an inverted yield curve—one that slopes downward—implies that investors expect short-term rates to decline over time. Historically, yield curve inversions have often preceded recessions, as they signal expectations of monetary easing or economic slowdown. 

The expectations theory provides a clear framework for understanding several regularities in bond markets. It explains why interest rates across different maturities tend to move together and why the yield curve typically steepens when short-term rates are low and flattens or inverts when they are high. However, it struggles to account for two important facts. First, yield curves are usually upward sloping, even when short-term rates are not expected to rise. Second, long-term interest rates are, on average, higher than short-term rates over long periods. These limitations motivate alternative theories that introduce risk and market segmentation.

### Segmented markets theory

The **segmented markets theory** offers a different perspective by assuming that bonds of different maturities are not substitutes. Instead, each maturity has its own market, with its interest rate determined independently by supply and demand within that maturity segment.

Investors are assumed to have strong preferences for certain maturities, often favoring short-term bonds because they involve less price volatility and interest-rate risk. As a result, the demand for short-term bonds is higher, driving up their prices and pushing down their yields relative to longer-term bonds. This maturity preference naturally produces a typically upward-sloping yield curve, since investors require higher yields to hold longer-term securities.

While the segmented markets view explains the usual upward slope of the yield curve, it fails to explain why interest rates of different maturities tend to move together or why the slope of the curve changes systematically with the level of short-term interest rates. For these reasons, the theory provides valuable intuition about market segmentation but cannot fully describe the dynamic behavior of yields over time.

### Liquidity premium theory

The **liquidity premium theory** blends the core ideas of the previous two approaches. It assumes that bonds of different maturities are substitutes, but not perfect substitutes. Investors generally prefer short-term bonds because they are more liquid and less exposed to price risk. Consequently, long-term bonds must offer a liquidity premium (also called a term premium) to compensate investors for the additional risk of holding them.

Formally, the yield on an $n$-period bond at time $t$ is expressed as:

$$
i_{nt} = \frac{i_t+i^e_{t+1}+i^e_{t+2}+...+i^e_{t+(n-1)}}{n} + l_{nt}
$$

where $l_{nt}>0$ is the liquidity (term) premium for the $n$-period bond at time $t$, which is positive and rises with the term to maturity of the bond.

Even if future short-term interest rates are expected to remain constant, a positive liquidity premium can still generate an upward-sloping yield curve. The steeper the yield curve, the greater either the expected rise in short-term rates or the magnitude of the liquidity premium.

The liquidity premium theory thus reconciles the strengths of its predecessors. It explains both the co-movement of short- and long-term interest rates and the usual upward slope of the yield curve. Moreover, it accounts for the empirical tendency of yield curves to steepen when short-term rates are low—reflecting optimism about future rate increases—and to flatten or invert when short-term rates are high, often signaling expectations of a slowdown.

| Shape of Yield Curve   | Interpretation                             |
| ---------------------- | ------------------------------------------ |
| Steeply upward sloping | Short-term rates expected to rise          |
| Mildly upward sloping  | Short-term rates expected to remain stable |
| Flat                   | Short-term rates expected to fall slightly |
| Inverted               | Short-term rates expected to fall sharply  |

Empirical evidence suggests that

- the spread between long-term rates and short-term rates contains information about predicting very short-term interest rate movements (over a few months) and very long-run interest rate movements, but little information about intermediate-term movements.
- the yield curve also has predictive power in regard to business cycles and inflation
- an upward sloping yield curve usually indicates loose monetary policy, and an inverted yield curve tight monetary policy

### A more rigorous treatment of expectations theory

In this section I extend the expectations theory to show how the holding period return is a function of a bond's yield following  <a href="https://www.jstor.org/stable/1833129" target="_blank">Campbell and Shiller (1987)</a>.

Consider a consol bond (infinite term bond) with coupon payment $C$ and price $P_t^L$.  The yield is defined as the interest rate which equates the price of the bond to its present discounted value of cash flows.  We write

$$
P_t^L = \sum_{j=1}^\infty \frac{C}{(1+R_{t}^L)^j}=\frac{C}{R_t^L}
$$

where $R_t^L$ is the yield on the consol bond.  The holding period yield on any coupon bond is defined by

$$
H_{t+1}=\frac{P_{t+1}-P_t+C}{P_t}
$$

Therefore, the holding period yield on the consol bond is
$$
H_{t+1}=\frac{(C/R_{t+1}^L)-(C/R_{t}^L)+C}{(C/R_{t}^L)}=\frac{(1/R_{t+1}^L)+1}{(1/R_{t}^L)}-1
$$

It will be convenient to take a first-order Taylor approximation around the steady state $R_{t}^L=R_{t+1}^L=R^L$.  The first-order approximation of the ratio $R_{t}^L/R_{t+1}^L$ is given by

$$
\begin{align*}
R_{t}^L/R_{t+1}^L &\approx \frac{R^L}{R^L}+\frac{1}{R^L}(R_{t}^L-R^L)-\frac{1}{R^L}(R_{t+1}^L-R^L) \\
&=1+\frac{R_{t}^L-R_{t+1}^L}{R^L}
\end{align*}
$$

Therefore, the approximate holding period yield si
$$
H_{t+1} \approx \frac{R_{t}^L-R_{t+1}^L}{R^L}+R_{t}^L
$$

Our no arbitrage condition equates the holding period yield to the short-term interest rate $R_t$

$$
\frac{R_{t}^L-R_{t+1}^L}{R^L}+R_{t}^L = R_t
$$

which can be rearranged to show the equilibrium long-term interest rate as a function of the short-term interest rate and expected long-term interest rate

$$
R_{t}^L=\frac{R_{t+1}^L}{1+R^L}+\frac{R^L}{1+R^L}R_t
$$

This difference equation can be solved forward to give the long-term rate as a function of expected short-term rates:

$$
R_{t}^L=\frac{R^L}{1+R^L}\sum_{j=0}^\infty \left(\frac{1}{1+R^L}\right)^j E_t R_{t+j}
$$

As we saw in the simple version, the long-rate is an average of future expected short rates.  Here, it is a weighted average where the weights are a function of the steady state interest rate $R^L$.  Indeed, these are weights because $\frac{R^L}{1+R^L}\sum_{j=0}^\infty \left(\frac{1}{1+R^L}\right)^j=1$.  This derivation has some extra insights though.  The persistence of temporary changes in short-term rates matter for the long-rate.  Suppose, for example, that short-term rates are governed by an AR(1) process $R_t=\rho R_{t-1}+\epsilon_t$.  We know that $E_t R_{t+j}=\rho^j R_t$ so we can write the expectations theory predicted long-term rate as

$$
R_{t}^L=\frac{R^L}{1+R^L}\sum_{j=0}^\infty \left(\frac{1}{1+R^L}\right)^j \rho^j R_t = \frac{R^L}{1+R^L-\rho}R_t
$$

when shocks to the short-term rate are completely transitory $(\rho=0)$, there is a very small effect on the long-rate.  When shocks are near permanent ($\rho \rightarrow 1$), however, there is a one-for-one response to the long-rate.

We can also derive predictions about the how the spread forecasts future changes in short-term interest rates.  We rewrite the approximate expectations theory to say

$$
R_t^L-R_t=\sum_{j=1}^\infty E_t \Delta R_{t+j}
$$

When the spread is high, short-term rates are expected to increase in the future.  When the spread is low, short-term rates are expected to fall.  Permanent changes changes in short-term rates have no effect on the spread.

Since it is well known that long rates are higher than predicted by the expectations theory, most empirical studies include a liquidity or risk premium.  We then have

$$
R_{t}^L=\frac{R^L}{1+R^L}\sum_{j=0}^\infty \left(\frac{1}{1+R^L}\right)^j E_t R_{t+j} + L
$$

where $L=\frac{R^L}{1+R^L} \sum_{j=0}^\infty \left(\frac{1}{1+R^L}\right)^j E_t l_{t+j}$ is the liquidity premium.
