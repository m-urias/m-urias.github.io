# A Simple Macro Model of Monetary Policy

We now develop a three-equation model-—commonly known as the IS–PC–MP model—-to describe short-run dynamics in the economy. This framework provides a compact yet powerful way to connect the conduct of monetary policy with real economic activity and inflation dynamics.  In this model, monetary policy influences the short-term real interest rate, which affects aggregate demand and output through investment and other spending channels. The model thus bridges the gap between <a href="olg-banking.html">our earlier OLG model of banking</a> (which explains how banks and reserves underpin interest rate determination) <a href="monetary-policy-systems.html">our model of the interbank market</a>, where the central bank sets the short-term nominal interest rate.

Here we use a three-equation model, known as the IS-PC-MP model, to describe the short-run dynamics of the economy.  We describe the effects of monetary policy as changes to the short-term real interest rate which in turn affects various components of real output.  This description provides a link between monetary policy conducted through the banking system as described in <a href="olg-banking.html">our OLG model of baking</a> and our <a href="monetary-policy-systems.html">model of the interbank market</a>.  In particular, we describe how changes in the short-term nominal interest rate, as described in our <a href="monetary-policy-systems.html">model of the interbank market</a>, translate into changing the short-term *real* interest rate, which in turn affects the real economy.  

Changes in nominal interest rates, when prices are slow to adjust, translate into changes in real interest rates that directly influence real activity. Moreover, inflation expectations feed back into actual inflation, which motivates the need for a monetary policy rule—-such as the Taylor rule-—to stabilize output and inflation over the business cycle.

We proceed by describing each of the three equations one at a time and then analyze the dynamic macreconomic system these three equations constitute. 

## The IS Curve

The channel through which monetary policy affects short-run output is the real interest rate. The central bank can alter overnight nominal interest rates through open market operations in the reserves market. These short-term nominal rates influence longer-term rates via the yield curve. Because prices are sticky in the short run, nominal rate changes translate into real rate changes-—so that $R_t \neq r=MPK$.

When real interest rates fall, investment becomes more attractive, stimulating aggregate demand and increasing output. Conversely, higher real rates discourage investment and slow the economy.

This **monetary policy transmission mechanism** is summarized by:

$$
\downarrow \text{interest rate} \Rightarrow \uparrow \text{investment} \Rightarrow \uparrow \text{output}
$$

Although monetary policy also affects other channels—such as consumption, credit spreads, and exchange rates—we focus here on the investment channel for clarity.

The **IS curve** represents combinations of real interest rates and output levels that are consistent with equilibrium in the goods market.  To build the IS curve, begin with the national accounting identity

$$
Y_t=C_t+I_t+G_t+EX_t-IM_t
$$

which states that the total output of the economy can be used for consumption, investment, government spending, and net exports. We specify “mini theories” for each expenditure component. We assume that, in the short run, consumption, government spending, exports, and imports are each a constant fraction of potential output $\bar{Y}$.

$$
\begin{align*}
C_t &= a_c \bar{Y}_t \\
G_t &= a_g \bar{Y}_t \\
EX_t &= a_{ex} \bar{Y}_t \\
IM_t &= a_{in} \bar{Y}_t \\
\end{align*}
$$

where $\bar{Y}_t$ is long-run output determined by fundamental factors—-technology, labor, and capital—-that lie outside our short-run model.

The theory of investment, however, is a bit different.  We assume that investment responds to the real cost of financing capital projects:

$$
I_t = a_i\bar{Y}_t+b(R_t-r)\bar{Y}_t
$$

where $R_t$ is the short-term real interest rate relative to the long-run real interest rate $r$. When current interest rate equals long-run interest rate $R=r$, investment is a constant fraction of long-run output $I_t = a_i \bar{Y}_t$. When $R>r$, financing investment projects is costly so investment decreases below it's long-run level. When $R<r$, financing investment projects is cheap so investment rises above it's long-run level.  Recall that the long-run real interest rate $r=MPK$ is, once again, determined by our long-run theories and therefore exogenous with respect to our short-run model.


Substituting our "mini theories" into the national accounting identity gives:

$$
\begin{align*}
\frac{Y_t}{\bar{Y}_t}&=\frac{C_t}{\bar{Y}_t}+\frac{I_t}{\bar{Y}_t}+\frac{G_t}{\bar{Y}_t}+\frac{EX_t}{\bar{Y}_t}-\frac{IM_t}{\bar{Y}_t} \\
\frac{Y_t}{\bar{Y}_t} &= a_c+a_i-b(R_t-r)+a_g+a_{ex}-a_{im} \\
\underbrace{\frac{Y_t-\bar{Y}_t}{\bar{Y}_t}}_{\tilde{Y}_t} &= \underbrace{ a_c+a_i+a_g+a_{ex}-a_{im}-1}_{a}-b(R_t-r)
\end{align*}
$$

```{admonition} Definition: IS Curve

:class: definition

Our **IS curve** shows combinations of the output gap and real interest rate consistent with goods market equilibrium:

$$
\tilde{Y}_t = a-b(R_t-r)
$$

where  $\tilde{Y}_t$ is the output gap, $a$ collects the exogenous components of expenditure, and $b$ measures how sensitive investment (and therefore output) is to changes in the real interest rate.In normal times $a=0$, but shocks to aggregate demand--such as fiscal stimulus $\uparrow a_g$ or export surges $\uparrow a_{ex}$-- cause $a \neq 0$.  Monetary policy is summarized by $R_t$.

```

<figure>
    <img src="_static/figs_smm/is_curve.png"
         alt="IS Curve"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    The IS curve gives an equilibrium relationship between the short-term real interest rate and the output gap.
    </figcaption>
</figure>


The IS curve accommodates two types of shocks: 
1. **Aggregate demand shocks**: Temporary changes in exogenous components of expenditure ($C,I,G,EX,IM$) cause shifts in the intercept $a$.  A positive shock ($a>0$) pushes output above potential ($\tilde{Y}_t>0$), while a negative shock ($a<0$) pushes output below potential ($\tilde{Y}_t<0$).  Graphically, these appear as shifts of the IS curve.
2. **Interest rate shocks**.  When the central bank alters $R_t$ away from it's long-run value, the economy moves *along* the IS curve.  A rate cut stimulates output and raises output $\tilde{Y}_t>0$m, while a rate hike reduces investment causing $\tilde{Y}_t<0$.

This framework allows us to trace the real effects of monetary policy, while clearly distinguishing them from other sources of demand-driven fluctuations.  The IS curve captures the aggregate demand channel of monetary policy. Yet, it is not itself a monetary relation-—it shows how spending responds to real rates, regardless of what determines those rates. To close the model, we need an additional relation describing how the central bank sets the nominal interest rate and how inflation expectations link the nominal and real rates.

## A Simple Monetary Policy

How do we determine the short-run real interest rate?  For now, we assume that the central bank can target whatever real interest rate it desires. The central bank intervenes in the market for reserves to change the nominal interest rate (recall our discussion in <a href="monetary-policy-systems.html">Monetary Policy Systems</a>.)

Changes in nominal interest rates influence the real interest rate because we assume that **inflation displays inertia, or stickiness, in the short run**

$$
R_t = i_t-\pi_t
$$

Formally, we model monetary policy by saying that $R_t$ is a **policy variable** determined with complete discretion by the central bank.  We will call this our **MP curve**, although at this stage it is a trivial horizontal line representing whatever value the central bank picks for $R_t$. Nevertheless, once the central bank chooses it's policy for the real interest rate, the IS curve tells us the short-run equilibrium output gap.

<figure>
    <img src="_static/figs_smm/is_mp_curve.png"
         alt="IS Curve"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    The MP curve represents the monetary authority choosing the real interest rate $R_t$.  Given $R_t$, the IS curve shows the equilibrium output gap.
    </figcaption>
</figure>

The figure above depicts a **long-run equilibrium** where the short-term real rate equal the long-run real rate, $R_t=\bar{r}$, and thus output is at potential, $\tilde{Y}_t=0$.

What distinguishes this short-run description of monetary policy from our long-run models is price rigidity. Prices do not adjust instantaneously to changes in the money supply or the nominal interest rate. Instead, there is a lag between monetary policy actions and their effect on prices.  Eventually, prices do adjust, returning the economy to its long-run equilibrium—-a process captured by the models we studied earlier. In the short run, however, this sluggish adjustment gives monetary policy a powerful influence over real activity.

## The Phillip's Curve

In 1958, economist A.W. Phillips published a landmark empirical paper that examined the relationship between unemployment and wage growth in the UK.  He discovered that periods of low unemployment were associated with rapid increases in wages, while periods of high unemployment corresponded to slow wage growth. Subsequent studies found similar negative relationships between unemployment and inflation in other countries. This empirical regularity became known as the *Phillip's Curve*.

The intuition behind the Phillips Curve is straightforward. When labor markets are tight—that is, when the unemployment rate is low—firms face difficulty hiring workers and therefore raise wages to attract and retain labor. These higher wage costs are passed on to consumers in the form of higher prices, generating inflation. In contrast, when unemployment is high and labor markets are slack, wage growth slows and inflation tends to ease. During the 1960s, this apparent trade-off between unemployment and inflation was interpreted as a potential policy tool: governments could, in principle, achieve lower unemployment by tolerating somewhat higher inflation.

However, in 1967, Milton Friedman and Edmund Phelps identified a crucial theoretical flaw in this interpretation. They argued that firms and workers care about real wages—not nominal ones. When people expect prices to rise, they negotiate higher nominal wages in advance to maintain their purchasing power. As a result, both wages and inflation increase one-for-one with expected inflation, in addition to responding to the tightness of the labor market. Friedman and Phelps also introduced the concept of a natural rate of unemployment—the rate consistent with perfectly flexible prices and competitive markets.

Consider the real wage rate

$$
w = W/P
$$

where $W$ is a nominal wage rate and $P$ is the price level.  Or in terms of growth rates

$$
g_w = g_W-\pi
$$

Expansionary monetary policy can temporarily raise output and lower unemployment, leading to higher nominal wage growth. But as workers and firms begin to anticipate inflation, they will increase nominal wages in line with prices to preserve real purchasing power. In the long run, the attempt to sustain unemployment below its natural rate will fail, as inflation expectations adjust upward. Each new round of expansionary policy would be met with higher inflation expectations, causing inflation to accelerate without producing lasting employment gains.  Friedman’s reasoning implied that policymakers could not permanently exploit a trade-off between inflation and unemployment. Attempts to push output above potential ($\tilde{Y}>0$) would lead only to accelerating inflation.

Friedman and Phelps formalized this insight with the **expectations-augmented Phillips Curve**:

$$
\pi_t = \pi_t^e - \omega(U_t-U_n)
$$

where $U_t$ is the actual unemployment rate, $U_n$ is the natural rate of unemployment, and $\pi_t^e$ is expected inflation.  In the long run, expected inflation tends to equal actual inflation--because you can't continually fool the public--$(\pi_t^e = \pi_t)$, eliminating the tradeoff.  Nevertheless, in the short run, unexpected inflation can temporarily reduce unemployment below its natural level.

Starting in the 1970s, the apparent relationship between unemployment and inflation broke down. Periods of low unemployment were accompanied by high and persistent inflation—stagflation—rather than stable trade-offs. This empirical evidence strongly supported the Friedman–Phelps framework, and both economists later received the Nobel Prize in recognition of their contributions. Modern discussions of this concept often refer to the **Non-Accelerating Inflation Rate of Unemployment (NAIRU)**—the unemployment rate consistent with constant inflation.

### Incorporating Price Shocks

In practice, inflation can also be influenced by supply-side shocks—such as changes in oil prices or import costs—that affect firms’ pricing decisions independently of the labor market. To account for this, the expectations-augmented Phillips Curve is often written as:

$$
\pi_t = \pi_t^e - \omega(U_t-U_n) + \rho
$$

where $\rho$ captures temporary inflationary pressures from price shocks.

### From Unemployment to Output

We can relate inflation to output by recalling that output depends on the number of employed workers:

$$
Y_t = A_t K_t^{1/3}((1-u_t)L_t)^{2/3}
$$

where $u_t$ is the unemployment rate so $(1-u_t)L_t$ is the number of employed workers in the economy.  Expressing deviations in output relative to potential, the Phillips Curve can equivalently be written in terms of the output gap $\tilde{Y}$.:

$$
\pi_t = \pi_t^e - \nu \tilde{Y}_t+ \rho
$$


This form of the Phillips curve highlights how inflation depends on both expected inflation and real activity. When output exceeds potential ($\tilde{Y}>0$), inflation tends to rise; when output is below potential ($\tilde{Y}<0$), inflation declines.

### Adaptive Expectations and Short-Run Dynamics

To introduce interesting dynamics, we assume that agents form adaptive expectations, meaning they base forecasts of future inflation on past observations. In the simplest case, expectations are purely backward-looking:

$$
\pi_t^e=\pi_{t-1}
$$

Substituting this into the Phillips Curve gives:

```{admonition} Definition: Phillip's Curve

:class: definition

$$
\pi_t = \pi_{t-1} - \nu \tilde{Y}_t+ \rho
$$

```

This is the specification used in our short-run macroeconomic model. Inflation today depends on last period’s inflation, the current output gap, and temporary supply shocks.  Later, we will explore alternative expectations mechanisms, including rational expectations, where agents use all available information to forecast inflation.


## Our Basic Macro Model

```{admonition} Simple Macro Model
:class: definition

Now we have a simple macro model with three equations:

$$
\begin{align*}
\text{IS Curve}: \quad \tilde{Y}_t &= a-b(R_t-r)\\
\text{MP Curve}: \quad R_t &= \bar{R}_t \\
\text{Phillip's Curve}: \quad \Delta \pi_t &= \nu \tilde{Y}_t+o
\end{align*}
$$

- The IS curve determines equilibrium output gap for a given interest rate
- The MP curve shows that monetary policy chooses the interest rate with complete discretion
- The PC shows the rate of change in inflation given the output gap

```

We can use this three equation model to investigate some macroeconomic events in history or predict how the economy responds to future shocks.  For example, let's consider what monetary policy under Paul Volcker did in the 1970s.  During the 1970s inflation in the US was very high (reaching over 12\% by 1980), and chairman Paul Volcker was in charge of bringing inflation down.  He implemented contractionary monetary policy:
- Contractionary monetary policy means $\uparrow R_t$
- Higher interest rates depress investment demand which generates a negative output gap $\tilde{Y}_t<0$
- Negative output gap represents low aggregate demand which causes inflation to decelerate $\Delta \pi_t <0$

<figure>
    <img src="_static/figs_smm/is_mp_pc_disinflation.png"
         alt="Disinflation under IS-MP-PC model"
         width="600"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    Contractionary monetary policy shifts the MP curve upward indicating a higher short-term interest rate.  High interest rates depress investment causing short-run output to fall below potential output.  The negative output gap puts downward pressure on prices through the Phillips curve.
    </figcaption>
</figure>

Once inflation reaches the central bank's target, $\bar{\pi}$, then monetary policy can lower the real interest rate back to its long-run level, pull the output gap back to zero, and stabilize inflation.  The impulse response functions for this monetary policy shock are displayed below.

## Monetary Policy Rules and Stabilization

Up to this point, our framework has described how output and inflation evolve in response to monetary policy. Yet, real-world central banks do not simply choose a one-time real interest rate and then passively observe the economy’s response. Instead, they continually update their policies in response to new information about prices, output, and inflation expectations.

Modern central banks operate under what is often referred to as a **dual mandate**: achieving maximum employment and price stability. These objectives can be represented by a **loss function**, which captures the trade-off between stabilizing output and keeping inflation near its target:

$$
\min \mathcal{L} = \tilde{Y}_t^2+\phi (\pi_t - \bar{\pi})^2
$$

where $\bar{\pi}$ is the central bank's inflation target and $\phi$ measures how much the central bank cares about deviations of inflation from its target relative to the output gap.  Below is a visualization of the loss function:

<figure>
    <img src="_static/figs_smm/loss.png"
         alt="Disinflation under IS-MP-PC model"
         width="600"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    Quadratic loss function of the central bank
    </figcaption>
</figure>

Clearly, loss is minimized (the bottom of the bucket above) when real output is at its long-run level $\tilde{Y}=0$ and inflation is at its target level $\bar{pi}=\pi$.  Deviations from this point create a policy trade-off between maintaining price stability and stabilizing the real economy. The curvature of the loss function represents the degree of this trade-off.  We can better visualize this trade-off by considering level curves (or contour plots) of the loss function--slices through the three-dimensional "loss surface" at different magnitudes of loss.

<figure>
    <img src="_static/figs_smm/loss_visual_1.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    Quadratic loss function of the central bank
    </figcaption>
</figure>

The central bank understands that inflation and output are linked through the Phillips Curve. Therefore, its optimization problem can be expressed as:

$$
\begin{align*}
\min\hspace{2mm} &\mathcal{L} = \tilde{Y}_t^2+\phi (\pi_t - \bar{\pi})^2 \\
\text{s.t.}& \quad \pi_t = \pi_{t-1}+\nu \tilde{Y}_t + o
\end{align*}
$$

<figure>
    <img src="_static/figs_smm/loss_visual_2.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    Quadratic loss function of the central bank
    </figcaption>
</figure>

When inflation expectations equal the target $(\pi_{t-1}=\bar{\pi})$, the central bank's optimal policy produces an output gap of zero.

<figure>
    <img src="_static/figs_smm/loss_visual_3.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    Quadratic loss function of the central bank
    </figcaption>
</figure>

If inflation expectations rise above target, the central bank optimally creates a negative output gap—-slowing economic activity to bring inflation down.

<figure>
    <img src="_static/figs_smm/loss_visual_4.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    Quadratic loss function of the central bank
    </figcaption>
</figure>

For any given current level of inflation expectations $\pi_{t-1}$ we can represent the optimal monetary policy response.  The collection of all these optimal responses traces out the MP curve.

<figure>
    <img src="_static/figs_smm/loss_visual_5.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    Quadratic loss function of the central bank
    </figcaption>
</figure>

When the central bank places greater weight on inflation stabilization ($\uparrow \phi$), the MP curve becomes steeper, reflecting a stronger willingness to create negative output gaps to fight inflation.

<figure>
    <img src="_static/figs_smm/loss_visual_6.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    Quadratic loss function of the central bank
    </figcaption>
</figure>

An economy in which the output gap has strong influences on inflation has a high $\nu$ which generates a steeper Phillip's curve and consequently a flatter MP curve.

Enough fiddling with graphs, let's just solve the central bank's problem mathematically. Substituting the Phillips constraint into the loss function yields:

$$
\min (\frac{\pi_{t}-\pi_{t-1}-o}{\nu})^2+\phi(\pi_{t}-\bar{\pi})^2
$$

Taking the first-order condition gives the optimal monetary policy rule:

$$
\tilde{Y}_t=-\nu \phi(\pi_t-\bar{\pi})
$$

We now have a more sophisticated **monetary policy rule** that represents the central bank's dual objective of stabilizing prices and encouraging the long-run trajectory of the economy given that there exists a short-run equilibrium relationship between inflation and the output gap coming through the Phillips curve. A flatter Phillips curve $\downarrow \nu$ implies that the central bank will fight inflation less aggressively in that it does not induce such a large negative output gap.  A central bank that cares more about inflation $\uparrow \phi$ will fight inflation aggressively in that it induces large negative output gaps whenever inflation is above target.

To make the model more realistic, assume that changes in the interest rate affect output with a one-period lag:

$$
\tilde{Y}_t = a-b(R_{t-1}-r)
$$

Empirically, investment and consumption respond to policy changes with a delay of one to two quarters. Combining this lagged IS curve with the Phillips and MP curves gives our short-run macro model:

$$
\begin{align}
\text{IS Curve}: \quad \tilde{Y}_t &= a-b(R_{t-1}-r)\\
\text{MP Curve}: \quad \tilde{Y}_t &= -\nu \phi(\pi_t-\bar{\pi}) \\
\text{Phillip's Curve}: \quad \Delta \pi_t &= \nu \tilde{Y}_t+o
\end{align}
$$

- The IS curve determines equilibrium output gap for a given interest rate
- The MP curve shows that monetary policy adjusts interest rates according to a strict rule based on achieving its dual mandate
- The PC shows the rate of change in inflation given the output gap

These three equations jointly determine short-run fluctuations in output, inflation, and the policy interest rate.

### Monetary Policy Rule in Interest Rates

Let's clarify how the central bank implements it's optimal monetary policy rule by adjusting interest rates. Combining the MP rule with the Phillips Curve gives:

$$
-\left(\frac{1}{\nu \phi}+1 \right)\tilde{Y}_t=\pi_{t-1}+o-\bar{\pi}
$$

which represents the solution to the central bank's problem.  For any observed inflation $\pi_{t-1}$, the central bank optimally chooses an interest rate in order to implement output gap $\tilde{Y}_t$.Substituting the solution to the central bank's problem into the IS curve gives

$$
R_{t-1}-r = \frac{a}{b}+\frac{\nu \phi}{b(\nu^2 \phi+1)}(\pi_{t-1}+o-\bar{\pi})
$$

This shows how the central bank adjusts the real interest rate in response to deviations of inflation from target.


For example, with $a=o=0$ and $\nu = \phi = b =1$:

$$
R_{t-1} - r = \frac{1}{2}(\pi_{t-1}-\bar{\pi})
$$

meaning that a 1\% increase in inflation above the target call for a 0.5\% increase in the real rate.


In practice, we observe nominal interest rates rather than real ones. Using the Fisher relation,

$$
i_t = \pi^e_t + R_t
$$

and assuming adaptive expectations $\pi_t^e=\pi_{t-1}$ we can write

$$
i_t = \pi_{t-1}+R_t
$$

Substituting the optimal policy rule gives:

$$
i_t = r+\frac{a}{b}+\pi_{t-1}+\frac{\nu \phi}{b(\nu^2 \phi+1)}(\pi_{t-1}+o-\bar{\pi})
$$

Under the simplifying assumptions $a=o=0$ and $\nu=\phi=b=1$ we have a simple prediction for the short term nominal interest rate given inflation data

$$
i_t = r + \pi_{t-1}+\frac{1}{2}(\pi_{t-1}-\bar{\pi})
$$

A 1% increase in inflation therefore requires a 1.5% rise in the nominal interest rate—-a policy response often described as “leaning against the wind.” This is obvious in that to manipulate real interest rates, the central bank has to adjust nominal rates by more than inflation.

<figure>
    <img src="_static/figs_smm/Taylor_2.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    Predicted Fed funds rate for the interest rate rule $i_t = r + \pi_{t-1}+\frac{1}{2}(\pi_{t-1}-\bar{\pi})$ where $r=2\%$ and $\bar{\pi}=2\%$. 
    </figcaption>
</figure>




A very popular monetary policy rule, that has been empirically very accurate for explaining central bank policy around the world, is the **Taylor rule**:

$$
R_t - r = \varphi_1(\pi_t-\bar{\pi})+\varphi_2\tilde{Y}_t
$$

where $\varphi_1, \varphi_2$ captures the objectives of the central bank for fighting inflation versus keeping output on it's long-run trajectory. Our model generates such a rule naturally when we allow for lags in the Phillips Curve—reflecting the fact that output affects inflation with a delay.

## Our IS-MP-PC Macro Model

We now have a complete framework linking output, inflation, and monetary policy. The model consists of three equations:

$$
\begin{align*}
\text{IS Curve}: \quad \tilde{Y}_t &= a-b(R_{t-1}-r)\\
\text{MP Curve}: \quad \tilde{Y}_t &= -\nu \phi(\pi_t-\bar{\pi}) \\
\text{Phillip's Curve}: \quad \pi_t &= \pi_{t-1}+\nu \tilde{Y}_t+o
\end{align*}
$$

The IS curve determines the equilibrium output gap for a given real interest rate. The MP curve captures how monetary policy adjusts interest rates in pursuit of its dual mandate—stabilizing inflation and output. The Phillips curve describes how inflation evolves over time, rising when output exceeds potential and falling when output is below potential.

<figure>
    <img src="_static/figs_smm/ISMPPC_1.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    </figcaption>
</figure>


### Event \#1: An Inflation Shock

Suppose that at time $t=0$, the economy experiences an increase in the price of oil or another cost shock that raises inflation. Formally, this is represented as a one-period shock $o>0$ so that

$$
\pi_0 = \pi_{-1}+\nu \tilde{Y}_0+o
$$

Graphically, this corresponds to an upward shift of the Phillip's curve which crosses the point $(\tilde{Y}_t,\pi_t)=(0,\pi_{-1}+o)$.

<figure>
    <img src="_static/figs_smm/ISMPPC_2.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    </figcaption>
</figure>


Initially the economy was at it's long-run equilibrium. The inflation shock raises current inflation by the magnitude of the shock. Observing $\pi_0>\bar{\pi}$, the central bank respondsby setting a interest rate $R_0$ in order to induce lower future output gap $\tilde{Y}_1$. The optimal response is represented graphically by the intersection of the MP curve with the new Phillip's curve 

<figure>
    <img src="_static/figs_smm/ISMPPC_3.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    </figcaption>
</figure>

As time passes, inflation remains above target--say $\pi_1 > \bar{\pi}$.  The new inflation rate generates a fresh Phillips curve,

$$
\pi_2 = \pi_1+\nu \tilde{Y}_1
$$

where expected inflation $\pi_2^e = \pi_1$ lower than before, but still above the long-run target.  Graphically, this curve crosses the point $(\tilde{Y}_1,\pi_2)=(0,\pi_1)$. The central bank again optimally raises rates to $R_1$--this time less aggressively-- resulting in a smaller but still negative output gap $\tilde{Y}_2$. This iterative process continues: inflation remains above target, the central bank maintains restrictive policy, and output gradually returns to potential.

<figure>
    <img src="_static/figs_smm/ISMPPC_4.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    </figcaption>
</figure>

<figure>
    <img src="_static/figs_smm/ISMPPC_5.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    </figcaption>
</figure>


### Event \# 2: An Aggregate Demand Shock

Next, suppose that at time $t=0$, aggregate demand rises sharply—-for example, due to a surge in exports from an overseas boom. This can be represented as a one-period positive demand shock
$a>0$:

$$
\tilde{Y}_t = a-b(R_{t-1}-r)
$$

Graphically, the IS curve shifts rightward.

<figure>
    <img src="_static/figs_smm/ISMPPC_6.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    </figcaption>
</figure>



At the existing interest rate $r$, the positive shock generates a positive output gap $\tilde{Y}_0>0$ and higher inflation $\pi_0>\bar{\pi}$. Observing inflation above target, $\pi_0 > \bar{\pi}$---which is graphically illustrated by a new Phillip's curve---the central bank increases the policy rate to dampen activity.  Knowing that the shock is temporary, it sets a real rate $R_0$ that returns output to potential once the IS curve shifts back next period. If, however, the shock persisted, the bank would need to raise rates further to contain inflation. (Note that if the aggregate demand shock lasted more than one period, then the IS curve would not shift back to its original position and this would force the central bank to set an even higher interest rate to induce it's optimal output gap).

<figure>
    <img src="_static/figs_smm/ISMPPC_7.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    </figcaption>
</figure>

Over subsequent periods, inflation remains above target and the Phillips curve shifts gradually downward as policy tightens. Output converges back toward potential, and the economy returns to its long-run equilibrium.

<figure>
    <img src="_static/figs_smm/ISMPPC_8.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    </figcaption>
</figure>


<figure>
    <img src="_static/figs_smm/ISMPPC_9.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    </figcaption>
</figure>


### Event \# 3: Lowering the Inflation Target

Now consider a change in the central bank’s long-run inflation target—an event akin to the Volcker disinflation of the early 1980s. Formally, this is a reduction in $\bar{\pi}$ in the MP curve,

$$
\tilde{Y}_t = -\nu \phi(\pi_t-\bar{\pi}')
$$

Graphically, the MP curve shifts downward, intersecting the new long-run inflation target $\bar{\pi}'$.  At the prevailing inflation rate, $\pi_0 >\bar{\pi}'$, so the central bank raises interest rates to induce a negative output gap and lower inflation.

<figure>
    <img src="_static/figs_smm/ISMPPC_10.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    </figcaption>
</figure>

As inflation falls, expectations begin to adjust: households and firms observe $\pi_1$ and revise their forecasts, $\pi_2^e=\pi_1$.  The Phillip's curve shifts downward, the central bank gradually eases policy, and the output gap closes. Eventually, the economy stabilizes at a new equilibrium with lower steady-state inflation.

<figure>
    <img src="_static/figs_smm/ISMPPC_11.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    </figcaption>
</figure>

<figure>
    <img src="_static/figs_smm/ISMPPC_12.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    </figcaption>
</figure>

### Inflation-Output Loops

Across these scenarios, a consistent dynamic emerges: the economy tends to trace counter-clockwise loops in inflation–output space. Following a positive demand shock, output rises first, inflation follows, and the central bank subsequently tightens policy. Conversely, after a negative demand shock, output and inflation fall together before policy turns expansionary. Inflation shocks trigger similar patterns through monetary tightening or easing. These loops capture the lagged and cyclical nature of stabilization policy.


<figure>
    <img src="_static/figs_smm/loops.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    </figcaption>
</figure>

### Financial Frictions and Risk Premiums

To capture financial instability, we can augment the model with a risk premium $f$ that drives a wedge between the policy rate and the lending rate faced by firms:

$$
R_t = R_t^{ff} + f
$$

Here $R_t^{ff}$ is the short-term real interest rate set by the central bank, while $R_t$ is the effective borrowing rate in financial markets. During the 2008–09 financial crisis, for example, even as the Federal Reserve cut its policy rate near zero, the rate faced by firms rose due to heightened risk premiums. Graphically, this can be interpreted as a contractionary shift in the IS curve:

$$
\tilde{Y}_t=a-b(R_{t-1}^{ff}+f-r)
$$

A rise in $f$ therefore acts like a negative demand shock, reducing output and inflation despite accommodative monetary policy.

During the financial crises in 2008 and 2009, financial market broke down due to solvency and liquidity reasons and business lending dried up. Even though the Federal Reserve cut interest rates down to near zero, the rate at which firms and households borrow to finance investment was rising instead of falling. We could represent this phenomena in our model as a rise in risk premiums $\uparrow f$ which causes real lending rates to rise.

<figure>
    <img src="_static/figs_smm/spread.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    </figcaption>
</figure>

### Expectations and Forward Guidance

Up to this point, we have assumed that expectations are adaptive—that is, agents forecast future inflation using the most recent observation:

$$
\pi_t^e = \pi_{t-1}
$$

However, this ignores the central bank’s announced target $\bar{\pi}$. 

Consider that the central bank's announced inflation target $\bar{\pi}$ does not appear in our equation of how agents form expectations.  A more realistic approach is to assume rational expectations, meaning that households and firms form forecasts based on all available information, including monetary policy commitments. Under rational expectations, the credibility of the central bank becomes crucial: its willingness to fight inflation directly influences inflation expectations.

We have the following model

$$
\begin{align*}
\text{IS Curve}: \quad \tilde{Y}_t &= a-b(R_{t-1}-r)\\
\text{MP Curve}: \quad \tilde{Y}_t &= -\nu \phi(\pi_t-\bar{\pi}) \\
\text{Phillip's Curve}: \quad \pi_t &= \pi_t^e+\nu \tilde{Y}_t+o \\
\text{Financial Frictions}: \quad R_t &= R_t^{ff} + f
\end{align*}
$$

Since actual inflation depends partly on expectations,, credible monetary policy can lower inflation expectations and reduce the short-run output costs of disinflation. If the central bank announces a credible plan to lower long-run inflation from 4% to 2%, then agents may immediately revise expectations to $\pi_t^e=2\%$. The Phillips curve shifts downward right away, and the economy reaches the new equilibrium with minimal output loss.
 
<figure>
    <img src="_static/figs_smm/ISMPPC_13.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    </figcaption>
</figure>

This highlights the importance of **forward guidance**—-clear communication about the central bank’s future actions—-and the idea of **constrained discretion**, where the central bank retains short-run flexibility while committing to a long-run target.

Reality is probably somewhere in between naive expectations ($\pi_t^e = \pi_{t-1}$) and rational expectations ($\pi_t^e = E[\pi_t \vert \text{all information}]$).  Our models generates two extreme predictions about the dynamics of the macroeconomy so that you can think about reality being some convex combination of those two extremes.  Academic papers which try to model such a reality may assume some form of **dynamic learning** where households and businesses observe a limited set of information and make sophisticated forecasts (like an econometrician would) based on their own model of the economy. 

### Advanced Macro Models

Modern macroeconomics builds on this framework by introducing **microfoundations**, persistence, and numerical methods. First, each component of the model—the IS curve, Phillips curve, and monetary policy rule—is derived from optimizing behavior under nominal rigidities. Second, shocks are modeled as persistent stochastic processes, such as

$$
a_t = \rho a_{t-1}+\epsilon_t
$$

where $\rho=0$ means the shock is temporary, $\rho=1$ means the shock is permanent, and $\rho \in (0,1)$ precisely measures the persistence of the shock. Finally, the model is solved computationally, yielding what is known as a **Dynamic Stochastic General Equilibrium (DSGE) model**.


Writing down such a model is called a **dynamic stochastic general equilibrium model (DSGE)** and is solved numerically. 

Conceptually you can think about combining every model we've studied so far (production function, capital accumulation, equilibrium unemployment, IS curve, Phillip's curve, monetary policy rule) introducing a 1 percent shock of your choice and numerically solving the trajectory of the economy over time.

<figure>
    <img src="_static/figs_smm/Smets_Wouters.png"
         alt="Disinflation under IS-MP-PC model"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    </figcaption>
</figure>

While quantitative estimation of such models is beyond our current scope, the qualitative insights remain central. The IS–MP–PC framework helps us understand short-run dynamics and stabilization policy, the model of reserves explains how the central bank implements interest rates, and the Solow model remains essential for long-run growth analysis.
