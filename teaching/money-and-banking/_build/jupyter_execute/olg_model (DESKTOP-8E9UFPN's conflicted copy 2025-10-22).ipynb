# Overlapping Generations Model of Money, Inflation, Assets

## OLG Model of Money

First, we motivate the need for currency in an economy.  Second, we develop a model with credit.  Third, we show how fiat money can overcome the frictions inherent in an economy with imperfect commitment.

### Wicksell's Triangle

Consider an economy with three people: A, B, and C (Ann, Bill, and Charlie); three time periods: 1, 2, 3 (morning, afternoon, night); and three time-dated goods: $c_1,c_2,c_3$.  Assume preferences and endowments such that everyone strictly prefers a good held by someone else:

$$
\begin{align*}
U_A &= \beta c_3 + c_1 \quad  \text{and} \quad (y_1,y_2,y_3)=(0,0,y) \\
U_B &= \beta c_1 + c_2 \quad  \text{and} \quad (y_1,y_2,y_3)=(y,0,0) \\
U_c &= \beta c_2 + c_3 \quad  \text{and} \quad (y_1,y_2,y_3)=(0,y,0) \\
\end{align*}
$$

where $\beta \in (0,1) $ is discount factor.  For example, Ann most prefers morning bread $c_1$, but also enjoys her own evening bread $c_3$ albeit only a fraction $\beta$ as much.

Although stylized, this structure highlights the trading frictions that give rise to some medium of exchange.  In particular, notice that there are no bilateral mutually beneficial trades.  That is, we cannot find any pair of agents where there is a double coincidence of wants.  If we pair Ann and Bill, for example, Ann wants the morning bread which Bill can produce, but Bill doesn't want the night bread that Ann produces.  Every pairwise combination yields the same problem. Without trade autarkic utility is $\beta y$: each person only eats the bread they can produce.  If, however, bread could be allocated to those who most value it, then each agent could realize utility $y$.  This discrepancy between Pareto utility--that which is socially optimal--and autarkic utility demonstrates that there exist multilateral gains from trade.  Given lack of double coincidence of wants, the problem is this: What trading arrangement achieves Pareto utility $y$?

The following figure proposes such a trading arrangement.  Charlie produces afternoon bread and gives it Bill.  Bill gives his morning bread to Ann.  And Ann gives her night bread to Charlie.  Everyone gets the good they most desire.  But notice that this trading arrangement is feasible only if Bill can *credibly promise* to engage in this multilateral exchange after receiving the afternoon bread from Charlie.  Is this a reasonable assumption?

Notice from the preferences that Bill still enjoys his own morning bread, albeit less than Charlie's afternoon bread.  After receiving afternoon bread from Charlie, why should Bill not simply take it and also eat his own morning bread?  We could imagine that when Charlie and Bill meet, Bill says something like, "Hey Charlie, you give me your afternoon bread now and I *promise* to deliver my morning bread to Ann, so you can get your night bread from her later."  But once Charlie hands over the bread to Bill, Bill may simply renege on his promise and enjoy his own bread too!  Charlie is no fool, he anticipates this and therefore refuses to trade with Bill in the first place thereby shutting down the entire multilateral trading arrangement.

There is no feasible multilateral arrangement if agents cannot credibly keep their promises.  Or using economic language, if there is **lack of commitment**.  Besides an interesting theoretical conundrum, this problem strikes as being a very realistic friction in real world interactions.  Why should we believe someone's promise if there is there is only benefit and no cost of them reneging on said promise.  In this example, we have not mentions any punishment to Bill, or any outside enforcement of the aggreement struck by the three parties.  In there absence, Bill should never keep is promise and everyone understands that; thereby destroying any chance of a multilateral trading arrangement.  This small economy is doomed to autarky.

Classical economic models implicitly assume that there exists **perfect commitment**: all agents make promises that are always kept.  One example of this assumption manifest are Arrow-Debreu securities.

An interesting corollary to this example is this: Lack of double coincidence of wants is not sufficient to generate need for money.  Many textbooks motivate the need for currency by citing lack of double coincidence of wants.  The story is often told that once there exists lack of double coincidence of wants then currency arrives to save the day.  But this story is too simple.  Lack of double coincidence of wants is necessary but not sufficient.  There also must be lack of commitment.  After all, Wickell's triangle provides an example where there is lack of double coincidence of wants, but if agents can extend each other "credit" (a promise to deliver a future trade), then the Pareto optimal allocation can be achieved.  If we forced there to exist only multilateral gains from trade, then it is possible that Charlie would accept morning bread from Bill even though he doesn't want it, because it knows that he can trade morning bread with Ann for night bread later.  In this case, morning bread circulates a medium of exchange!  Charlie values it not because he wants to eat it, but because he expects to be able to use it to execute a trade later.  

To understand the role that credit and liquid assets play in an economy, we must take seriously the reality of lack of commitment.  Our goal in developing the overlapping generations (OLG) model is to show how **fiat money** can improve economic allocations.  To give money a role to play, our model must acknowledge the trading frictions which prevent barter trades. We understand why fiat money is valuable if we understand the frictions it overcomes.

With this in mind, we proceed as follows: First, we develop an OLG economy without frictions to identify what is Pareto optimal allocation in our environment.  Second, we introduce trading frictions which prevent the economy from obtaining a good allocation.  Third, we introduce fiat currency and show that it valuable because it facilitate good allocations.

### An Economy with Record Keeping

We begin by specifying the environment which characterizes our **overlapping generations model** (OLG).  

**Agents and timing:** Time is infinite and discrete: $t=1, 2, 3, \ldots$.  The number $N_t$ of agents are born in period $t$ and live for two periods. We say a person born in period $t$ is young in period $t$ and old in period $t+1$.  Agents die at the end of old age.  Additionally, there exist some *initial old* $N_0$ in period 1, who live for just one period.

**Goods:** There is one nonstorable good; meaning that it perishes at the end of the period in which it was produced.  This good cannot be saved over time. Young agents work to produce $y$ units of the good, while the old cannot work.

**Preferences:** We specify the preferences over young and old consumption $(c_{1,t}, c_{2,t+1})$ in terms of a **utility function** $u(\cdot):\mathbb{R^+}\to \mathbb{R}$.  We assume that $u(\cdot)$ is increasing, differentiable, and concave.

Notice the fundamental problem that exists in this stylized environment: agents value consumption when old but are unable to produce any goods.  The only way that an old person can obtain consumption is if they trade with a young person who *can* produce goods.  But what can the old exchange with the young?  They have nothing.  Can the old *promise* to repay the young person later?  Certainly not: the old will be dead.  This economic environment delivers both lack of double coincidence of wants and lack of commitment.  Although highly stylized, it is a wonderful environment in which to explore the ramifications key frictions that give money a valuable role in the economy.

```{admonition} Definition: Feasible Allocations
:class: definition

An **allocation** is the consumption of young and old in each period:
	
$$ \{c_{1,t},c_{2,t} \}_{t=1}^{\infty} $$

An allocation is **feasible** if 

$$ \underbrace{N_t c_{1,t}+N_{t-1}c_{2,t}}_{\text{consumption in period $t$}} \leq \underbrace{N_t y}_{\text{output in period $t$}} $$

```
Notice that the description of our economy is *dynamic*.  We specify the entire path of consumption for both young and old agents into the infinite future.  The feasibility constraint simply states that the economy cannot eat more than it produces.  At time $t$ there exist $N_t$ young individuals (born this period) and $N_{t-1}$ old individuals (born last period).  Once we specify the amount each generation consumes $(c_{1,t},c_{2,t})$ we know aggregate consumption $C_{1,t}+C_{2,t}=N_t c_{1,t}+N_{t-1}c_{2,t}$.  This aggregate consumption can be no larger than the total output that the young generation produces $Y_t=N_t y$.

For simplicity let's assume a constant population growth rate $n=N_t/N_{t-1}$.  Then the feasibility constraint is given by

$$
c_{1,t}+\frac{c_{2,t}}{n} \leq y
$$

The figure below plots the feasibility set for a constant population growth rate $n$:

<figure>
    <img src="_static/figs_olg/feasible_set2.png"
         alt="Feasible allocations"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    The set of feasible allocations with a constant population growth rate $n$.
    </figcaption>
</figure>

Having defined which allocations are feasible, we now investigate which allocations are good.  That is, which allocation maximizes the welfare of society?  To tackle this problem, we introduce the construct of the **social planner**.  The social planner is a hypothetical agent whose goal is to maximize the welfare of society.  Clearly no such person exists in reality, but we appeal to this fiction in order have a well defined *best allocation*.  We need such a benchmark by which to judge the allocations that a decentralized economy can achieve.

The problem of the central planner who wants to maximize welfare is given by,

$$
\max_{c_1,c_2} u(c_1,c_2) \quad \text{s.t.}\quad c_1+\frac{c_2}{n}\leq y
$$ (central_planner_problem)

We refer to this allocation as the **golden rule allocation**: that which maximizes welfare. Graphically, the solution to the planner's problem is simple: it is the point of tangency between the feasible set line and agents' indifference curve.

<figure>
    <img src="_static/figs_olg/golden_rule2.png"
         alt="Golden rule allocation"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    Golden rule allocation.
    </figcaption>
</figure>

We denote this best outcome for the economy by $(c_1^*,c_2^*)$.  Can the economy achieve the golden rule allocation through **decentralized trades**?	If each agent acts in its own self interest, what sort of allocations can this economy generate?  To model decentralized trades, we consider **competitive equilibria**.  Competitive equilibria have three defining features:

1. Individuals make mutually beneficial trades
2. Individuals take prices as given
3. Markets clear

Before introducing trading frictions, let's describe an environment where agents can engage in trades without money.  Instead of perfect commitment, we assume there exists some record keeping technology.  The record is a ledger which records all the trades in the economy.  We assume, for now, that this record keeping technology is perfect and costless.  

With record keeping, consider the following trading arrangement:
1. A young person gives up goods to an old person in exchange for a record of the transfer.  The consumption given to an old person is denoted $\Phi_t$, so that the **budget constraint of the young** is

$$
c_{1,t}+\Phi_t\leq y
$$

2. When old, the person shows the record of transfer and acquires consumption goods by receiving a transfer $\Phi_{t+1}^R$.  The **budget constraint of the old** is

$$
c_{2,t+1}\leq \Phi_{t+1}^R
$$

In this proposed inter-generational trading arrangement, a young person pays effectively pays $\Phi_t$ to goods when young to obtain $\Phi_{t+1}^R$ goods when old.  We can define a new variable $x$ to denote the implicit rate of return on the young person's "purchase of a receipt":

$$
\Phi_t x_{t+1}=\Phi_{t+1}^R
$$

where $x_{t+1}$ is the rate of return.  With this variable in hand, we can define the **lifetime budget constraint** facing a young person by combining the two period budget constraints:

$$
c_{1,t}+\frac{c_{2,t+1}}{x_{t+1}}\leq y
$$

The lefthand side is the lifetime present value of consumption while the righthand side is the lifetime present value of income.  The figure below illustrates the consumption/savings decision of every agent:

<figure>
    <img src="_static/figs_olg/record_keeping.png"
         alt="Equilibrium with record keeping"
         width="250"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    Consumption/savings decision with record keeping.
    </figcaption>
</figure>

Notice that the rate of return $x_{t+1}$ serves as a discount rate when discounting future consumption.  It remains to find the equilibrium value for this rate of return.  In computing it, we will assume **stationarity** and **perfect foresight**.  Stationary equilibria are those which do not depend on time: $(c_{1,t},c_{2,t}) = (c_1,c_2)$.  Perfect foresight says that agents perceive all available information and there is no uncertainty.  

We appeal to market clearing to write

$$
\underbrace{N_{t-1}x_t\Phi_{t-1}}_{\text{demand for transfers}}=\underbrace{N_t(y-c_{1t})}_{\text{supply of transfers}}
$$

so that

$$
x_t=\frac{N_t(y-c_{1,t})}{N_{t-1}\Phi_{t-1}}
$$


Using $\Phi_{t}=y-c_{1,t}$ and stationarity $c_{1,t}=c_1 \forall t$, 

$$
x_t=\frac{N_t}{N_{t-1}}=n
$$

The equilibrium rate of return on gifts the old is equal to the population growth rate.  


We've now computed the equilibrium of this record-keeping economy.  Is this record-keeping equilibrium efficient? If the planner's problem coincides with the decentralized equilibrium, then the solution is said to be **efficient**.

The budget line in the decentralized economy 

$$
c_1 + \frac{c_2}{n} = y
$$

is identical to the planner's.


Perfect record keeping can implement the golden rule allocation!

Let's investigate in more depth why the allocation we computed is indeed an *equilibrium* outcome.  That is, is it in every agents own self interest to give gifts to the old in exchange for a receipt which entitled them to consumption when old?  Are there any profitable alternative actions agents can take?  Consider the implicit game and enforcement in our environment.  Young people at time $t$ take the actions of future young people as given.  They cannot persuade or compel them to take any particular action.  We imagine there exist two actions: transfer $\Phi=c_2^*$ to the old generation or don't. If the young person expects $\Phi^R=c_2^*$ as the action chosen by next period's young person, then the transfer when young is justified as a best response.  Therefore, record keeping can sustain punishment that makes $(c_1^*, c_2^*)$ into a Nash equilibrium



Thus far we have delivered the following results:
- The OLG environment temporally separates agents so that there is no double coincidence of wants.
- Record keeping enables punishment for a person who does not participate in the intergenerational transfer program
- Record keeping acts as a social accounting system, where trading history provides incentives to the young to trade with the old
- Record keeping allows the decentralized economy to achieve the golden rule allocation (best feasible allocation).

### Introducing Fiat Money

Having described an economy with record keeping, we now want to investigate how the economy behaves in its absence.  Suppose that record keeping is too costly (e.g. young people can fake trading with old people to obtain goods when they are old).   A young person at time $t$ can only trade with young people of the same generation or old people of the previous generation.  Trading with the young is pointless, since there is no double coincidence of wants, and the old are not around the following period.  In this case, autarky is the only possible equilibrium without any other policy or institution.  The young consume their entire endowment when young and nothing when old.  Clearly a bad outcome. Can fiat currency improve welfare? If so, what determines the the value of fiat currency?

```{admonition} Definition: Fiat Money

:class: definition

**Fiat money** has two defining characteristics.
- Inconvertibility: cannot convert fiat money into anything else.
- Instrinsic uselessness: never wanted for its own sake or used in a production process.

```
	
Can pieces of paper with no inherent worth improve the allocation of our economy? To begin, consider a fixed supply $M$ of fiat money which is uniformly distributed among the initial old so that each person begins with $M/N$ units of fiat money.

Because fiat money, by definition, is not backed by any real asset its value is entirely derived from its expected future value.  Fiat money has no value if it is known with certainty that it has no value at any future date $T$.  If agents know that it is certainly worthless at time $T$, then no one would be willing to hold it at time $T-1$.  Ergo, it has no value at time $T-1$.  Similarly, if fiat money has no value at time $T-1$ then no one is willing to hold it at time $T-2$.  By induction, we can see that fiat money will have no value in the present if there exists any finite date $T$ in the future where it certainly has no value.  In this case, the only equilibrium allocation is autarky.  Therefore, a priori we suppose that money has positive value in all future periods and this value is denoted by $\nu_t$.  It remains to be seen if there exists equilibria with $\nu_t>0$.  

It is worth recognizing now that if the value of one unit of fiat currency is $\nu_t$, then the **price** of one unit of good must be $p_t=1/\nu_t$.  Therefore, if we say that money has value we can equivalently say $\nu_t>0$ or $p_t <\infty$.  

When describing the equilibrium, it is convenient to work with a variable called **real money balances**.  Real money balance is simply the real value (measured in units of consumption good) of a given stock of fiat money.  Mathematically, it is defined as

$$
q_t\equiv \nu_t m_t
$$

We now specify the problem facing a young agent.  A young agent chooses how much to consume and how much to save (in fiat currency) to maximize lifetime utility.

$$
\begin{align*}
\max_{c_{1,t},c_{2,t+1}} & u(c_{1,t},c_{2,t+1}) \\
s.t. \quad &c_{1,t}+\nu_t m_t \leq y \\
&c_{2,t+1}\leq \nu_{t+1}m_t
\end{align*}
$$

The money market must clear every period to satisfy a competitive equilibrium

$$
\underbrace{\nu_t M_t}_{\text{suppy}}=\underbrace{N_t(y-c_{1,t})}_{\text{demand}} \quad \forall t
$$

or

$$
\nu_t=\frac{N_t(y-c_{1,t})}{M_t}
$$


```{admonition} Definition: Monetary Equilibrium

:class: definition

A **monetary equilibrium** is a consumption path $\{c_{1,t},c_{2,t} \}_{t=1}^{\infty}$ and sequence of nominal prices $\{p_t\}_{t=1}^{\infty}$ such that $p_t < \infty$ for all $t$ where agents are behaving optimally and markets clear.

```


```{admonition} Definition: Stationary Monetary Equilibrium

:class: definition

A **stationary** monetary equilibrium is a monetary equilibrium where the consumption path does not change over time $(c_{1,t},c_{2,t+1})=(c_1,c_2)$.

```

The **rate of return on money** measures how the value of money changes over time:

$$
\begin{align*}
\frac{\nu_{t+1}}{\nu_t}&=\frac{\frac{N_{t+1}(y-c_{1,t+1})}{M_{t+1}}}{\frac{N_t(y-c_{1,t})}{M_t}}\\
&=\frac{N_{t+1}/M_{t+1}}{N_t/M_t}\\
\end{align*}
$$

In a stationary equilibrium with a constant population and money supply

$$
\nu_{t+1}=\nu_t
$$

The **quantity theory of money** predicts that price level is exactly proportional to quantity of money.  Let's investigate whether or not our OLG model supports this claim or not.

The price level is

$$
p = \frac{1}{\nu}=\frac{M}{N(y-c_1)}
$$

Hence, price level is proportional to money stock.


In this environment, money is also **neutral**: consumption $(c_1, c_2)$ does not depend on $M$.


What role is fiat money playing in our economy?  Fiat currency allows people to trade for goods they desire despite an absence of double coincidence of wants. Young people sell some of their goods for fiat money and use this money to buy goods when they are old. Fiat money is intrinsically useless---it is not consumed and it does not produce anything that can be consumed---but has value because it allows agents to execute trades.  Fiat money serves a similar function as record keeping---according to Kocherlakota ``money is memory."

### Optimality of a monetary equilibrium

Having identified the monetary equilibrium of our economy, we can now ask to what extent it is a "good equilibrium."  That is, to what degree does it deliver maximum welfare to the citizens?  To measure this, we need only compare the allocation of the decentralized monetary economy to that of the central planner.  The central planner, after all, maximizes welfare by definition; thereby serving as a benchmark by which to judge the outcomes of decentralized equilibria.  We can simplify the problem even further.  Since the decentralized economy and the planner's problem share the same objective function---to maximize utility---we can guarantee the two solutions coincide if and only if the constraint of the optimization problems are the same.  Said plainly, does the feasibility line of the planner coincide with the budget line of the decentralized equilibrium?  If so, then the decentralized monetary economy achieves the Pareto optimal outcomes.  If not, then the planner could improve on the allocation.


Recall that the rate of return on money with a constant stock of fiat currency is

$$
\frac{\nu_{t+1}}{\nu_t}=\frac{N_{t+1}}{N_t}=n
$$

The agent's budget line is therefore

$$
c_1 + \frac{c_2}{n} \leq y
$$

which coincides with social planner! Hence the decentralized monetary economy implements golden rule!


What did we do?
- We gave the initial old worthless pieces of paper and generated an efficient outcome!
- In our model, fiat money improves the allocation of resources.

## Money, Inflation, and Assets

In our simple model thus far, we have shown how money is **essential**: it improves the allocation of the decentralized economy relative to its absence.  Now we want to investigate how the decentralized economy responds to changes in the supply of fiat money.  How are economic decisions altered?  Is there an optimal stock or growth rate of fiat money?  Can the government use fiat money to create revenue for itself?  How is inflation affected by changes in the supply of money?  Can money coexist with alternative assets?

This questions lay at the heart of perennial questions in macroeconomics related to inflation, interest rates, liquidity and central bank policy.  It is important to develop a rigorous model of the long-run relationships between these concepts before delving into their details later.

How do we measure money? Which assets can be called ``money”?

According to the Federal Reserve of the US the **money supply** is the total amount of money--cash, coins and balances in bank account--in circulation.  There are several different **monetary aggregates** which measure the money supply.  These monetary aggregates categorize difference assets based on their liquidity.
- Monetary aggregate M0 is currency held by the non-bank public.
- The **monetary base**, or **MB**, is the sum of M0 and reserve balances, or deposits held by banks and other depository institutions in their accounts at the Federal Reserve.  
- Monetary aggregate **M1** is the sum of M0 and deposits, and other checkable deposits (OCDs).  Deposits include balances held in checking accounts and other highly liquid deposits used for transactions.  It measures currency held outside of the U.S. Treasury, Federal Reserve Banks, and the vaults of depository institutions.  
- Monetary aggregate **M2** is the sum of M1 and small-denomination time deposits (less than \$100,000) and retail money market mutual funds.
- Monetary aggregate **M3** is the sum of M2 and large-denomination time deposits (greater than \$100,000), balances in institutional money market mutual funds, repos issued by depository institutions, and Eurodollars held by US residents at foreign branches of US banks.

Data on the monetary aggregates are released every month in the Federal Reserve's H.6statistical release, <a href="https://www.federalreserve.gov/releases/h6/current/default.htm" target="_blank">Money Stock Measures</a>. 


When introducing newly printed money into our OLG model, we will assume a constant money supply growth rate so that $M_t=zM_{t-1}$.  In principle, there can be any path for the money supply $\{M_t\}_{t=1}^{\infty}$ we like, as it is an exogenous variable.  But given that we want to characterize our equilibrium on pen and paper, we stick with the constant money supply growth rate $z$.  Therefore,  newly printed fiat money is given by the expression

$$
M_t-M_{t-1}=\left(1-\frac{1}{z}\right)M_t
$$

We further assume that new money is transferred lump-sum to each old person and worth $a_t$ units of the consumption good.  The aggregate amount of newly printed money therefore satisfies

$$
N_{t-1}a_t=\left(1-\frac{1}{z}\right)\nu_t M_t
$$

where the lefthand side is the total real value of money distributed to the old and the righthand side is the total real value of new printed money.

When agents make their consumption/savings decision, they realize the transfer from the government of newly printed money when old.  This is reflected in their old period budget constraint,

$$
c_{1,t}+\nu_t m_t\leq y \\
c_{2,t+1}\leq \nu_{t+1}m_t+\textcolor{blue}{a_{t+1}}
$$

The period budget constraints are combined to obtain the lifetime budget constraint:

$$
c_{1,t}+\frac{\nu_t}{\nu_{t+1}}c_{2,t+1}\leq y+\frac{\nu_t}{\nu_{t+1}}\textcolor{blue}{a_{t+1}}
$$

The figure below shows the decentralized allocation with money supply growth.  Notice that agents take the rate of return on money $\nu_{t+1}/\nu_t$ *as given*.  Any individuals decision about how much money to save has no effect its equilibrium rate of return.  An individual agent is simply too small relative to the entire money market.  So what we have found so far is the optimal consumption/savings decision for any given rate of return on fiat money.  It now remains to compute the *equilibrium rate of return* on money so we can find the *equilibrium allocation* of decentralized economy.

The rate of return of fiat money in stationary equilibrium with constant population is given by

$$
\frac{\nu_{t+1}}{\nu_t}=\frac{N_{t+1}(y-c_1)/M_{t+1}}{N_t(y-c_1)/M_t}=\frac{M_t}{M_{t+1}}=\frac{1}{z}
$$

Notice that the gross inflation rate equals money growth rate

$$
\frac{p_{t+1}}{p_t}=\frac{\nu_t}{\nu_{t+1}}=z
$$

- Government spending shocks and supply shocks can produce short-run inflation but not sustained inflation
- Example: spike in oil prices raises general goods prices; however, after the initial shock, prices are stable

Newly printed money lowers the value of of existing money, serving as a tax on money holders.

- Higher growth rate of money decreases savings
- Higher growth rate of money distorts the tradeoff agents see between consumption today and consumption tomorrow
- Inflation tax should incentivize agents to seek alternative assets...we will add this to the model in a bit

Intuitively, we know that if inflation effectively taxes agents (and there is no negative externality associated with fiat currency), then the equilibrium should be inefficient.

- Feasible set is still $c_1 + c_2 \leq y$ (printing money is costless), which differs from budget set
- Hence, equilibrium is inefficient
- Inflation distorts the intertemporal trade-off between young and old consumption

Welfare cost of inflation

- Inflation does not destroy any goods but induces a suboptimal combination of $(c_1, c_2)$
- Drop in value of fiat money also reduces value of money balances of initial old
- More generally, inflation induces people to consume fewer cash goods and more non-cash goods
- In a production economy, inflation can discourage production of cash goods and hence reduce output


Golden rule monetary policy in a growing economy:
Suppose population grows at rate $n$: $N_{t+1}=nN_t$

$$
\frac{\nu_{t+1}}{\nu_t}=\frac{N_{t+1}}{N_t}\frac{M_t}{M_{t+1}}=\frac{n}{z}
$$

Lifetime budget constraint

$$
c_1+\frac{z}{n}c_2\leq y+\frac{z}{n}a
$$

Feasibility constraint is $c_1+c_2/n\leq y$ so the decentralized equilibrium is inefficient.

Economists have argued that the central bank should maintain a constant price level in order to stabilize the economy.   

- Since $p_{t+1}/p_t=z/n$, fixing the price level requires $n=z$: expand money at the same rate as population growth
- Implies that budget set is $c_1+c_2\leq y+a$
- But feasibility constraint is still $c_1+c_2/n\leq y$

This is a poor policy prescription.

- With $z=n$ a person receives one more good tomorrow for each good is sacrifices today
- This is not the true state of the economy: the economy can offer $n$ goods to the old for each good sacrificed by the young
- The true tradeoff faced by the economy is not reflected in the individual's budget constraint causing a misallocation of consumption over time

The best policy is that which maximizes welfare and reaches the golden rule allocation.

- Golden rule is to maintain a fixed stock of fiat money regardless of population growth ($z=1$)
- With population growth, this implies \textit{deflation}
- Intuition: a growing economy offers $n$ goods to the old in exchange for each good sacrificed when  young, but this message is not incorporated into budgets at a constant price level






## Money Coexisting with Alternative Assets

Our model thus far only allows individuals to consume at a later time by holding fiat money.  Fiat money is the only store of value.  In reality, individuals hold many assets in their portfolio.  How can we replicate this obvious fact within our theoretical environment?  

We introduce one alternative asset available to agents: capital.  Capital differs from fiat money in that it is **productive** and must be **produced** (in contrast to land and natural resources)

The U.S. <a href="https://www.bea.gov/" target="_blank">Bureau of Economic Analysis (BEA)</a> categorizes capital stock into several broad types based on the nature of the assets. These include private fixed assets, government fixed assets, and consumer durable goods.  **Private fixed assets** consist of equipment and structures, including owner-occupied housing, located in the United States and owned by private business or nonprofit institutions.  Within private fixed assets, BEA further classifies capital stock by type, including equipment, structures, and intellectual property products. **Government fixed assets** include equipment and structures owned by federal, state, and local government entities, with coverage extending worldwide for national defense equipment and U.S. embassies.  **Consumer durable goods** are items purchased by households for non-business use with a life expectancy of at least three years, such as cars, household appliances, and furniture.

#### An OLG Model with Capital

To begin, let's assume that capital is the *only* asset in our OLG environment.  Because capital has productive uses, it allows the old to produce some amount of capital.  For each unit of capital saved when young $k_t$, it allows the old to produce $x k_t$ units of consumption good.  For simplicity, we assume that capital completely depreciates at the end of the period in which it is used.  We will relax this assumption later.

With capital available as an investment, the period budget constraints are

$$
\begin{align*}
&c_{1,t}+k_t \leq y \\
&c_{2,t+1}\leq x k_t
\end{align*}
$$

which are combined into the lifetime budget constraint

$$
c_{1,t}+\frac{c_{2,t+1}}{x}\leq y
$$

Can capital, or other assets such as debt, coexist alongside money in equilibrium?  There is a fundamental economic/financial principle called **rate of return dominance**.  Said simply, if two assets are perfect substitutes then they will be held if and only if their rates of return are equal.  In reality two assets are almost never *perfect* substitutes as they differ in terms of the risk or liquidity.  However,  if two assets were indeed perfect substitutes, then both are held in a portfolio only if they provide the same rate of return.  If one asset held a higher rate of return than another, then the individuals would simply allocate all of their wealth to the high-return asset.

In the environment we've described thus far, capital and fiat money are indeed perfect substitutes.  We have not introduced any randomness to our model, so they are both risk free assets.  We have not introduced any frictions (e.g. informational, search) which would render the assets with different liquidity.  In our case, both capital and fiat currency can coexist if and only if they provide the same rate of return,

$$
\nu_{t+1}/\nu_t=x
$$

This is a somewhat unsatisfactory result given that the rate of return on capital $x$ is exogenous.  That is, capital and fiat currency can only coexist if the modeler chooses the value of $x$ to exactly coincide with $\nu_{t+1}/\nu_t$.  Furthermore, any small perterbation in either rate of return would result in zero demand for the lower return asset.  

To generalize our model and allow more than a knife edge case of assets co-existing, we allow the rate of return on capital to be endogenous.  To do so, assume that output is produced using capital according to a production function $f(k)$.  Moreover, assume that this production function is concave so that the marginal product of capital $f'(k)$ is diminishing with respect to the stock of capital.

When both fiat currency and capital exist and potential savings vehicles, we know the equilibrium rate of return on money is $\nu_{t+1}/\nu_t=n/z$ and the equilibrium rate of return on capital is $f'(k^*)$ where $k^*$ is agents' choice of savings in the form of capital.  Since both assets are still perfect substitutes, they are both valued is and only if 

$$
f'(k^*)=\frac{n}{z}
$$

Since the population growth rate $n$ and the money supply growth rate $z$ are exogenous (chosen by the central bank or government according to criteria outside the scope of the present model), agents will choose to accumulate capital up until the point its marginal return is equal to the rate of return on fiat money.  

Consider the effects of a permanent increase in money growth from $z$ to $z'$. This policy lowers rate of return on fiat money from $n/z$ to $n/z'$ and induces people to hold capital instead of money.  They simply chase the higher return asset.  Of course as they accumulate more capital it's rate of return falls until it is, once again, equal the rate of return on money $f'(k^*)=\frac{n}{z'}$.  This theoretical substitution of capital for money in reaction to higher anticipated inflation is called the **Tobin effect**. A larger capital stock increases output in the following period

$$
GDP_t=N_ty+N_{t-1}f(k_{t-1})
$$

This suggests that the central bank/government can stimulate output by introducing inflation.  However, there are at least two issues with this policy
- Capital stock may not be at optimal level (rate of return may be too low)--higher output does not necessarily raise welfare
- Measuring the magnitude of the Tobin effect in real economies is difficult.  There exist endogeneity problems and possibly a non-linear relationship.
		
Nonetheless, this is one theoretical explanation for why deviations from the Friedman rule can be optimal (Altermatt, Wipf 2022).

### The Fisher Equation

In order to discuss how inflation and interest rates affect our real economy, it is useful to discuss the **Fisher equation**.  Named after economist Irving Fisher, this equation describes the relationship between nominal interest rates, real interest rates, and inflation.  All financial contracts are expressed in nominal terms--dollars, for example, in the US.  When a business want to borrow money from a bank, the loan contract specifies the interest payments in dollars.  However, we have already seen that the real value of those dollars can change over time when the money supply grows.  The Fisher equation corrects the nominal rate for inflation.  

Let $i_t$ denote gross nominal interest rate in period $t$ (in units of money), $r_t$ be gross real interest rate (in units of goods), and $p_t$ be the price of a good in fiat money.  The Fisher equation says,

$$
\begin{align*}
i_t&=r_tp_{t+1}/p_t\\
i_t&=r_t(1+\pi_t)\\
(i_t-1)&=(r_t-1) + \pi_t + (r_t-1)\pi_t\\
i_t-1&\approx r_t-1 + \pi_t
\end{align*}
$$

The last line follows from $r_{t}\pi_t$ is a very small value, so a useful approximation of the Fisher equation is that the nominal interest rate is equal to the real interest rate plus the inflation rate.


The Fisher equation can be understood as a no-arbitrage condition between nominal and real bonds.

Applying the Fisher equation to our OLG model we have that

$$
i_t-1 = r_t-1 + \frac{z}{n}-1
$$

since the equilibrium inflation rate in our model is $\pi_t=\frac{\nu_t}{\nu_{t+1}}-1=\frac{z}{n}-1$.  Often it is assumed that the real-interest rate is determined by forces outside the control of monetary policy.  This is an example of the classical dichotomy which posts that real variables (e.g. output, employment) and nominal variables (e.g. money supply, prices) are determined separately.  The real interest rate is thought to be determined by the level of technology and capital stock.  Consider a macro model like Solow where the equilibrium interest rate is equal to the marginal product of capital $r=f'(k)$.  The real rate of return on capital projects is independent of money supply or the price level.  If this assumption is true, then the Fisher equation implies that the nominal interest rate and inflation move one-for-one with each other.  That is, if the central bank increases the growth rate of the money supply inducing inflation, the nominal interest rate should rise by exactly the same magnitude as inflation.

The figure above shows that the nominal interest rate and inflation do not move one-for-one.  The different between the two--the real interest rate--varies substantially over time.  One popular explanation is that prices are sticky in the short-run.  This means that monetary policy does not immediately affect inflation, so changes in the nominal interest rate cause changes in the real interest rate.  Given enough time, inflation manifests in response to the policy and the real interest rate returns to its natural level.  There is a lot of evidence that this story carries water, but we do not yet have the tools in hand to incorporate nominal rigidities into our model.  Instead, we will work within the environment already described to posit an alternative explanation as to why nominal rates and inflation do not move one-for-one.

We continue to suppose that output is produced from capital according to an increasing and concave function $f(k)$ and households can either save in capital or fiat currency.  Rate of return equality demands that both assets must provide the same return in equilibrium.  If the central bank increases inflation via faster money growth, then households optimally respond by substituting capital for fiat money.  Inflation lowers the return to fiat money so households shift their portfolio of assets away from fiat money and towards capital.  But since capital is subject to diminishing marginal returns, this lowers the rate of return on capital--or the real interest rate. Nominal interest rate thus responds less than one-for-one to rise in anticipated inflation.

### Optimal allocation in our OLG economy

Having described the equilibrium of our economy, we now ask what is the optimal allocation?  In particular: How much capital *should* the economy accumulate?  Can the free market achieve the optimal allocation?

We continue to assume that capital and fiat money are both available to households, and output is produced according to $f(k)$. Recall that when discussing optimality, we are imagining a hypothetical social planner who is not subject to the frictions inherent in the decentralized marketplace, and instead can allocate resources however they like.  The only constraint is that they cannot conjure resources from nothing--*ex nihilo nihil fit*.  The planner is subject of a feasibility constraint:

$$
N_t c_{1,t}+N_{t-1}c_{2,t}+N_tk_t\leq N_ty+N_{t-1}f(k_{t-1})
$$

where total consumption and investment cannot exceed total output.  Aggregate consumption consists of those $N_t$ individuals born today consuming $c_{1,t}$ and $N_{t-1}$ old individuals born yesterday consuming $c_{2,t}$. Aggregate output is the endowment $y$ of each $N_t$ young born today plus the production of the $N_{t-1}$ old who accumulated $k_{t-1}$ units of capital yesterday.  Total investment is simply the amount of capital $k_t$ that $N_t$ young people choose to save today.  Notice that if we denote aggregate consumption by $C$, aggregate investment by $I$, and aggregate output by $Y$ we get a very familiar looking equation:

$$
C+I=Y
$$

This is our national income accounting identity!  The absence of government spending $G$ should not be surprising, since we have not including any such thing in our OLG environment.  Later we will include a government sector and recover the all too familiar macro identity.

To solve the planner's problem, we first divide the feasibility constraint by $N_t$, to express everything in per capita terms, and consider only stationary allocations:

$$
\begin{align*}
c_1+\frac{c_2}{n}+k\leq y + \frac{f(k)}{n}\\
c_1+\frac{c_2}{n}\leq y + \left[\frac{f(k)}{n}-k\right]
\end{align*}
$$

The right hand side represents output net of costs of investment.  The planner finds the optimal allocation in two steps: first maximize output net of investment, then maximize welfare by choosing the distribution of consumption among the young and old based on preferences.  Maximizing output net of investment immediately obtains the optimality condition $f'(k)=n$.  Capital should be accumulated until its marginal product is precisely equal to the population growth rate.  Intuitively, this is because the marginal return on each units of investment is $f'(k)/n$, the marginal product of capital deflated by the population growth rate, while the marginal cost of each unit of investment is the forgone consumption of one unit.

Since the real interest rate is defined as $f'(k)=r$, we simply need to check whether or not the real interest rate equals the population growth rate. If $f'(k)<n$ then there capital over-accumulation, while if $f'(k)>n$ there is capital under-accumulation.  In the decentralized economy there is nothing which guarantees that $f'(k)=n$, so it is almost surely inefficient.

What type of policies can correct this inefficiency.  First, a constant stock of fiat money prevents capital over-accumulation.  If the central bank chooses $z=1$, then the equilibrium features $f'(k)=n$ which is exactly the optimality condition.  So in this economy the optimal monetary policy is to choose a constant money supply.  Alternative policies, which we do not go into detail here include: Social security (taxing young to pay old people) reduces capital for saving and can prevent capital over-accumulation. Opposite transfer scheme (taxing old to pay young) encourages capital if there is under-accumulation

Although we've made progress in explaining why real-world households hold both fiat money and alternative assets, there is a deeper question we've yet to contend with: in reality, fiat money has a lower rate of return than capital or private debt. Why is fiat money still valued? We will address this question more seriously when discussing banks.
