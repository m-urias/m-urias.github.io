# OLG Model of Banking


We have considered environments in which assets were perfect substitutes for fiat money and found that when fiat money offers a lower rate of return than other assets, it is not valued in equilibrium. We now explore how fiat money may nonetheless be valued because of its liquidity—its ability to facilitate consumption quickly.  

Building on this, we will show why banks are valuable to the economy: they issue liquid liabilities and reconcile mismatches between the maturities of assets and the liquidity needs of agents. To do so, we modify our overlapping generations (OLG) framework to include the possibility of “early” and “late” consumption. This allows us to highlight how money provides liquidity services and how banks can replicate these services through the creation of inside money.

First, we introduce a three-period OLG environment to accommodate the notion of "early" versus "late" consumption where there exists a market for private debt.  Second, we show how liquidity constraints give rise to a role for fiat money.  Third, we combine both insights to study the coexistence of inside and outside money.  Lastly, we investigate how monetary policy interacts with the banking sector to affect the real economy.

## Long-Term Capital Investment and Private Debt

We begin once again  with the frictionless economy, now modified to incorporate "early" and "late" consumption.  Later we will introduce limited commitment so that fiat currency plays a role and yields a liquidity premium relative to other assets.

The OLG economy now features *three-period-lived* agents, corresponding to the young, middle-aged, and old. Each agent is endowed with $y$ units of the consumption good when young but receives nothing thereafter. A production technology allows young agents to transform $k$ units of output into $f(k)$ when old--thus capital requires two periods to mature.

### Private Debt and Collateral

Agents can borrow or lend in a market for private debt, denoted $a$, at interest rate $R$.  Debt is *collateralized* by capital,and we assume that capital is perfectly observable and fully seizable in the event of default.  Hence, lenders can verify borrowers’ holdings of capital and are willing to extend credit up to the collateral value.

At each date $t$, the young decide how much to consume and how much to save.  Savings can take the form of either bonds ($a_t>0$), yielding a one-period rate of return $R$, or capital, giving a two-period return of $f(k)$.  The middle-aged generation consumes out of bond returns and may borrow ($a{t}<0$) from the current young,while the old generation consumes the output from their capital and repays prior debt.  Borrowing and lending therefore occur across generations, providing motives for borrowing and lending even with identical agents.  

### The Consumer's Problem

Each individual maximizes lifetime utility:

$$
u(c_{1,t})+\beta u(c_{2,t+1})+\beta^2 u(c_{3,t+2})
$$

subject to the period budget constraints given by:

$$
\begin{align*}
&\text{young:} \quad &&c_{1,t} + a_{1,t} + k_t  = y_t \\
&\text{middle-aged:} \quad &&c_{2,t+1} + a_{2,t+1} = Ra_{1,t} \\
&\text{old:} \quad &&c_{3,t+2} \leq f(k_t) + Ra_{2,t+1}
\end{align*}
$$

where $a_{1,t}$ and $a_{2,t+1}$ denotes net bond-holdings when young and middle-aged, respectively.  We impose $a_{t+2}=0$ to rule out Ponzi schemes.  

At any date $t$, three overlapping generations coexist:

| Generation    | Life stage  | Variable                          | Budget constraint                  |
| ------------- | ----------- | --------------------------------- | ---------------------------------- |
| born at ($t$)   | young       | chooses $(a_{1,t}, k_t)$                | $c_{1,t} + a_{1,t} + k_t = y_t$        |
| born at ($t-1$) | middle-aged | has $a_{1,t-1}$ from last period and chooses $a_{2,t}$  | $c_{2,t} + a_{2,t} = R a_{1,t-1}$      |
| born at ($t-2$) | old         | consumes $(f(k_{t-2}) + R a_{2,t-1})$ | $c_{3,t} = f(k_{t-2}) + R a_{2,t-1}$ |

Combining the three constraints yields the lifetime budget constraint:

$$
c_1 + R^{-1}c_2 + R^{-2}c_3 \leq y-k+R^{-2}f(k)
$$

The agent first chooses $k$ to maximize lifetime wealth (the right-hand side), and then allocates consumption to maximize lifetime utility.  Maximizing wealth implies,

$$
f'(k^{*})=R^2
$$

Thus, in equilibrium, the marginal product of capital equals the square of the interest rate.  Optimal portfolio choice equates the two-period rate of return on capital with the two-period return on bonds.  Said differently, the equilibrium interest rate satisfies $R=\sqrt{f'(k)}$; which equates the one-period return on capital and bonds.  Lifetime wealth is therefore,

$$
W=y-k^*(R)+R^{-2}f(k^*(R))
$$

The choice of $k^*$ depends only on technology and the interest rate $R$, and not on the agent's utility function.  Utility maximization, taking lifetime wealth as given, yields the familiar Euler equations:

$$
u'(c_1)=\beta R u'(c_2) = \beta^2 R^2 u'(c_3)
$$

which states the marginal utility across every period is equalized.  The Euler equations, in combination with the wealth constraint, characterizes the desired consumption plan $(c_1^*,c_2^*,c_3^*)$.

Finally, market clearing requires,

$$
k^* + c_1^*+c_2^*+c_3^*=y+f(k^*)
$$

Total available output (endowment plus capital returns) must equal aggregate consumption and investment. This gives us one equation in one unknown: the equilibrium interest rate $R^*$.

### Interpretation and Walras' Law

In this equilibrium, “money” takes the form of private bonds backed by capital returns. Both bonds and capital earn the same equilibrium return $R^*=(f'(k^*(R^*)))^{1/2}$.  Let's investigate an interesting implication of Walras' law: market clearing in the goods market is equivalent to market clearing in the bond market. 

Summing across all agents’ budget constraints yields:

$$
c_{1,t}​+c_{2,t​}+c_{3,t}​+k_t​+a_{1,t}​+a_{2,t}​=y_t​+f(k_{t−2}​)+R(a_{1,t−1}​+a_{2,t−1​})
$$

Rearranging, so that the goods market appear on the left hand side and the bond market appears on the right hand side, we have

$$
c_{1,t}​+c_{2,t​}+c_{3,t}​+k_t-(y_t+f(k_{t−2}​))=R(a_{1,t−1}​+a_{2,t−1​})-(a_{1,t}+a_{2,t})
$$

Goods market clearing demands that the left hand side equal zero, which also implies that $R(a_{1,t−1}​+a_{2,t−1​})=(a_{1,t}+a_{2,t})$.  This condition says that net financial flows across generations must be zero: repayments of old debt plus new borrowing must perfectly balance the goods transactions between generations.  We can make this more intuitive by restricting our analysis to steady state where $a_{1,t}=a_{1,t-1}$ and $a_{2,t}=a_{2,t-1}$.  In this case the bond market clearing simplifies to $(1-R)(a_{1,t}+a_{2,t})=0$.  Since $R=1$ would be degenerate, the economically meaningful case is

$$
a_{1,t}+a_{2,t}=0
$$

This restores the intuitive bond-market clearing condition — total borrowing equals total lending within a period.

## Illiquid Capital and Fiat Money 

We now consider the opposite case of the previous model: suppose that the return to capital investment, $f(k)$ is *noncollateralizable*. In other words, the value of an individual’s capital cannot be verified or seized by others. This friction prevents the creation of private credit markets because borrowers could always default without consequence. As a result, the market for private debt collapses.

Formally, agents now face the constraints $a_t,a_{t+1} \geq 0$.  Since no debt would be repaid in equilibrium, there is no incentive to hold or issue private debt, and thus $a_t=a_{t+1}=0$. In such an environment, fiat money acquires a role: it becomes the only asset capable of transferring purchasing power across time.

This setup highlights two distinct motives for holding assets. Fiat money provides liquidity, allowing consumption in the near term, while capital offers a higher rate of return but only after two periods. This captures a realistic feature of modern economies: money is valued not because it yields the highest return, but because it facilitates timely exchange and consumption. Our goal is to build a model in which both fiat money and capital are valued in equilibrium—even though capital dominates money in terms of long-run return.

### Portfolio Choice and the Role of Fiat Money

We continue to work with the three-period OLG model introduced earlier, now without private debt. Instead, there exists a constant nominal stock of fiat money, $M_0$, initially held by the middle-aged generation.

Each individual maximizes expected lifetime utility 

$$
u(c_{1,t})+\beta u(c_{2,t+1})+\beta^2 u(c_{3,t+2})
$$

subject to the budget constraints,

$$
\begin{align*}
&\text{young:} \quad &&c_{1,t} + q_t + k_t  = y_t \\
&\text{middle-aged:} \quad &&c_{2,t+1} + q_{t+1} = (\nu_{t+1}/\nu_t)q_t \\
&\text{old:} \quad &&c_{3,t+2} \leq f(k_t) + (\nu_{t+2}/\nu_{t+1})q_{t+1}
\end{align*}
$$

Here $q_t$ and $q_{t+1}$ denote real money balances when middle-aged and old, respectively, and $\nu_t=1/p_t$ is the value of fiat currency.  

The agent faces a tradeoff between two assets: capital is illiquid but yields a high return, whereas fiat money is liquid but yields a low return.  Individuals therefore sell some of their initial endowment for fiat money to finance early consumption. However, they will also hold capital in order to finance late consumption if and only if capital offers a superior rate of return:

$$
f'(y-c_1-q) > \nu_{t+2}/\nu_{t}
$$

The right-hand side represents the two-period gross real return on money.  To simplify the analysis, assume that production is linear

$$
f(k) = Xk, \quad X >n^2
$$

so that capital yields a constant two-period return $X$.  The budget constraints then become,

$$
\begin{align*}
&\text{young:} \quad &&c_{1,t} + q_t + k_t  = y_t \\
&\text{middle-aged:} \quad &&c_{2,t+1}  = (\nu_{t+1}/\nu_t)q_t \\
&\text{old:} \quad &&c_{3,t+2} \leq Xk_t
\end{align*}
$$

Combining these gives the lifetime budget constraint,

$$
c_{1,t}+\frac{\nu_t}{\nu_{t+1}}c_{2,t+1}+\frac{1}{X}c_{3,t+2}\leq y
$$

This expression shows that the real rate of transformation between consumption in different stages of life depends on both the rate of return on capital ($X$) and the evolution of the value of fiat money ($\nu_{t+1}/\nu_t$).

Our model admits a world where money and capital are not perfect substitutes.  Capital’s illiquidity prevents it from serving as a transaction medium, even though it yields a higher return. Fiat money, in contrast, is used more frequently despite its lower yield.    

To formalize this intuition, define **velocity** as the ratio of the value of transactions involving an asset to the total stock of that asset. Fiat money has velocity equal to one, since the entire stock of money is exchanged in each period. Capital, by contrast, has a velocity below one because it is traded less frequently. In a stationary equilibrium, the velocity of capital equals:

$$
\frac{N_tk_t}{N_tk_t+N_{t-1}k_{t-1}}=n/(n+1)<1
$$

where $n$ is the population growth rate. Thus, fiat money circulates more rapidly than capital, mirroring real-world differences in the liquidity of financial assets.  

## Banking in the OLG Economy

We now enrich the model by introducing a special type of institution which can commit to honouring debt contracts: a bank. Because banks can commit to repay, they are able to issue a new form of liquid asset--**inside money**.  Inside money refers to money created by the private sector, such as demand deposits or bank credit, which is backed by private loans. This contrasts with outside money, like fiat currency, which is issued by the government. We will see that the coexistence of inside and outside money depends on profitability and credibility conditions for banks.

### The Deposit Contract

Suppose a bank offers the following deposit contract. In period $t$, the bank borrows one unit of good from a young depositor in exchange for a promised rate of return $n$. The bank invests that single good into the creation of capital, which yields $X>n^2 after two periods. To honor its promise at $t+1$, the bank borrows from the next generation of young depositors, rolling over its liabilities one period at a time.   Depositors are willing to accept this contract if it offers at least as high a return as fiat currency, $n\geq n/z$, where $z$ is the growth rate of the money supply.  Banks find the arrangement profitable since they earn $x>n^2$ on their investment.  Consequently, inside money replaces outside money: deposits dominate currency as the preferred liquid asset.

Consumers thus divide their saving between direct investment in capital (to finance consumption when old) and deposits in banks (to finance middle-age consumption).  Deposits are used by the banking sector to finance investment projects.  Although abstract, this description captures some salient features of modern economies.  Most highly liquid assets are not held in physical cash, but rather as deposits with some financial intermediary.  Most transactions occur by digitally debiting or crediting deposit accounts.  Banks value these deposits because they can be used to finance loans to businesses which are capital projects that yield some future rate of return.  Banks indeed have an asset maturity mismatch on their balance sheets where their liabilities (deposits) are short-term and their assets (loans/investments in capital) are long-term.

Of course, if there exist profits in the business of banking then surely more banks will begin operations.  Absent any barriers, continued entry into the banking business erodes profits until banks offer a one-period interest rate $r=x\equiv X^{1/2}$. Hence, in competitive equilibrium, banks earn zero economic profit.

### Portfolio Choice with Banks

Each household now decides how much to save in bank deposits, $h_t$, and in capital, $k_t$. Deposits are liquid, providing returns after one period, while capital is illiquid, paying off after two periods. The individual’s problem is:

$$
\begin{align*}
c_{1,t}+h_t+k_t &\leq y \\
c_{2,t+1} \leq r h_t \\
c_{3,t+1} \leq X f(k_t)
\end{align*}
$$

An equilibrium consists of a path $\{c_{1,t},c_{2,t+1},c_{3,t+1},k_t\}_{t=1}^{\infty}$ and a deposit rate $r$ such that agents maximize utility and banks earn zero profits.

## Central Banking in the OLG Economy

We now introduce a **central bank** into the model for two reasons.  First, in the previous model fiat currency was not valued in equilibrium once banks existed.  Not only is this counterfactual, but it also implies that prices would have to be expressed in a different unit of account. Introducing a central bank that imposes **reserve requirements** can restore a role for fiat money.  Second, this framework allows us to analyze how policy instruments—money growth, reserve ratios, and lending facilities—affect the economy.

### The Banking Sector Balance Sheet

We continue with our three-period lived OLG environment with banks and illiquid capital, but add a reserve requirement.  Reserves, in practice, are vault cash and deposits held within Federal Reserve banks.  We suppose each bank is required to hold a fraction $\gamma$ of their deposits as reserves.  The central bank controls both the reserve ratio $\gamma$ and the growth rate of the money supply $z$.  Assume that capital pays return $X>(n/z)^2$ two periods after its creation. 

Letting total deposits in the banking system be $H$, the balance sheet of the banking sector is:

| **Assets**                      | **Liabilities**      |
|---------------------------------|----------------------|
| Reserves ($\gamma H$ )         | Deposits \( H \)     |
| Interest-bearing assets ( $(1-\gamma)H$ ) | Net worth \( 0 \) |
| **Total assets ( $H$)**        | **Total liabilities ( $H$ )** |

Banks hold the minimum required reserves $\gamma H$, investing the remainder $(1-\gamma)H$ in interest-bearing assets such as capital.

### Money Market Equilibrium

Only banks hold fiat currency, since deposits dominate currency in return. The demand for fiat money arises entirely from the reserve requirement.  Equilibrium in the money market is therefore,

$$
\nu_t M_t=\gamma N_t h_t
$$

where the right hand side is real money demand by banks to satisfy their reserve requirements and the left hand side is real money supply. Solving for the price level gives,

$$
p_t=\frac{M_t}{\gamma N_t h_t}
$$

A higher reserve requirement or larger deposit base raises demand for reserves, thereby increasing the value of money (i.e., lowering the price level). Thus, the familiar quantity theory relationship holds: a one-time increase in the money supply produces a proportional increase in the price level.

### Seigniorage

Since fiat money is not held by households, all newly printed currency accrues to the government as seigniorage revenue:

$$
\begin{align*}
S_t&=\nu_t(M_t-M_{t-1})\\
	&=\nu_t M_t\left(1-\frac{1}{z}\right)\\
	&=\gamma N_t h_t\left(1-\frac{1}{z}\right)
\end{align*}
$$

Seigniorage rises with the reserve requirement $\gamma$, the real stock of bank deposits $N_th_t$, and the rate of monetary expansion $z$.

### Equilibrium Interest Rate

A competitive equilibrium requires that banks earn zero profits.  Therefore, the equilibrium deposit rate is the weighted average of the returns on reserves and capital.  

$$
\begin{align*}
r&=\gamma\frac{n}{z}+(1-\gamma)x\\
&=x-\gamma\left[x-\frac{n}{z}\right]
\end{align*}
$$

where $x=X^{1/2}$ is the one-period return on capital. Increases in the reserve ratio or the money growth rate $z$ reduce return on deposits.

### Output and Money Multipliers

Aggregate output in this economy equals the sum of the endowment of the young and the payoffs from both direct and indirect capital investment:

$$
Y_t=N_ty+N_{t-2}Xk_{t-2}+N_{t-2}X(1-\gamma)h_{t-2}
$$

We define the monetary aggregate $M1$ as the total nominal stock deposits, whereas the stock of reserves is the **monetary base**.  Given reserve ratio $\gamma$ we have that

$$
M1_t=\frac{M_t}{\gamma}
$$

The term $1/\gamma$ is the familiar **money multiplier**.  The price level can then be expressed in terms of $M1$ as

$$
p_t=\frac{M_t}{\gamma N_th_t}=\frac{M1_t}{N_th_t}
$$

The quantity theory holds if $N_t h_t$ does not depend on rate of return. If demand for deposits, $N_th_t$ is affected by the rate of return on deposits, then the change in the price level depends on the tool used to change $M1$.  Specifically: a one-time increase in monetary base does not affect $r$; permanent increase in monetary base reduces $r$; and lower $\gamma$ increase $M1$ and raises $r$.

### Central Banking Lending Facilities

In practice, banks that fall short of their required reserves can obtain funds in several ways: by selling assets, borrowing from other banks in the **interbank market**, or borrowing directly from the central bank via a **lending facility**.  The purpose of central bank lending is to permit banks to meet reserve requirements without precipitously selling off interest-bearing assets.  

Let $\delta$ denote the fraction of a bank's reserves financed through central bank borrowing, and $\Gamma_t^B$ the total nominal amount of borrowed reserves.  Total reserves are then $M_t+\Gamma_t^B$ implying:

$$
\delta=\frac{\Gamma_t^B}{\Gamma_t^B+M_t}
$$

The balance sheet of the banking sector becomes,

| **Assets**                      | **Liabilities**      |
|---------------------------------|----------------------|
| Reserves ($\gamma H$ )         | Deposits \( H \)     |
| Interest-bearing assets ( $(1-\gamma)H$ ) | Loan from central bank ($\delta \gamma H$) |
| 								 | Net worth \( 0 \) |
| **Total assets ($\gamma \delta H+H$)**        | **Total liabilities ( $\gamma \delta H+H$ )** |

### Price Level and Borrowed Reserves

The market for reserves now clears when total reserve demand equals the sum of non-borrowed and borrowed reserves:

$$
\delta \gamma N_th_t+\nu_tM_t=\gamma N_th_t
$$

so that the equilibrium price is

$$
p_t=\frac{M_t}{\gamma(1-\delta)N_th_t}
$$

More central bank lending increases the supply of reserves, lowering the value of money (raising the price level). Effectively, lending reduces the “effective” reserve ratio from $\gamma$ to $\gamma(1-\delta)$, allowing greater deposit creation.  

The market for reserves clears when

$$
\gamma M1_t=\delta \gamma M1_t+M_t
$$

so that the money multiplier now becomes

$$
M1_t=\frac{M_t}{\gamma(1-\delta)}
$$

By requiring banks to hold less vault cash and supplying borrowed reserves, it allows banks to issue more deposits per unit of reserves.  Therefore, the money multiplier increases to $1/(\gamma(1-\delta))$.  

### Endogenous Borrowing and the Discount Rate

In reality, the central bank does not fix $\delta$ directly.  Instead, it chooses the interest rate on reserve loans (the **discount rate**), call it $\psi$, and allows banks to decide how much to borrow.  We can therefore write the equilibrium deposit rate as,

$$
r=\gamma\frac{n}{z}+[1-\gamma(1-\delta)]x-\psi \delta \gamma
$$

But notice that we have knife-edge equilibria with respect to borrowed reserves: if $\psi>x$, then no bank borrows; if if $\psi<x$, then bank demand unbounded amount of borrowed reserves; and if $\psi=x$ then banks are indifferent between borrowing or not resulting in indeterminate amount of borrowed reserves and, by extension, the price level.  To obtain a well-defined solution, we can assume diminishing returns to capital: $f(k)$ where $f(0)=0, f'(0)=\infty$, and $f''(k)<0$.  In this case, banks borrow reserves up to the point where the marginal return on capital equals the cost of borrowing:

$$
f'(k)=\psi
$$

Since total interest-bearing assets are $k=\Gamma^B+(1-\gamma)H$, this condition determines equilibrium borrowing $\Gamma^B$ and, in turn, $\delta = \Gamma^B/(\Gamma^B+M)$.  Substituting these values gives the equilibrium interest rate and deposit holdings consistent with optimal behavior by both banks and households.
