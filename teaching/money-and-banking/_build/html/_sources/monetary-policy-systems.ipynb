# The Market for Reserves and Monetary Policy Systems


## The Market for Reserves

On any given day some banks will have more reserves than they want, and some banks will have less reserves than they want. The **interbank money markets** are where banks borrow and lend reserves to each other using short-term financing arrangements. In the US the interbank money market is called the **Fed Funds Market** and the equilibrium interest rate that banks charge each other for reserves is called the **Fed Funds Rate**.

Banks demand reserves for two reasons:
1. Regulatory requirements: Many central banks require member commercial banks to hold a certain fraction of their liabilities in the form of reserves. This is called the **required reserve ratio**. In China it is usually around 12%.
2. Payments: Banks need to hold reserves to settle payment requests that come through the payments network. Banks must hold the required reserve amount stipulated by law, but will also choose to hold some **excess reserves** as insurance against unanticipated payment requests.

The central bank can manipulate overnight interest rates by intervening in the market for bank reserves. Different monetary policy tools have different affects on the market for reserves, and we will discuss each after providing a theoretical description of the market for reserves.  

First, let's describe the demand function for bank reserves.  The quantity of reserves demanded by banks equals the sum of required reserves and excess reserves.  There is an opportunity cost of holding excess reserves: the interest rate that could be earned by lending them out (fed funds rate) less the interest earned on excess reserves ($i_{or}$).  When the Fed funds rate is high, it very costly to hold excess reserves because those reserves could have been lent out to earn a high rate of return.  When the Fed funds rate is low, it is not so costly for banks to hold excess reserves.  Therefore, there is a **negative relationship between the total demand for bank
reserves and the overnight lending rate**.

Second, let's describe the supply of bank reserves.  The central bank is a monopoly issuer of reserves.  The total supply of reserves equals the sum of nonborrowed reserves (supplied through open market operations) and borrowed reserves (supplied through the discount window). We will denote the cost of borrowing from the Fed is the discount rate $i_d$.  If $i_{ff} < i_d$ then banks will not borrow from the Fed and thus borrowed reserves will be zero. If, on the other hand, $i_{ff} > i_d$ then banks will want to keep borrowing from the fed and lending out the proceeds.

Hence, we have a vertical supply curve for reserves--representing the monopoly control of reserves in the banking system by the central bank--and a downward sloping demand curve for reserves.  This demand curve has a horizontal portion where the fed funds rate is equal to the discount rate--indicating the point at which there would be infinite borrowing from the Fed--and another horizontal protion at the interest on excess reserves rate--where the opportunity cost of holding reserves vanishes.

### Open market operations

Open market operations affect the quantity of non-borrowed reserves in the banking system
- an open market purchase adds non-borrowed reserves
- an open market sale subtracts non-borrowed reserves

The interest paid on reserves $i_{or}$ sets a floor for the federal funds rate

Open market operations fall into two broad categories:
- Dynamic open market operations: intended to change the level of reserves and the monetary base.  These are conduct through the outright security purchases or sales.
- Defensive open market operations (sterilization): intended to offset movements in other factors that affect reserves and the monetary base.  These are conducted through repos and reverse-repos

### Recent developments in the US

Before 2008, the Federal Reserve operated monetary policy under a \textbf{corridor system}: there was an upper bound and lower bound on where short-term interest rates could go. The upper bound was the discount rate.  If short-term market rates were to rise about the upper bound, banks could borrow at the Fed and lend at the higher market interest rate.  Arbitrage would force these rates down to the borrowing rate at the Fed.  The lower bound was the interest on excess reserves (zero).  The lower bound is binding since banks would rather earn 0 percent holding reserves than a negative return from lending..

In November 2008, the Fed raised interest on excess reserves (IOER) rate above short-term market interest rates and triggered a shift to a \textbf{floor system}. A floor system removes the opportunity cost to banks of holding excess reserves by offering the banks a deposit rate at the Fed---the IOER rate---that is equal to or abvoe short-term market interest rates. If short-term rates fall below IOER rate, then banks could borrow at those lower rates and park the funds at the Fed.  Arbitrage, then, should cause short-term rates to converge to the IOER rate.  In theory, this allows the Fed to use it's balance sheet as a tool of monetary policy while still maintaining control over short-term interest rates.

Under a floor system, monetary policy is not set by a market interest rate, but by an administrative interest rate: the IOER rate.  Monetary policy is not tied to the supply of reserves, but instead to the quantity of reserves demanded by banks which the central bank controls through changes to the IOER. The central bank sets the IOER rate higher enough that banks' demand for reserves becomes perfectly elastic with respect to inter-bank interest rates. Changes in the IOER rate, rather than open-market operations, is the main instrument of monetary policy

Prior to the financial crisis the Federal Reserve did not pay interest on excess reserves (the foor on the Fed funds rate was at 0), but in October 2008 the Fed decided to pay a positive interest rate on excess reserves thereby raising the floor. In September 2013 the Fed began a program of taking in money from a wide range of non-bank financial institutions and paying interest. The technical name for this programme is the Overnight Reverse
Repurchase Agreement Facility (ON RRP). Now the ON RRP rate is the effective floor.  The Fed says, "*Any counterparty that can use the ON RRP facility should be unwilling to invest funds overnight with another
counterparty at a rate below the ON RRP rate, just as any depository institution eligible to earn interest on reserves should be unwilling to invest funds overnight with another counterparty at a rate below the
interest rate on excess reserves.*"

Why did the US central bank move to a floor system during the Financial crisis? According to Ben Bernanke, "*...the risk that our emergency lending, which had the side effect of increasing bank reserves, would lead short-term interest rates to fall below our federal funds target and thereby cause us to lose control of monetary policy...Until this point we had been selling Treasury securities we owned to offset the effect of our [emergency] lending on reserves (the process called sterilization). But as our lending increased, that stopgap response would at some point no longer be possible because we would run out of Treasuries to sell.*"

Some critics worry that the floor system undermines the monetary transmission mechanism and alters bank portfolios away from positive NPV capital projects.

Here is a summary of the ample reserves system currently in place in the US.

### China's monetary policy system

China's implementation of monetary policy works more like the US prior to the financial crisis. There are scarce reserves. The central bank's short-term lending facility provides a ceiling on the
overnight interest rate. Interest on excess reserves provides a floor on the overnight interest rate. The equilibrium interest rate is determined by adjusting the supply of reserves through the repo market
In addition China makes use of it's medium term lending facility (MLF) to adjust 1-year rates.

### A simple model of central bank policy under floor and corridor systems

We now build a simple model to demonstrate how the central banks' operating system interacts with the market for loans.  We assume that the markets for reserves and bank loans are both competitive.  Since this makes banks price takers, the key mechanism linking the markets for reserves and loans is that banks will optimally banks equate the margins on bank reserves and bank loans.

We continue to use the theoretical model for bank reserves as described above.  The market for loans is described as follows:

- let $S^*_L$ denote the loan supply schedule based on bank fundamentals like funding costs, capital levels, regulatory requirements, and bank culture.
- let $S^0_L$ denote the loan supply schedule reflect the actual supply of loans taking into account the riskiness of borrowers
- the risk premium creates a wedge between $S_L^*$ and $S^0_L$
- the supply curves are upward sloping for the usual reasons: as the marginal return on loans increases, banks are willing to increase the quantity of loans supplied
- the demand for bank loans is downward sloping for the usual reasons: as the cost of financing bank loans falls, the quantity of bank loans demanded will rise.

- the intersection of the ``fundamental" supply curve $S^*_L$ and the demand curve for loans determines the risk-free short-run interest rate on loans, $i_{rf}$.
- the intersection of the loan supply curve and the demand curve for loans determines the observable loan interest rate $i_L$
- the spread between $i_L$ and $i_{rf}$ can be viewed as the net interest margin.

Equilibrium in both markets obtain where the overnight interest rate $i_{ST}$ is equal to the risk-free short-term interest rate $i_{rf}$. Said differently, the marginal risk-free return on loans is equal to the marginal cost of funding in equilibrium.

#### A positive shock to loan demand

Let's use our model to investigate what happnes when there is a sudden surge in demand for bank loans.  After the positive loan demand shock banks realize that higher demand for loans increases the marginal risk-free return on loans. Assuming constant risk premiums, the loan interest rate will proportionally rise from $i_L$ to $i_{L}^1$. The marginal risk-free return on loans exceeds the marginal cost of funding these loans ($i_{rf}^1 > i^_{ST}$)so banks respond by increasing the quantity of loans supplied from $L_0$ to $L_1$.

As a consequence of the higher equilibrium quantity of loans there is higher demand for bank reserves to meet the now-higher required (and precautionary) reserve amounts. This shifts the demand curve for reserves to the right until the marginal cost of funding loans just equals the marginal risk-free return ($i_{st}^1 = i_1$)

Note that the quantity of reserves in the banking system is unchanged, but the composition of reserves does change.  As the quantity of loans supplied increases, excess reserves get turned into required reserves.

#### Loosening monetary policy in a corridor system

Here we consider the central bank conducts an open market purchase to increase the supply the bank reserves. The excess liquidity causes a decline in the interbank interest rate from $i^1_{ST}$ to $i^2_{ST} = i^0_{ST}$

After the central bank lowers the overnight interest rate banks find that their marginal risk-free return on loans is higher than the marginal funding cost $(i^*_1 > i^2_{ST}$) and banks increase the supply of loans until the marginal conditions are again equalized.

#### Modeling the Fed's move to a floor system

- Before the financial crisis the Federal Reserve of the US set a target rate for $i^0_{ST}$ (according to a Taylor rule) 
- To achieve the target rate the Fed endogenously adjusted the supply of reserves to changes in demand for reserves
- The market for bank loans drove activity in the market for bank reserves, and the Fed responded to offset shifts in reserve demand to achieve its target rate

The Fed moved to a floor system in November 2008:

- the recession caused a sharp decline in loan demand causing a leftward shift from $D^1_L$ to $D^2_L$
- the marginal risk-free return on loans fell from $i_1^*$ to $i^*_2$ and the loan interest rate fell from $i^1_L$ to $i^2_L$
- the marginal cost of funding loans exceeded the marginal risk-free return ($i_{ST} > i_2^*$) and banks responded by reducing the quantity of loans supplied from $L_1$ to $L_2$
- now there is less demand for bank reserves causing $i_{ST}$ to start declining

Some critiques of the floor system:

- Setting the IOER rate above other short-term interest rates may include setting the IOER above the natural interest rate and inadvertently tightening the stance of monetary policy
- The floor system may influence banks' portfolios toward reserves and away from other assets
- The IOER rate spread effectively creates a quasi-liquidity trap above the zero lower bound

{Costs of a floor system}
Banks may reduce lending given a positive return on excess reserves:

- once $IOER_2 > i_{ST} = i^*_2$) the marginal cost of funding is higher than the risk-free marginal return on bank lending
- the production cost of loans have risen, and this incentivizes banks to reduce the supply loans they provide
- the loan supply curve shifts left until the marginal cost of funding equals the risk-free marginal return on loans ($IOER^2 = i_3^*$)
- the equilibrium quantity of loans fall from $L_2$ to $L_3$
