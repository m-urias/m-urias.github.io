$$
Y_t = Y_t^{\text{trend}}+Y_t^{\text{cycle}}
$$

As economists we distinguish between the long-run and short-run in the following way:
- Long-run GDP, called **potential GDP**, represents the level of production at which resources are being fully and sustainably utilized.
- Short-run GDP, called **actual GDP**, is the realized level of output taking into account any shocks that are hitting the economy.

When the economy is in a long-run equilibrium

$$
\text{Actual GDP}=\text{Potential GDP}
$$

When the economy is pushed off it's long-run equilibrium by shocks

$$
\text{Actual GDP} \neq \text{Potential GDP}
$$

Our theory of long-run GDP comes from theories of economic growth such as the Solow model. Our theory of long-run inflation comes from the OLG model. Now we want to develop a theory to explain short-run fluctuations of the economy away from its long-run trajectory.

In our model, we will describe short-run output in terms of percentage deviations of actual output from potential output

$$
\tilde{Y}_t = \frac{Y_t-\bar{Y}_t}{\bar{Y}_t}
$$
