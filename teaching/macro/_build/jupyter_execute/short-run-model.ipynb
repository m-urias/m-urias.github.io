# A Short-Run Model of the Macroeconomy

## The IS Curve

The key channel through which short-run output will be affected in our model is through changes in the real interest rate:

- The central bank can intervene in the market for reserves to manipulate overnight nominal interest rates
- Changes in overnight nominal interest rates translate into changes to longer term nominal interest rates via the yield curve
- Changes in nominal interest rates translate into changes into real interest rates in the short-run because prices respond with a lag (this is why $r\neq MPK$ in the short-run)
- Changes in the real interest rate affect investment decisions and therefore aggregate output

Our simple \textbf{monetary policy transmission mechanism}:

$$
\downarrow \text{interest rate} \Rightarrow \uparrow \text{investment} \Rightarrow \uparrow \text{output}
$$

Of course there are other ways that monetary policy can affect the real economy, but we will restrict our attention to this channel.



The **IS curve **represents a short-run equilibrium in the goods market wherein higher interest rates are associated with lower levels of output due to less investment.


Now let's build the IS curve...

Begin with the national accounting identify

$$
Y_t=C_t+I_t+G_t+EX_t-IM_t
$$

which says that the total output of the economy can be used for consumption, investment, government spending, and net exports.


Now we want to develop a ''mini theory' for each component on national expenditure

We assume (for now) that consumption, government spending, exports and imports are all a constant fraction of long-run output
\begin{align*}
C_t &= a_c \bar{Y}_t \\
G_t &= a_g \bar{Y}_t \\
EX_t &= a_{ex} \bar{Y}_t \\
IM_t &= a_{in} \bar{Y}_t \\
\end{align*}
where $\bar{Y}_t$ denotes the level of long-run, or potential, output.  Remember that $\bar{Y}_t$ is determined outside of the short-run model: it is determined by Solow/Romer/Equilibrium Unemployment.

The theory of investment is a bit different.

$$
I_t = a_i\bar{Y}_t+b(R_t-r)\bar{Y}_t
$$

which says that short-run investment depends on the short-term real interest rate $R_t$ relative to the long-run real interest rate $r$

- If current interest rate equals long-run interest rate $R=r$ then investment is just a constant fraction of long-run output $I_t = a_i \bar{Y}_t$.
- If $R>r$ then financing investment projects is  costly so investment decreases below it's long-run level.
- If $R<r$ then financing investment projects is cheap so investment rises above it's long-run level. 
- Recall that the long-run real interest rate $r=MPK$ is, once again, determined by our long-run theories and therefore exogenous with respect to our short-run model.


Now we can substitute our ''mini theories" into the national accounting identity:
\begin{align*}
\frac{Y_t}{\bar{Y}_t}&=\frac{C_t}{\bar{Y}_t}+\frac{I_t}{\bar{Y}_t}+\frac{G_t}{\bar{Y}_t}+\frac{EX_t}{\bar{Y}_t}-\frac{IM_t}{\bar{Y}_t} \\
\frac{Y_t}{\bar{Y}_t} &= a_c+a_i-b(R_t-r)+a_g+a_{ex}-a_{im} \\
\underbrace{\frac{Y_t-\bar{Y}_t}{\bar{Y}_t}}_{\tilde{Y}_t} &= \underbrace{ a_c+a_i+a_g+a_{ex}-a_{im}-1}_{a}-b(R_t-r)
\end{align*}

Our IS curve is given by

$$
\tilde{Y}_t = a-b(R_t-r)
$$

- The lefthand side is the percentage deviation of current GDP from potential GDP
- The parameter $a$ includes all the exogenous components of expenditure.  Notice that $a=0$ in normal times.  Although certain \textbf{economic shocks} (e.g. fiscal stimulate $\uparrow a_g$) could cause $a \neq 0$.  
- The parameter $b$ measures how sensitive investment, and therefore short-run output, is to changes in the real interest rate
- Monetary policy is summarized by $R_t$

Two possible shocks to consider:

- Aggregate demand shocks: Any event which temporarily alters the exogenous component of $C,I,G,EX,IM$.  Positive shocks cause $a>0 \Rightarrow \tilde{Y}_t>0$ temporarily while negative shocks cause $a<0 \Rightarrow \tilde{Y}_t<0$.  Graphically, this will appear as \textit{shifts} of the IS curve.
- Interest rate shocks: Monetary policy decides to alter $R_t$ away from it's long-run value.  Lowering the interest rate stimulates investment and causes $\tilde{Y}_t>0$ temporarily, while raising the interest rate reduces investment causing $\tilde{Y}_t<0$.  Graphically, this will appear as \textit{movement along} the IS curve.

### A Digression on our Mini Theories: Consumption
How do economists think about households' decision to consume versus save?

- Households care about their satisfaction, called \textbf{utility}, over their entire lifetime
- Households forecast their future income in order to make consumption decisions today
- What matters for your consumption/savings choice is the expected present value of your lifetime income

Consider a simple model where an individual makes a decision about how much to consume when young versus when old.


An individuals lifetime utility is given by

$$
U = u(c_y)+\beta u(c_o)
$$

where $(c_y,c_o)$ the amount of consumption when young and old respectively and $\beta \in (0,1)$ is a subjective discount factor reflecting how patient this individual is.


Of course, the amount of consumption you can enjoy in each period of your life is constrained by income:
\begin{align*}
c_{\text{y}}+s_{\text{y}}=y_{\text{y}} \\
c_{\text{o}} = y_{\text{o}}+(1+R)s_{\text{y}}
\end{align*}
where $R$ is the market interest rate that you can earn on your savings.

The problem that an individual solves is
\begin{align*}
\max \hspace{2mm} &u(c_y)+\beta u(c_o) \\
\text{s.t.} \quad &\underbrace{c_y + \frac{c_o}{1+R}}_{\text{PV of consumption}}=\underbrace{y_y+\frac{y_o}{1+R}}_{\text{PV of income}}
\end{align*}

<div style="text-align: center;">
<img src="_static/figs_SR/consumption.png" 
     alt="GDP growth across countries" 
     width="400">
</div>

The optimal amount of consumption when young and old is given by

$$
u'(c_y)=\beta(1+R)u'(c_o)
$$

which is called the \textbf{Euler equation}: it says that an individual must be indifferent between consuming one more unit today versus saving that unit and consuming it in the future.  

- If $u'(c_y) > \beta(1+R)u'(c_o)$ then the marginal value of consuming when young is higher than the marginal value of saving and consuming later.  So the individual should consume more when young.
- If $u'(c_y) > \beta(1+R)u'(c_o)$ then the marginal value of saving income today, earning a rate of return, and consuming when old is higher than the marginal value of consuming when young.  So the individuals should save more and consume less when young.
- Only when $u'(c_y) = \beta(1+R)u'(c_o)$ is the individuals behaving optimally and therefore has no reason to alter it's consumption plan.


In general, a higher interest rate $R$ increases the marginal value of savings and therefore reduces consumption today.


However, consider the following utility function

$$
u(c) = \ln c
$$

With this utility function the Euler equation becomes

$$
\frac{c_o}{c_y}=\beta (1+R)
$$

And substituting this back into our lifetime budget constraint obtains
\begin{align*}
c_y = \frac{1}{1+\beta}[y_y+\frac{y_o}{1+R}] \\
c_o = \frac{\beta}{1+\beta}(1+R)[y_y+\frac{y_o}{1+R}]
\end{align*}

Notice that with log utility the desired amount of consumption when young is simply a constant fraction of expected lifetime income.  If we use the notation $\bar{Y}$ to denoted expected lifetime income then we have

$$
c_y = \frac{1}{1+\beta}\bar{Y}
$$

which is exactly what our mini-theory said about consumption:

$$
C_t = a_c \bar{Y}_t
$$

The key lesson, and the one embodied in our mini theory, is that current consumption depends on the present value of lifetime income and not on current income.  This is sometimes called the \textbf{permanent-income hypothesis}.

In general, changes in the interest rate alter the the present expected value of lifetime income and therefore alter present consumption.  If we wanted to include this channel into our mini theory we could write

$$
C_t = a_c \bar{Y}_t-b_c(R_t-r)\bar{Y}_t
$$

(You should try to derive a modified IS curve using this mini-theory for consumption).

### A Digression on our Mini Theories: Investment
How do economists think about investment?

- Firms purchase capital goods to produce output that they can sell in the marketplace.
- We assume that firms maximize their expected lifetime profits.
- We don't always have to solve a profit maximization problem.  Instead we can appeal to the notion of arbitrage (or ''thinking on the margin'') to deduce when a firm is optimally investing


Consider investing in equipment (e.g. stove and a portable cart) to start my noodle business in Pingshan.  I have two alternative uses of my money:

- Put my money in Pingan bank and earn interest rate $R$
- Buy equipment for my business.  Let the price of equipment be given by $p_k$.

If I'm maximizing my personal wealth, then at the margin both alternatives should yield the same return

$$
\underbrace{R \times p_k}_{\text{return from bank account}} = \underbrace{MPK+\Delta p_k}_{\text{return from capital}}
$$


For clarity, let's normalize $p_k=1$ so that we have

$$
MPK = R-\Delta p_k
$$

which says that I should invest in capital up until the point at which the marginal product of capital is equal to the difference between the interest rate and the capital gain (or loss).


Another reason why my equipment will fall in value over time is due to depreciation.  As I use my equipment over time, the wear and tear gradually decreases it's market value.  We can account for this by writing

$$
MPK = \underbrace{R + \delta - \Delta p_k}_{\text{user cost of capital}}
$$

The \textbf{user cost of capital} is the total cost to a firm of using one more unit of capital.  

Notice that if there is no capital gains or depreciation then we are back to our standard condition that

$$
MPK = R
$$

Consider the introduction of a tax $\tau$ on business income.

$$
\underbrace{R}_{\text{cost of funds}} = \underbrace{(1-\tau)MPK-\delta+\Delta p_k}_{\text{return from investing in capital}}
$$

The business income tax raises the user cost of capital

$$
MPK = \underbrace{\frac{R+\delta-\Delta p_k}{1-\tau}}_{\text{user cost of capital}}
$$

Because extra output produced by capital is taxed, the marginal product of capital must be that much higher in order for the investment to be profitable.

How do we get from desired capital to investment?


Suppose output is produced according to the production function

$$
Y = AK^{1/3}L^{2/3}
$$

so that the marginal product of capital is

$$
MPK = \frac{1}{3}\frac{Y}{K}
$$

Letting $r$ denote the user cost of capital we have that firm's optimal capital stock is given by

$$
\frac{Y}{K} = 3r
$$

Capital evolves over time according to 

$$
\Delta K_{t+1}=I_t-\delta K_t
$$

or equivalently

$$
\frac{\Delta K_{t+1}}{K_{t}}=\frac{I_t}{Y_t}\frac{Y_t}{K_t}-\delta
$$

Finally, substitute out the optimal capital $Y/K=3r$ and solve for the investment rate

$$
\frac{I_t}{Y_t}=\frac{g_K+\delta}{3r}
$$

This equation says that the optimal investment rate depends on:

- the growth rate of the capital stock $\Delta K_{t+1}/K_t=g_K$
- The depreciation rate $\delta$
- The user cost of capital $r$


How does this relate to what we've studied so far?

Recall that in Solow we took the investment rate as exogenous

$$
\frac{I_t}{Y_t}=s
$$

Now we have a simple theory about where that investment rate comes from.  This answers some questions about what type of events might alter the savings rate (an increase in corporate taxes for example).

$$
\frac{I_t}{Y_t}=\frac{g_K+\delta}{3r}, \quad r=\frac{R+\delta-\Delta p_k}{1-\tau}
$$


But wait...there's more!

- Our consumption Euler equation pins down the long-run interest rate $R$
- Our Solow model pins down the long-run growth rate of output and capital
- Our condition that $MPK = r$ pins down the capital-output ratio
- Our optimal investment rule $I/Y=(g_k+\delta)/3r$ pins down the investment-output ratio and therefore also the consumption-output ratio (since $Y_t=C_t+I_t$).


### A Digression on our Mini Theories: Government Spending

The most important concept for thinking about fiscal policies is the government budget constraint

$$
G_t+(1+i_t)B_t=T_t+B_{t+1}+\Delta M_{t+1}
$$

which says the government spending and interest payment on Treasury bonds have to be paid using taxes, bonds, or seigniorage revenue.
To understand the \textit{dynamic} constraint the government is subject to, consider that the government budget constraint has to hold at all point in time.  In particular it is true that

$$
G_{t+1}+(1+i_{t+1})B_{t+1}=T_{t+1}+B_{t+2}+\Delta M_{t+2}
$$

or rearranged,

$$
B_{t+1}=\frac{T_{t+1}+B_{t+2}+\Delta M_{t+2}-G_{t+1}}{1+i_{t+1}}
$$

Since the budget constraint must hold at time $t$ and time $t+1$ we can combine them by substituting $B_{t+1}$ above into our time $t$ budget constraint:

$$
G_t+(1+i_t)B_t=T_t+\Delta M_{t+1}+\frac{T_{t+1}+B_{t+2}+\Delta M_{t+2}-G_{t+1}}{1+i_{t+1}}
$$

Or course the government budget constraint also has to hold at time $t+2$ which gives us

$$
B_{t+2}=\frac{T_{t+2}+B_{t+3}+\Delta M_{t+3}-G_{t+2}}{1+i_{t+2}}
$$

which can again be substituted out to give
\begin{align*}
&G_t+(1+i_t)B_t= \\
&T_t+\Delta M_{t+1}+\frac{T_{t+1}+\Delta M_{t+2}-G_{t+1}}{1+i_{t+1}}+\frac{T_{t+2}+B_{t+3}+\Delta M_{t+3}-G_{t+2}}{(1+i_{t+1})(1+i_{t+2})}
\end{align*}

Continuing process over and over again eventually gives

$$
\underbrace{(1+i_t)B_t}_{\text{PV of initial debt}} + \underbrace{\sum_{s=t}^\infty R_{t,s} G_s}_{\text{PV of government spending}} = \underbrace{\sum_{s=t}^\infty R_{t,s} T_s}_{\text{PV of tax revenue}} + \underbrace{\sum_{s=t}^\infty R_{t,s} \Delta M_{s+1}}_{\text{PV of seigniorage}}
$$

where

- $R_{t,s}=\frac{1}{\prod_{\nu=t+1}^s (1+r_{\nu})}$ is the market discount factor
- $\lim_{s \rightarrow \infty} R_{t,s} B_s =0$ to rule out asset bubble in government bonds

The important lesson is this: although the government never has to run a balanced budget in any given year, it does have to run a balanced budget in the present value sense.

- The amount that the government can borrow is limited by the amount it can credibly expect to pay back.
- The amount that it can pay back depends on the growth rate of the economy, since GDP is in some sense the total potential tax base.
- If the economy is growing very fast then the debt-to-GDP ratio will decline for a given path of government spending
- If the economy is growing more slowly than expected, then future tax revenue is less certain and will almost certainly cause the yield on the bond to increase 

Here is the way we are thinking about government spending in our mini theory:

- When the economy is on it's long-run equilibrium (balance growth path if you like) then the present discounted value of GDP is given by $\sum_{s=t}^{\infty}\bar{Y}_s/(1+r)^s$
- This effectively determined the present value of tax revenue and therefore the present value of feasible government spending
- This long-run level of government spending is $G_t=a_g \bar{Y}_t$
- Of course the government may wish to increase or decrease spending ($\uparrow \downarrow a_g$), but this can only happen temporarily since the government budget constraint must be satisfied in the present value sense.


### A Digression on our Mini Theories: Net Export

In the long-run, we expect arbitrage in goods markets to guarantee that identical goods have the same price in any two countries

$$
1=\underbrace{q^g_{US/EUR}}=\underbrace{E_{\$/€}P^g_{EUR}} \times \hspace{-15mm}\underbrace{\frac{1}{P^g_{US}}}
$$

This is called the **law of one price**.  Of course we are abstracting from trade costs, but this is a reasonable starting point.  (Recall the Big-Mac index we discussed at the beginning of the course).

Rather than considering a single good, we could imagine a bundle of goods should cost the same between two countries in which case we could write

$$
1=q_{US/EUR}=\frac{E_{\$/€}P_{EUR}}{P_{US}}
$$

This is called **absolute purchasing power parity** (PPP).  

Absolute PPP says that the nominal exchange rate is determined by the ratio of price levels:

$$
E_{\$/€} = \frac{P_{US}}{P_{EUR}}
$$

\textbf{Relative PPP} says that the rate of depreciation depends on the difference between the inflation rates of two countries

$$
\frac{\Delta E_{\$/€}}{E_{\$/€}}=\pi_{US}-\pi_{EUR}
$$

These two theories empirically do reasonably well in the long-run, but not in the short-run.  This is a long-run theory of exchange rates.

We already developed some tools to delve a bit deeper into this long-run theory of exchange rates.


Use the quantity theory of money to say

$$
\frac{\Delta E_{\$/€}}{E_{\$/€}}=\pi_{US}-\pi_{EUR} = (g^{US}_M-g^{US}_Y)-(g^{EUR}_M-g^{EUR}_Y)
$$

which says that long-run changes in the nominal exchange rate are driven by differences in growth rates of the money supply and growth rates of GDP between two countries.

A short-run theory of exchange rates comes from uncovered interest rate parity (UIP)

$$
(1+i_{\$})=(1+i_{€})\frac{E^e_{\$/€}}{E_{\$/€}}
$$

which says that holding dollar deposits in the US must give the same rate or return as converting dollars to Euros and holding deposits in the EU and then converting them back to dollars.


Uncovered interest rate parity is convenient to use in an approximate form

$$
i_{\$}=i_{€}+\frac{\Delta E^e_{\$/€}}{E_{\$/€}}
$$

which says that the expected depreciation of a currency depends on the nominal interest rate differential between two countries.

In a long-run equilibrium we have both relative PPP and UIP holding
\begin{align*}
\frac{\Delta E_{\$/€}}{E_{\$/€}}&=\pi_{US}-\pi_{EUR} \\
\frac{\Delta E^e_{\$/€}}{E_{\$/€}} &= i_{\$}-i_{€}
\end{align*}
so that 

$$
\pi_{US}-\pi_{EUR}=i_{\$}-i_{€}
$$

which implies that 

$$
r^e_{US}=r^e_{EUR}
$$

In the long-run real interest rates between countries should equalize!

In the short-run, however, we think that prices don't adjust instantaneously so that PPP need not hold.  Therefore, changes in the nominal exchange rate can affect the real exchange rate

$$
q_{US/EUR}=\frac{E_{\$/€}P_{EUR}}{P_{US}}
$$

Consider this chain of reasoning:

- The central bank increases the nominal interest rate $i_{\$}$
- In the short-run this will also increase the real interest rate above it's long-run level $R>r$
- The high rate of return attracts financial flows to the US causing the dollar to appreciate via UIP
- Because prices don't adjust immediately, the real exchange rate appreciates as well meaning that goods in the US are relatively more expensive
- Relatively more expensive US goods causes US exports to decline and imports to increase


If we wanted to incorporate these effects into our mini theory we could write

\begin{align*}
NX_t &= a_{nx}\bar{Y}_t-b_{nx}(R_t-r)\bar{Y}_t+b_{nx}(R_t^w-r)\bar{Y}_t \\
&=a_{nx}\bar{Y}_t-b_{nx}(R_t-R^w_t)\bar{Y}_t
\end{align*}
(You should try to derive a modified IS curve using this mini-theory for net exports.)


Notice that in the long-run $R_t=r=R_t^w$ and we have that net exports to output is at it's constant level $a_{nx}$ determined by theories from trade outside of our macro model.

### A Simple Monetary Policy in our Macroeconomy

We have an IS curve that gives us an equilibrium relationship between the output gap and the real interest rate.

$$
\tilde{Y}_t=a-b(R_t-r)
$$

How do we determine the short-run real interest rate?


We assume that the central bank can target whatever real interest rate it would like.  

- The central bank intervenes in the market for reserves to change the nominal interest rate (recall our model of the market for reserves)
- Changes in nominal interest rates influence the real interest rate because we assume that \textbf{inflation displays inertia, or stickiness, in the short run}

$$
R_t = i_t-\pi_t
$$


Formally, we model monetary policy by saying that $R_t$ is a policy variable determined with complete discretion by the central bank.


We will call this our **MP curve**, although for now it is a trivial curve since it is just whatever value the central bank picks for $R_t$.


Nevertheless, once the central bank chooses it's policy for the real interest rate, the IS curve tells us the short-run equilibrium output gap.

Our model of the short-run so far:
\begin{align*}
\text \quad \tilde{Y}_t&=a-b(R_t-r) \\
\text{MP Curve} \quad R_t &= \bar{R}_t
\end{align*}

### The Phillips Curve

In 1958, economist A.W. Phillips published a famous empirical paper that examined the relationship between unemployment and wage growth in the UK.  He found that periods of low unemployment were associated with rapid rises in wages, whereas periods of high unemployment were associated with low growth in wages. Other economists found that this negative correlation between unemployment and inflation exists in many countries, and was dubbed the *Phillip's Curve*.

</div>

We already developed a theory to think about this relationship...

The idea behind the Phillips curve is that when labor markets are *tight*---unemployment rate is low---firms have difficulty hiring workers and raise wages to attract labor.  The rise in wages is passed on in the form of higher prices, which increases inflation.  The Phillips curve implied that policymakers could trade-off inflation for lower unemployment, and was very influential in the 1960s.


<div style="text-align: center;">
<img src="_static/figs_SR/Phillips_inflation.png" 
     alt="GDP growth across countries" 
     width="400">
</div>


In 1967, Milton Friedman and Edmund Phelps pointed out a severe theoretical flaw in the Phillps curve analysis: firms care about *real* wages, not *nominal* wages.  When workers and firms expect the price level to be rising, they will adjust nominal wages upward so that the real wage does not decrease: wages and overall inflation will rise one-for-one with increases in expected inflation, as well as respond to tightness in the labor market.  The Friedman-Phelps analysis also postulated the existence of a **natural rate of unemployment**: the unemployment rate that would exist with perfectly flexible prices and competitive markets.

From Milton Friedman's 1967 AEA address:


"At any moment of time, there is some level of unemployment which has the property that it is consistent with equilibrium in the structure of real wage rates. At that level of unemployment, real wage rates are tending on the average to rise at a ``normal” secular rate."



This is easy to understand in the context of our model on equilibrium unemployment.

- The unemployment rate consistent with a steady state wage is the steady state unemployment rate...

$$
u = \frac{s}{s+f(\theta)}
$$

- ...and from our Solow model we know GDP per capita can grow over time (the balanced growth path) if technological progress occurs...

$$
g_{Y/L}=g
$$

- ...so along a balanced growth path the steady state unemployment rate produces a ``normal" rise in real wages.




From Milton Friedman's 1967 AEA address:


"A lower level of unemployment is an indication that there is an excess demand for labor that will produce upward pressure on real wage rates. A higher level of unemployment is an indication that there is an excess supply of labor that will produce downward pressure on real wage rates...The``natural rate of unemployment” in other words, is the level that would be ground out by the Walrasian system of general equilibrium equations, provided there is imbedded in them the actual structural characteristics of the labor and commodity markets, including market imperfections."




We saw the relationship between "excess demand for labor" or "excess supply for labor" and it's impact on the wage before:

$$
w = \beta(y+\theta \kappa)+(1-\beta)b
$$


- Excess demand for labor indicates a tight labor market ($\uparrow \theta$) which puts upward pressure on wages
- Excess supply of labor indicates a loose labor market ($\downarrow \theta$) which puts downward pressure on wages


From Milton Friedman's 1967 AEA address:


"You will recognize the close similarity between this statement and the celebrated Phillips Curve. The similarity is not coincidental. Phillips’ analysis of the relation between unemployment and wage change is deservedly celebrated as an important and original contribution. But, unfortunately, it contains a basic defect---the failure to distinguish between nominal wages and real wages."




From Milton Friedman's 1967 AEA address:


"Implicitly, Phillips wrote his article for a world in which everyone anticipated that nominal prices would be stable and in which that anticipation remained unshaken and immutable whatever happened to actual prices and wages. Suppose, by contrast, that everyone anticipates that prices will rise at a rate of more than 75 per cent a year ... Then wages must rise at that rate simply to keep real wages unchanged. An excess supply of labor will be reflected in a less rapid rise in nominal wages than in anticipated prices, not in an absolute decline in wages."




- During the sample period that Phillips used, 1861-1913 and up to 1950, inflation was very low
- Most people did not anticipate substantial inflation
- The inflation that occurred between 1860 and 1950 was the same amount that occurred between 1970-1979
- Milton Friedman points out that it is real wages which matter, not nominal wages (just like in our model).  So if people expect inflation in the future, they will increase nominal wages and this does not reflect excess demand for labor




The wage rate we studied in our model of the labor market was

$$
w = W/P
$$

where $W$ is a nominal wage rate and $P$ is the price level.  Or in terms of growth rates

$$
g_w = g_W-\pi
$$

Expansionary monetary policy may temporarily increase output (leading to higher wages) but eventually leads to inflation---and people know that!  Firms and households will expect inflation and therefore increase nominal wages at the same rate of inflation in order to keep real wages the same.



From Milton Friedman's 1967 AEA address:


"Restate Phillips’ analysis in terms of the rate of change of real wages---and even more precisely, anticipated real wages---and it all falls into place."


Friedman argued that policy attempts to keep unemployment below it's natural rate were doomed to fail:

- Expansionary monetary policy would initially cause an increase in expenditure and output
- Workers will start to anticipate higher prices and demand higher nominal wages which puts upward pressure on unemployment
- In order to keep unemployment below the natural rate monetary policy would need another round of expansionary policies, but this would be accompanied by even higher inflation expectations
- Efforts to keep output above potential for a sustained period of time would lead to accelerating inflation
- If Friedman's reasoning is correct, than attempts to exploit the original Phillips curve are doomed to fail: the Phillips curve relationship in the data should break down.

The expectations-augmented Phillips curve shows that the negative correlation between unemployment and inflation breaks down when the unemployment rate remains at or below the natural rate of unemployment. Starting in the 1970s, after a period of very low unemployment rates, the negative correlation between unemployment and inflation disappeared.  It seems Friedman and Phelps were right! (they both got Nobel Prizes).  Today, many policymakers talk about the \textbf{non-accelerating inflation rate of unemployment (NAIRU)}: the unemployment rate at which inflation is constant.


The Friedman reasoning suggested a Phillips curve that looks like this:

$$
\pi_t = \pi_t^e - \omega(U_t-U_n)
$$

which became known as the \textit{expectations-augmented Phillips curve}.

- There is long-run unemployment determined by fundamentals $U_n$
- In the long-run, expected inflation will gravitate toward realized inflation.  You can't continually fool the public so $\pi^e_t \approx \pi_t$
- There is a short-run trade-off between unemployment and inflation




Let's take a naive approach to inflation expectations for simplicity

$$
\pi_t ^e= \pi_{t-1}
$$

In this case our Phillips curve is

$$
\pi_t = \pi_{t-1}-\omega(U_t-U_n) \qquad \text{or} \qquad \Delta \pi_t = -\omega(U_t-U_n)
$$


- When unemployment is below its long-run level, firms are producing more than their long-run output and prices rise by more than usual
- When unemployment is above its long-run level, firms are producing less than their long-run output and prices rise more slowly
- When unemployment is at its long-run level, the economy is neither booming nor slumping and the inflation rate remains steady (NAIRU)

How does the CBO estimate the NAIRU?  A more realistic assumption about inflation expectations is that they are formed by some weighted average of historical observations

$$
\pi_t^e =\sum_{i=1}^T \beta_i \pi_{t-i} , \quad \sum_{i=1}^T \beta_{i}=1
$$

so that the expectations augmented Phillips curve is

$$
\pi_t = \sum_{i=1}^T \beta_i \pi_{t-i}-\omega(U_t-U_n)
$$





You can run a regression of current inflation on past inflation and unemployment

$$
\pi_t =\alpha + \sum_{i=1}^T \beta_i \pi_{t-i}-\omega U_t + \epsilon_t
$$

where the intercept term is $\alpha = \omega U_n$.  So we have an estimate NAIRU given by

$$
U_n = \frac{\alpha}{\omega}
$$




There still exist supply shocks that can move inflation in the short-run that are independent of the labor market.  Often the Phillips curve includes these price shocks

$$
\Delta \pi_t = -\omega(U_t-U_n)+o
$$

The parameter $\rho$ is equal to zero in the absence of shocks.  Something like the oil price shocks in the 1970s would be modeled as $\rho>0$.



In economic models, it is the Phillips curve that serves as an \textit{aggregate supply curve} showing the relationship between equilibrium inflation and GDP.


We know there is an obvious relationship between GDP and unemployment which could (if you're so inclined) be modeled simply by our production function

$$
Y_t = A_t K_t^{1/3}((1-u_t)L_t)^{2/3}
$$

where $(1-u_t)L_t$ is the number of employed workers in the economy.



The Phillips curve that we will use in our short-run model of the macroeconomy is written as

$$
\Delta \pi_t = \nu \tilde{Y}_t+o
$$


- $\Delta \pi_t = \pi_t - \pi_{t-1}$ is the one-period change in inflation
- $tilde{Y}_t$ is the output gap
- $\rho$ is a price shock




Often sources of inflation are categorized in two ways:

- Cost-push inflation: results from negative supply shocks or a push for wage hikes that are beyond those justified by productivity gains
- Demand-pull inflation: results when policy makers pursue policies that increase aggregate demand

### Our Basic Macro Model
Now we have a macro model with three equations:

\begin{align*}
\text{IS Curve]}: \quad \tilde{Y}_t &= a-b(R_t-r)\\
\text{MP Curve}: \quad R_t &= \bar{R}_t \\
\text{Phillip's Curve}: \quad \Delta \pi_t &= \nu \tilde{Y}_t+o
\end{align*}

- The IS curve determines equilibrium output gap for a given interest rate
- The MP curve shows that monetary policy chooses the interest rate with complete discretion
- The PC shows the rate of change in inflation given the output gap



We can use this three equation model to investigate some macroeconomic events in history or predict how the economy responds to future shocks.


For example, let's consider what monetary policy under Paul Volcker did in the 1970s.  During the 1970s inflation was very high (reaching over 12\% by 1980), and chairman Paul Volcker was in charge of bringing inflation down.  He implemented contractionary monetary policy:

- Contractionary monetary policy means $\uparrow R_t$
- Higher interest rates depress investment demand which generates a negative output gap $\tilde{Y}_t<0$
- Negative output gap represents low aggregate demand which causes inflation to decelerate $\Delta \pi_t <0$

There are several other policy experiments you should try on your own:

- Changes to aggregate demand $a \neq 0$ (e.g. amazing progress in AI spurs a boom in investment $a>0$)
- Changes to the natural rate of interest $\uparrow \downarrow r$ (e.g. a permanent increase in corporate taxes increases the user cost of capital causing a decline in the natural rate of interest $\downarrow r$)
- Price shocks $o \neq 0$ (e.g. a shortage of microchip imports causing the price of microchips and all electronic goods that use microchips to temporarily increase $o>0$).
- Changes to the sensitivity of investment demand to interest rates $\uparrow \downarrow b$ (e.g. regulatory requirements on business loans become more strict making it more difficult for entrepreneurs to acquire bank credit $\downarrow b$).
- Changes to the slope of the PC $\uparrow \downarrow \nu$ (e.g. industries become more concentrated generating higher market power for firms and markups $\uparrow \nu$).



## Monetary Policy Rules and Stabalizing the Economy


Most central banks have a dual mandate: maximum employment and price stability.


Law of the People’s Republic of China on the People’s Bank of China, Article 3 says: “ aim of monetary policies shall be to maintain the stability of the value of the currency and thereby promote economic growth.”


We can represent the central banks goals by considering it minimizing a \textbf{loss function}:

$$
\min \mathcal{L} = \tilde{Y}_t^2+\phi (\pi_t - \bar{\pi})^2
$$

where $\bar{\pi}$ is the inflation target and $\phi$ represents how much the central bank cares about inflation deviating from the target relative to the output gap.



$$
\mathcal{L} = \tilde{Y}_t^2+\phi (\pi_t - \bar{\pi})^2
$$

Enough fiddling with graphs, let's just solve the central bank's problem mathematically.


Substitute the constraint into the objective function

$$
\min (\frac{\pi_{t}-\pi_{t-1}-o}{\nu})^2+\phi(\pi_{t}-\bar{\pi})^2
$$

The first-order condition to this unconstrained problem gives 

$$
\tilde{Y}_t=-\nu \phi(\pi_t-\bar{\pi})
$$

which represents the \textbf{optimal monetary policy rule}.


We now have a more sophisticated \textbf{monetary policy rule} that represents the central bank's dual objective of stabilizing prices and encouraging the long-run trajectory of the economy given that there exists a short-run equilibrium relationship between inflation and the output gap coming through the Phillips curve.

$$
\tilde{Y}_t=-\nu \phi(\pi_t-\bar{\pi})
$$


- A flatter Phillips curve $\downarrow \nu$ implies that the central bank will fight inflation less aggressively in that it does not induce such a large negative output gap.
- A central bank that cares more about inflation $\uparrow \phi$ will fight inflation aggressively in that it induces large negative output gaps whenever inflation is above target.



Make one final adjustment to our model that is more realistic: changes in the interest rate affect investment demand with a one-period lag.

$$
\tilde{Y}_t = a-b(R_{t-1}-r)
$$

There are empirical studies on the effects of monetary policy which suggest that macro variables like consumption, investment, and output respond to interest rate changes with a 1 to 2 quarter lag.


Our macro model is now given by these three equations:
\begin{align}
\text{IS Curve}: \quad \tilde{Y}_t &= a-b(R_{t-1}-r)\\
\text{MP Curve}: \quad \tilde{Y}_t &= -\nu \phi(\pi_t-\bar{\pi}) \\
\text{Phillip's Curve}: \quad \Delta \pi_t &= \nu \tilde{Y}_t+o
\end{align}

- The IS curve determines equilibrium output gap for a given interest rate
- The MP curve shows that monetary policy adjusts interest rates according to a strict rule based on achieving its dual mandate
- The PC shows the rate of change in inflation given the output gap




Let's clarify how the central bank implements it's optimal monetary policy rule by adjusting interest rates.


Combine the monetary policy rule (2) with the Phillip's curve (3) to write

$$
-\left(\frac{1}{\nu \phi}+1 \right)\tilde{Y}_t=\pi_{t-1}+o-\bar{\pi}
$$

which represents the solution to the central bank's problem.  For any observed inflation $\pi_{t-1}$, the central bank optimally chooses an interest rate in order to implement output gap $\tilde{Y}_t$.


Substitute the solution to the central bank's problem into the IS curve to write

$$
R_{t-1}-r = \frac{a}{b}+\frac{\nu \phi}{b(\nu^2 \phi+1)}(\pi_{t-1}+o-\bar{\pi})
$$

This shows which interest rate the central bank chooses in order to implement it's desired output gap $\tilde{Y}_t$.


Consider $a=o=0$ and $\nu = \phi = b =1$ as an example:

$$
R_{t-1} - r = \frac{1}{2}(\pi_{t-1}-\bar{\pi})
$$

which says that for a 1\% increase in inflation above the target, the central bank should increase the real interest rate by 0.5\%.



$$
R_{t-1}-r = \frac{a}{b}+\frac{\nu \phi}{b(\nu^2 \phi+1)}(\pi_{t-1}+o-\bar{\pi})
$$


- As $\phi$ increases, the central bank responds more to inflation
- As $\nu$ increases, the central bank responds less to inflation
- As $b$ increases,  the central bank responds less to inflation



Lastly, when we observe central bank policy we don't see $R_t$.  Rather, we see the nominal interest rate $i_t$.  How do you understand our monetary policy rule in terms of observable data?


Use the Fisher equation

$$
i_t = \pi^e_t + R_t
$$

and recall that for our macro model we assume that $\pi_t^e=\pi_{t-1}$ so we can write

$$
i_t = \pi_{t-1}+R_t
$$

Now our monetary policy interest rate rule becomes

$$
i_t = r+\frac{a}{b}+\pi_{t-1}+\frac{\nu \phi}{b(\nu^2 \phi+1)}(\pi_{t-1}+o-\bar{\pi})
$$



For simplicity consider $a=o=0$ and $\nu=\phi=b=1$ so that we have a simple prediction for the short term nominal interest rate given inflation data

$$
i_t = r + \pi_{t-1}+\frac{1}{2}(\pi_{t-1}-\bar{\pi})
$$

Notice that a 1\% increase in inflation demand that the central bank raise the nominal interest rate by 1.5\%.  The principle that the central bank must fight 1 unit rise inflation with a larger than 1 unit rise in the nominal interest rate is sometimes called \textbf{leaning against the wind}.  This is obvious in that to manipulate real interest rates, the central bank has to adjust nominal rates by more than inflation.

A very popular monetary policy rule, that has been empirically very accurate for explaining central bank policy around the world, is the Taylor rule:

$$
R_t - r = \varphi_1(\pi_t-\bar{\pi})+\varphi_2\tilde{Y}_t
$$

where $\varphi_1, \varphi_2$ captures the objectives of the central bank for fighting inflation versus keeping output on it's long-run trajectory.


We can derive such a Taylor rule in our model if we assume that output affects inflation with a one-period lag in the Phillip's Curve.

## A 3-Equation Macro Model


Our macro model is given by these three equations:
\begin{align*}
\text{IS Curve}: \quad \tilde{Y}_t &= a-b(R_{t-1}-r)\\
\text{MP Curve}: \quad \tilde{Y}_t &= -\nu \phi(\pi_t-\bar{\pi}) \\
\text{Phillip's Curve}: \quad \pi_t &= \pi_{t-1}+\nu \tilde{Y}_t+o
\end{align*}

- The IS curve determines equilibrium output gap for a given interest rate
- The MP curve shows that monetary policy adjusts interest rates according to a strict rule based on achieving its dual mandate
- The PC shows the rate of change in inflation given the output gap

$$

- Graphically this is represented by an upward shift of the Phillip's curve which crosses the point $(\tilde{Y}_t,\pi_t)=(0,\pi_{-1}+o)$

- Negative aggregate demand shock: Output decreases, inflation decreases, central bank increases output, increasing inflation over time.

- Negative inflation shock: Inflation decreases, central bank increases output, increasing inflation over time.

One ingredient that has been very important in macro models since the financial recession is some way of capturing shocks to the financial sector.


One very simple way (not sufficient for serious policy but okay for now) to think about financial crises, is to say that there exists sharp increase in risk premiums that causes interest rates to rise

$$
R_t = R_t^{ff} + f
$$

where $R_t$ is the short-term real interest rate at which firms borrow in financial markets, $R_t^{ff}$ short-term real interest rate chosen by the central bank, and $f$ is an exogenous risk premium.


During the financial crises in 2008 and 2009, financial market broke down due to solvency and liquidity reasons and business lending dried up. 


Even though the Federal Reserve cut interest rates down to near zero, the rate at which firms and households borrow to finance investment was rising instead of falling.


We could represent this phenomena in our model as a rise in risk premiums $\uparrow f$ which causes real lending rates to rise.

     alt="GDP growth across countries"

If we wanted to talk about shocks to financial markets in our macro model, we could add this equation:
\begin{align*}
\text{IS Curve}: \quad \tilde{Y}_t &= a-b(R_{t-1}-r)\\
\text{MP Curve}: \quad \tilde{Y}_t &= -\nu \phi(\pi_t-\bar{\pi}) \\
\text{Phillip's Curve}: \quad \pi_t &= \pi_{t-1}+\nu \tilde{Y}_t+o \\
\text{Financial Frictions}: \quad R_t &= R_t^{ff} + f
\end{align*}
But actually this is not difficult since we could incorporate these financial frictions into our IS curve

$$
\tilde{Y}_t = a-b(R_{t-1}^{ff}+f-r)
$$

so that shocks to risk premiums would appear exactly as a negative shock to aggregate demand. 


There is one more key idea to explore in our macro model: the role of expectations.


Up until now we have assumed that agents' forecast of the future is based on historical data

$$
\pi_t^e = \pi_{t-1}
$$

Consider that the central bank's announced inflation target $\bar{\pi}$ does not appear in our equation of how agents form expectations.


This is not reasonable recognizing that is one of the most important advancements in macroeconomics for which Robert Lucas was credited with a Nobel Prize.


**Rational expectations** assumed that people make use of all available information to make their best forecast of future economic variables.


In the presence of rational expectations, the central bank's willingness to fight inflation becomes a crucial determinant of expected inflation.


Since actual inflation is influenced by inflation expectations

$$
\pi_t = \pi_t^e+\nu \tilde{Y}_t
$$

policymakers can try to influence those expectations in order to reduce the costs of maintaining an inflation target.


Consider our macro model with rational expectations
\begin{align*}
\text{IS Curve}: \quad \tilde{Y}_t &= a-b(R_{t-1}-r)\\
\text{MP Curve}: \quad \tilde{Y}_t &= -\nu \phi(\pi_t-\bar{\pi}) \\
\text{Phillip's Curve}: \quad \pi_t &= \pi_t^e+\nu \tilde{Y}_t+o \\
\text{Financial Frictions}: \quad R_t &= R_t^{ff} + f
\end{align*}
and suppose that the central bank announces that it wishes to lower long-run inflation from 4\% to 2\%.

- If the central bank has some credibility, then firms and households will believe that the central bank will take action to enact it's monetary policy of $\bar{\pi}=2\%$.
- That is, people will adjust their expectations of future inflation to be in line with the central bank's policy announcement

$$
\pi_t^e = 2\%
$$

- This causes the Phillip's curve to shift down immediately to the new inflation target and the economy achieves a new long-run equilibrium without a costly recession

<img src="_static/figs_SR/ISMPPC_13.png" 
     alt="GDP growth across countries" 
     width="400">
</div>

This highlights the importance of **forward guidance**: the central bank's communication of what actions it plans to enact given future macroeconomic data.


A growing list of countries---the US, EU, Japan, UK, Australia, Brazil, Canada, New Zealand, and Sweden among them---announced a medium- to long-term inflation target to stabilize inflation expectations.


This is an example of **constrained discretion**: the central bank has the flexibility to respond to shocks in the short as it sees fit (perhaps deviating from a Taylor rule) but maintains a commitment to a particular inflation rate in the long-run.


Reality is probably somewhere in between naive expectations ($\pi_t^e = \pi_{t-1}$) and rational expectations ($\pi_t^e = E[\pi_t \vert \text{all information}]$).


Our model generates two extreme predictions about the dynamics of the macroeconomy so that you can think about reality being some convex combination of those two extremes.


Academic papers which try to model such a reality may assume some form of \textbf{dynamic learning} where households and businesses observe a limited set of information and make sophisticated forecasts (like an econometrician would) based on their own model of the economy. 


### Advanced Macro Models
To go beyond what we've learned in class toward frontier research in macro there are three preliminary steps:

- Provide **microfoundations** for each element of the IS curve and nominal rigidities
- Formalize the extent to which shocks are temporary versus permanent.  This can be done very simply by assume that shocks follow an AR(1) process

$$
a_t = \rho a_{t-1}+\epsilon_t
$$

where $\rho=0$ means the shock is temporary, $\rho=1$ means the shock is permanent $\rho \in (0,1)$ precisely measures the persistence of the shock.
- Learn some coding to solve the model

Writing down such a model is called a \textbf{dynamic stochastic general equilibrium model (DSGE)} and is solved numerically.

Conceptually you can think about combining every model we've studied so far (production function, capital accumulation, equilibrium unemployment, IS curve, Phillip's curve, monetary policy rule) introducing a 1 percent shock of your choice and numerically solving the trajectory of the economy over time.
