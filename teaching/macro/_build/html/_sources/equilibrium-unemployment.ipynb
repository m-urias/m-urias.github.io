# Equilibrium Unemployement

## The Labor Market

The **labor market** refers to the process by which workers find jobs, leave jobs, and the resulting wage that firms pay.  When analyzing the labor market, there are some key questions we want to address:

- How are aggregate labor market outcomes---unemployment, job vacancies, and employment---determined as equilibrium phenomena?  
- How are wages determined?
- What role does the labor market play in explaining business cycles and growth?


One of the defining features of the labor market we emphasize in this chapter is that trade in the labor market is a decentralized, uncoordinated activity that is costly for both firms and workers.  This is very different from the classical supply and demand story that characterizes other markets.  As such, we do not assume that there exists some market clearing price but rather that wages are the outcome of a negotiation between the firm and worker.  The process of finding appropriate employers and employees, negotiating wages, and/or returning to the labor market after an unsatisfactory job called **job creation** and **job destruction**.  These flows of workers finding jobs and losing jobs every month is a normal, dynamic feature of the labor market.

Two stylized facts about unemployment worth noting (in the US):

- Most unemployment spells are relatively short: Each month, about 25 percent of people who are unemployed will find job.  Therefore, the average unemployment spell is about 3 months
- Most of the total weeks of lost work are accounted for by people who are unemployed for a long period of time: In February 2016, almost 60 percent of the unemployed had been out of work for less than 14 weeks, but more than one quarter had been unemployed for more than 6 months

## Search and Matching Model of Unemployement

Given the dynamic character of the labor market observed in the data, we want to build a model that captures the following features:
- dynamic job creation and job destruction
- workers finding good jobs and firms finding good workers is not a costless endeavor
- unemployment always exists and is subject to fluctuations given shocks to the economy
- wages reflect the bargaining strength of firms/workers and the aggregate state of the labor market

Trade in the labor market is costly because of heterogeneities, frictions, and information imperfections.  Firms and workers have to spend resources before job creation and production can take place.  There is uncertainty about when good jobs arrive to job-seekers, and when good workers arrive to hiring firms.  The labor market does not clear automatically (contrary to a simple supply-and-demand analysis).

The **matching function** captures the implications of matching frictions without describing their causes explicitly: It gives the number of jobs formed at any moment in time as a function of the number of workers looking for jobs and the number of firms looking for workers.

$$
\mathcal{M} = m(\mathcal{U},\mathcal{V})
$$
where $\mathcal{U}$ is the number of unemployed (workers looking for jobs) and $\mathcal{V}$ is the number of vacant jobs (firms looking for workers).


We assume that the matching function is (i) increasing in both $\mathcal{U}$ and $\mathcal{V}$; (ii) concave (diminishing marginal matching); and has constant returns to scale.

The size of the labor force is $\mathcal{L}=\mathcal{U} + \mathcal{V}$.


What sort of real-world data do we have in mind?  For the US we have: 

- Current Population Survey (CPS): asks individuals about their labor market status and compute unemployment statistics.  This data can be easily accessed via FRED.
-  Help Wanted Index (HWI) is published every month by the Conference Board: measures the amount of “help wanted” advertisements published in major newspapers across the United States
- Job Opening and Labor Turnover Survey (JOLTS): monthly survey conducted by the Bureau of Labor Statistics (BLS) of nearly 16,000 business establishments measuring the amount of job advertisements online

Consider a **Cobb-Douglas matching function**

$$
\mathcal{M} = A \mathcal{U}^{a}\mathcal{V}^b
$$

You can estimate values of $a$ and $b$ that match observable data on unemployment, vacancies, and new jobs.

$$
\ln \mathcal{M}=\ln A + a \ln \mathcal{U}+b \ln \mathcal{V}
$$

Blanchard and Diamond (1990) estimated $a$ and $b$ using U.S. data by regressing the log of monthly national hiring on the log of unemployment and the log of vacancies (using Ordinary Least Squares). They show that one could not reject the hypothesis $a+b=1$, implying that CRS is an empirically supported assumption.


For our model, let's use the following matching function:

$$
\mathcal{M}=A\mathcal{U}^{1/2}\mathcal{V}^{1/2}
$$ (cobb_douglas_matching)

where $A$ measures the efficiency of the matching process. A country with widespread access to the Internet and employment agencies may have a higher value of $A$ than one that lacks such infrastructure. $A$ is also inversely related to the amount of heterogeneity and specialization in the economy. Economies with high specialization tend to have a lower value of $A$ because it takes more time for matches to form. $A$ could also be related to attitudes toward work.  For example, after Covid workers may become more picky about finding jobs that allow them to work from home with would be represented by a lower value of $A$.

The matching function can be used to derive the probability that a worker finds a job and the probability that a firm finds a worker. The matching probability of an unemployed worker is called the **job finding rate**:

$$
f(\theta) \equiv \frac{m(\mathcal{U},\mathcal{V})}{\mathcal{U}} = \frac{A\mathcal{U}^{1/2}\mathcal{V}^{1/2}}{\mathcal{U}}=A\sqrt{\theta}
$$ (job_finding_rate)

where $\theta = \mathcal{V}/\mathcal{U}$ is a measure of labor market conditions called **market tightness**: the relative number of available jobs to job-seekers. When there is a relatively small number of job-seekers and a large number of vacant jobs to choose from, it is easier for an unemployed worker to find a job.  Conversely, then there is relatively large number of job-seekers competing for few vacant jobs, it is very difficult for an employment worker to find a match.

During a period of time (a month, for example) a worker finds a job with probability $f$ and does not find a job with probability $1-f$.  The probability of a worker finding a job in the $n$th month is therefore $(1-f)^{n-1} f$.  If we want to find the average number of months it takes for a worker to find a job we compute

$$
\sum_{n=1}^{\infty}n (1-f)^{n-1} f=-1\cdot f \frac{d}{df}\sum_{n=1}^{\infty} (1-f)^{n}=\frac{1}{f}
$$

The average duration of unemployment is

$$
\frac{1}{f(\theta)}
$$

The matching probability of a firm is called the \textbf{vacancy filling rate}:

$$
q(\theta) \equiv \frac{m(\mathcal{U},\mathcal{V})}{\mathcal{V}} = \frac{A\mathcal{U}^{1/2}\mathcal{V}^{1/2}}{\mathcal{U}}=\frac{A}{\sqrt{\theta}}
$$ (vacancy_filling_rate)

By the same reasoning applied above the average duration it takes for a firm to fill a job is $\frac{1}{q(\theta)}$.

A tight labor market (high $\theta$) means there are many firms searching for few job-seekers.  This makes it difficult for firms to find workers resulting in a long average duration of a vacancy.  Having now characterized the process by which jobs are created, we need to say something about how jobs are destroyed.  Job destruction is modeled as a current match being separated.  For simplicity, let's assume an exogenous **job separation rate** $s$.  Job separations happen because the worker-firm match is not productive forever.  We need not be specific about the reasons why the worker-firm match is no longer valuable.  Some plausible reasons are that there is a negative productivity shock or a negative demand shock which makes the firm unable to generate profits any longer.

Workers can be in one of two states: unemployed or employed.  Workers move from unemployment to employment at rate $f$ and move from employment to unemployment at rate $s$.  Firms similarly fill vacancies at rate $q$ and lose workers at rate $s$.  We can mathematically represent these flows using an equation which tracks how the umber of unemployed persons changes over time

$$
\mathcal{U}_{t+1}-\mathcal{U}_t=\underbrace{s\mathcal{E}_t}_{\text{flows in}}-\underbrace{f(\theta) \mathcal{U}_t}_{\text{flows out}}
$$ (law_motion_unemployment)

or

$$
\mathcal{U}_{t+1} = \underbrace{(1-f(\theta))\mathcal{U}_t}_{\text{doesn't find job}} + \underbrace{s\mathcal{E}_t}_{\text{loses job}}
$$

Here are empirical measures of the rates at which workers leave unemployment (U2E transitions) and enter unemployment (E2U transitions). ### Steady State Unemployment

There exists a steady state unemployment rate where $\mathcal{U}_{t+1}-\mathcal{U}_t=0$:

$$
\begin{align*}
s\mathcal{E}_t &= f(\theta) \mathcal{U}_t \\
s(\mathcal{L}_t-\mathcal{U}_t)&=f(\theta) \mathcal{U}_t \\
\end{align*}
$$

gives a steady state unemployment *rate*

$$
u \equiv \frac{\mathcal{U}}{\mathcal{L}}=\frac{s}{s+f(\theta)}
$$

Using the Cobb-Douglas matching function {eq}`cobb_douglas_matching`, the steady state unemployment rate

$$
u = \frac{s}{s+A\sqrt{\theta}}
$$ (steady_state_u)

gives a long-run relationship between unemployment and the rates at which jobs are created and destroyed.  This negative relationship between unemployment and vacancies is called the **Beveridge curve**. Our steady state unemployment equation {eq}`steady_state_u` delivers a negative relationship between vacancies and unemployment.

## An Empirical Digression on the Beveridge Curve

There is a recent paper by Ahn and Crane called *Dynamic Beveridge Curve Accounting*.  Let's walk through some of their paper:

Assume the flow of new hires in a month $t$ is given by a Cobb-Douglas matching function:

$$
\mathcal{M}_t = A_t \mathcal{U}_t^{1-\beta}\mathcal{V}_t^{\beta}
$$

The job finding probability is therefore 

$$
f_t = A_t (\theta_t)^{\beta}
$$

The law of motion for unemployment is

$$
\mathcal{U}_{t+1} - \mathcal{U}_t = s_t(1-\mathcal{U}_t)-f_t \mathcal{U}_t
$$



Substituting the job finding rate into the law of motion for unemployment gives

$$
\mathcal{V}_t = \left[ \frac{s_t(1-\mathcal{U}_t)-\Delta \mathcal{U}_{t+1}}{A_t\mathcal{U}_t^{1-\beta}}\right]^{1/\beta}
$$

where $\Delta \mathcal{U}_{t+1} = \mathcal{U}_{t+1}-\mathcal{U}_t$.  If we consider a steady state equilibrium where $s_t,A_t,$ and $\mathcal{U}_t$ are constants, then we obtain 

$$
\mathcal{V} = \left[ \frac{s(1-\mathcal{U})}{A\mathcal{U}^{1-\beta}}\right]^{1/\beta}
$$

This is exactly the Beveridge curve we learned with a Cobb-Douglas matching function!

How would you go about estimating the Beveridge curve using real data? First, you need data on unemployment and vacancy rates.  Then, you need to choose parameters for matching efficiency $A_t$, separation rate $s_t$, and the elasticity of the matching function $\beta$.

Here is what the paper by Ahn and Crane do:

They can directly observe data on unemployment and vacancy rates from Census data and Job Openings and Labor Turnover Survey
They set the monthly job-finding probability, $f_t$ using data on the number of short-term unemployed each month 

$$
f_t = 1 - \frac{\text{Unemployed for more than one month}_{t+1}}{\text{Unemployed}_t}
$$

They choose $s_t$ so that the law of motion for unemployed holds exactly

## Determining equilibrium market tightness

We know that unemployment is determined by how tight the labor market is, but what determines market tightness $\theta$? To answer this question we consider the firms decision to open up vacancies or not.  We make the following assumptions about firms' decision to create vacancies:
- It is not free to advertise a job.  The firm incurs a **vacancy posting cost** $\kappa$ every period it advertises an open job
- Once a job is filled, it produces **output** $y$ every period until the job is destroyed
- The worker is paid a **wage** $w$ each period until the job is destroyed

A firm considers the following sequence of events: First, open a vacancy at cost $\kappa$.  Next, with probability $q(\theta)$ it finds a worker and generates profits $y-w$ until the job is destroyed; or with probability $1-q(\theta)$ is doesn't find a worker leave the vacancy open at cost $\kappa$.  Finally, on average, a job lasts $1/s$ periods.

A firm optimally opens a vacancy as long as the cost of posting the vacancy is less than the expected present discounted value of the profits from a filled job:

$$
\frac{\kappa}{q(\theta)}<\frac{y-w}{s}
$$

It takes an average of $1/q$ periods for a firm to find a worker, and each period it pays the vacancy posting cost $\kappa$ resulting in expected vacancy posting costs of $\kappa/q(\theta)$.  A job lasts for an average of $1/s$ periods, and each period the firm earns profits $y-w$ resulting in expected profits of $(y-w)/s$.  It is optimal for the firm to continue opening vacancies so long as the expected profits of doing so are larger than the expected costs.  If all firms in the economy realize it is profitable to open vacancies then we expect the number of vacancies to rise which causes the labor market to become tighter ($\uparrow \theta$).

On the other hand, if

$$
\frac{\kappa}{q(\theta)}>\frac{y-w}{s}
$$

then we expect firms to close vacancies and labor market tightness would fall ($\downarrow \theta$). An equilibrium level of labor market tightness is therefore determined where firms have no incentive to either create or close vacancies:

$$
\frac{\kappa}{q(\theta)}=\frac{y-w}{s}
$$ (job_creation)

We will call {eq}`job_creation` our **job creation condition**.  We can substitute out the vacancy filling rate {eq}`vacancy_filling_rate` to write

$$
\theta = \left[ \frac{A}{\kappa}\left(\frac{y-w}{s} \right) \right]^2
$$

With the equilibrium level of market tightness determined by the job creation condition, we can use the Beveridge curve to describe the equilibrium unemployment rate.

$$
\begin{align*}
\theta &= \left[ \frac{A}{\kappa}\left(\frac{y-w}{s} \right) \right]^2 \\
u &= \frac{s}{s+A\sqrt{\theta}}
\end{align*}
$$

To have a good model of the labor market we need an endogenous wage.  However, the wage can't be determined from standard market-clearing (demand=supply) arguments because workers and firms are meeting in pairs, not a centralized marketplace.  When a firm and worker meet each other they negotiate a wage contract.  We won't discuss the technical details of how economists model a bargaining process, but we can intuitively describe the issues over which workers and firms bargain and present a wage equation reflecting the solution to the bargaining problem.

A job contract generates two distinct sources of value for the firm each period until the match is destroyed:  (i) revenue $y$ from selling output; and (ii) avoiding the cost of advertising a vacancy in the future given by $f \frac{\kappa}{q}=\theta q \frac{\kappa}{q}=\theta \kappa$.  The total benefit to a firm of settling on a job contract is therefore

$$
y+\theta \kappa
$$

If an unemployed worker settles on a job contract they sacrifice some benefits $b$ which could  include unemployment benefits, the value of leisure, home production etc.

In a negotiation, the wage is bounded below by the worker's unemployment benefit $b$ and the value to the firm of forming a match $y+\theta \kappa$.  With this in mind let's propose the following **wage equation**:

$$
w = \beta(y+\theta \kappa)+(1-\beta)b
$$ (wage_equation)

where $0 \leq \beta \leq 1$ represents the relative bargaining power of the worker.  If the worker has all the bargaining power ($\beta=1$) then they extract the full value of the mach to the firm $w=y+\theta \kappa$.  If the firm has all the bargaining power ($\beta=0$) then the wage is driven down to $w=b$ where the worker is indifferent between remaining unemployed or working. Parameter $\beta$ is exogenous and may change if, for example, increased labor union activity strengthens workers' bargaining ability.
 
Notice how the state of the labor market, measured by tightness $\theta$ influences wages. When $\theta$ is high, it is very difficult for firms to attract workers which gives labor a better bargaining position in the wage negotiation.- When $\theta$ is high, it is very easy for workers to find another job, which puts firms in a worse bargaining position in the wage negotiation

The wage equation {eq}`wage_equation` gives a positive relationship between labor market tightness and wages.  The wage equation {eq}`wage_equation` and job creation condition {eq}`job_creation` jointly determine the equilibrium wage and market tightness.  The Beveridge curve {eq}`steady_state_u` then determines equilibrium unemployment.

$$
\begin{align*}
\theta &= \left[ \frac{A}{\kappa}\left(\frac{y-w}{s} \right) \right]^2 \\
w &= \beta(y+\theta \kappa)+(1-\beta)b \\
u &= \frac{s}{s+A\sqrt{\theta}}
\end{align*}
$$

This system of three equation in three unknowns $(\theta,w,\mathcal{U})$ can be represented graphically.  The figure on the left shows the equilibrium combination of $(\theta,w)$ that satisfies both the job creation condition and the wage equation; while the figure on the right shows the equilibrium unemployment rate $\mathcal{U}$ that obtains with the level of market tightness.

## Comparative Statics

With the model in hand we can perform some experiments to predict how the labor market responds to different shocks, such as:
- labor market policies that give better unemployment benefits
- unionization of workers
- an economic recession
- an economic boom (AI industry takes off)
- technological improvements that make it easier to post job advertisements
- etc...

First, let's investigate what happens when there is positive productivity shock.  Mathematically, a higher value of $x$ will affect the job creation condition {eq}`job_creation` and the bargained wage {eq}`wage_equation`.  Intuitively, it's effect on the wage equation comes from workers' ability to capture a fraction $\beta$ of the productivity gains; while it's effect on job creation comes from firms' ability to earn higher profits with more productive workers thereby incentivizing them to create more vacancies.  More vacancies increase the job finding rate and therefore reduce unemployment.  The net result from the positive productivity shock is therefore higher wages and lower unemployment.

</div>

Second, consider an increase in unemployment income $b$. Higher unemployment benefits increases workers' **reservation wage**: the wage which makes then indifferent between working or not.  As a result, firms find it less attractive to create jobs and therefore reduce the number of vacancies.  Fewer vacancies decreases the job finding rate and thus increases unemployment.  The net result is higher wages and higher unemployment (a higher $\beta$ produces similar results).  Note that we are gnoring any disincentive effects (i.e. in spite of higher unemployment income, the unemployed continue to seek employment with the same effort).

Third, consider an increase in the separation rate $s$.  Higher separation rates mean the life of a job is on average shorter. The job creation curve shifts downward and moves the equilibrium to a lower wage and lower market tightness along the wage curve. This results in higher unemployment, but the shift in the Beveridge curve outward making the effect on job vacancies ambiguous.  The net result is lower wages and higher unemployment.

Fourth, consider a decrease in the efficiency of the matching function $A$.  n exogenous decrease in the matching function reflects greater **mismatch** in the labor market---differences in the qualities of jobs and workers, differences in their location, and imperfect information about these and other relevant parameters.  Higher mismatch slows the speed of job creation.  The job creation curve shifts downward reflecting the increase in average hiring costs.  The wage falls as does market tightness. The Beveridge curve shifts outward. The net result is lower wages and greater unemployment.

Fifth, consider an increase in vacancy posting costs $\gamma$.  Higher vacancy posting costs reduces the incentive for firms to create new vacancies, so the job creation curve shifts inward. Higher vacancy posting costs are partly passed on to workers through higher wages, so the wage rule shifts outward. Both fewer vacancy creation, and more job seekers lead to lower market tightness; but the net effect on wages is ambiguous. The net result is higher unemployment, but an ambiguous effect on wages.

## The natural rate of unemployment

For the U.S., the Congressional Budget Office (CBO) publishes estimates of the **natural rate of unemployment**. The natural rate of unemployment (NAIRU) is the rate of unemployment arising from all sources except fluctuations in aggregate demand. Estimates of potential GDP are based on the long-term natural rate. (CBO did not make explicit adjustments to the short-term natural rate for structural factors before the recent downturn.) The short-term natural rate incorporates structural factors that are temporarily boosting the natural rate beginning in 2008. The short-term natural rate is used to gauge the amount of current and projected slack in labor markets, which is a key input into CBO's projections of inflation."

The natural rate of unemployment is seen as a long-run equilibrium unemployment rate. We can also use our model to think about short-run dynamics of unemployment when the economy is hit by a macroeconomic shock.  Suppose that the economy experiences a negative shock which temporarily reduces the value of output $y$:

- Firms realize lower profits and so close job openings
- Vacancies immediately drop ($\downarrow \theta$) creating a looser labor market
- Unemployment gradually adjusts, according to our law of motion for unemployment, toward a new lower steady state level
- Eventually the shock to output dissipates, causing firms to open vacancies, causing unemployment to gradually return to its initial steady state level
