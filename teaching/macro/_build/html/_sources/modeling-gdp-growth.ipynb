# Modeling GDP Growth

Economic growth brings with it a host of benefits:
- longer life expectancy
- lower infant mortality rates
- better education
- less starvation
- etc...

We want a simple model that can be used to explain and measure economic growth over a long period of time.  Let's begin by imagining an economy composed of a single company: a noodle shop. To cook noodles, the owner needs to employ people (\textbf{labor}) and rent some cooking equipment (\textbf{capital}). The more people that work at the shop, or the more equipment the shop rents, the more noodles that can be cooked and served.

A \textbf{production function} tells us how many bowls of noodles can be produced if $L$ workers are combined with $K$ equipment:

$$
Y = F(K,L) = \bar{A}K^{1/3}L^{2/3}
$$

where $\bar{A}>0$ is some parameter that we call **productivity**.  In this example, we could imagine productivity reflecting how skilled the workers are at cooking, or how advanced the equipment is.  A production function can take on any form, but here we restrict our attention to a popular functional form called **the Cobb-Douglas production function**:

$$
Y = \bar{A}K^{\alpha}L^{1-\alpha}
$$

We will explore why this function is useful later, but for now let's just understand some of its properties.  First, let's understand what the production function says.  Suppose that $\bar{A}=1$ to make things simple.  If my noodle shop employs 15 workers and 9 cooking equipment then I can produce

$$
Y = 9^{1/3}15^{2/3} = 12.65
$$

bowls of noodles.

Second, notice that our production function exhibits something called **constant returns to scale**---if we double the amount of each input, then we double the amount of noodles produced.

$$
F(2K,2L) = (2\times 9)^{1/3} (2 \times 15)^{2/3}=2 \times (9^{1/3}15^{2/3}) = 25.3
$$

different production functions could have **increasing returns to scale**---doubling inputs gets more than double the output---or **decreasing returns to scale**---doubling inputs gets less than double the output.  Third, the production function function exhibits **diminishing marginal products**---the extra output we get from increasing one input gradually diminishes. Suppose we want to hire one more person, holding the amount of equipment fixed.  How many more bowls of noodles can we produce with one extra worker?

$$
MPL = \frac{\partial F(K,L)}{\partial L}=\frac{2}{3}\bar{A}\left(\frac{K}{L} \right)^{1/3} = \frac{2}{3}\left( \frac{Y}{L}\right)
$$

where MPL is called the **marginal productivity of labor**.  Of course we could say the same thing about capital:

$$
MPK = \frac{\partial F(K,L)}{\partial K}=\frac{1}{3}\bar{A}\left(\frac{L}{K} \right)^{2/3} = \frac{1}{3}\left( \frac{Y}{K}\right)
$$

where MPK is called the **marginal product of capital**.  Why might the marginal product of labor be diminishing as we add more workers?  Suppose that we have 5 units of equipment.  The first five people we hire should be very productive, but as we add more workers, there is less for them to do with only the five machines among them.  Too many cooks in the kitchen!  If we want to think about an entire economy, instead of a single noodle shop, we can still use our production function:

$$
Y = \bar{A}K^{1/3}L^{2/3}
$$

where $L$ denotes the number of workers in the economy and $K$ the amount of equipment.  $Y$ is the predicted level of real GDP---the number of goods produced by this economy. $A$ is what we call **total factor productivity** which will reflect many many things (e.g. institutions, technology, market structure).

What determines if a country is rich or poor? It's amount of workers, capital, and productivity! In fact, we can use the model to express real GDP per capita

$$
y \equiv \frac{Y}{L}=\frac{\bar{A}K^{\alpha}L^{1-\alpha}}{L} = \bar{A}k^{1/3}
$$

where $k=K/L$ is the amount of capital per person. Real GDP per capita is determined by two things:

- Productivity $\bar{A}$
- Capital per person $k$

We can now use some data on capital per person and real GDP per capita to check how well these prediction hold up empirically.  We want to test whether or not our model prediction

$$
y=\bar{A}k^{1/3}
$$

holds in the data.  For now, let's assume that $\bar{A}=1$ for every country, so that GDP per capita should be determined entirely by the amount capital per person in the economy.

With no difference in productivity, the model predicts smaller differences in income across countries than we observe.  The model predicts that most countries should be substantially richer than they are. We should allow for productivity differences to play a role:

$$
y = \bar{A}k^{1/3}
$$

but unfortunately TFP is not an observable variable.  Instead, we can treat the model as true and use data on GDP per capita to compute the level of TFP for each country:

$$
\bar{A}=\frac{y}{k^{1/3}}
$$

We can think of $\bar{A}$ as a measure of how big the gap is between our model and the data.  It is often referred to as ``the residual" or ``a measure of our ignorance."

In the year 2014, the per capita GDP of five richest countries was 70 times higher than that of the five poorest.  Differences in capital per person explain a factor of about 5 of this difference, while difference in TFP explain a factor of 14:

$$
\underbrace{\frac{y_{\text{rich}}}{y_{\text{poor}}}}_{70}=\underbrace{\frac{\bar{A}_{\text{rich}}}{\bar{A}_{\text{poor}}}}_{14 \quad\times}\underbrace{\left( \frac{k_{\text{rich}}}{k_{\text{poor}}}\right)}_{5}
$$

Since TFP is roughly three times as important as capital per person in this accounting exercise, we say that TFP explain three-quarters o the differences in per capita GDP, while capital per person explains about one-quarter.  There are many possible sources of differences in TFP across countries:

- Human capital: the stock of skills that people accumulate to make them more productive
- Technology: frontier technology (e.g. AI, software, information technology) improves how an economy can transform its inputs into output
- Institutions: laws, regulations, government policies
- Misallocation: government policies or market environments that inhibit resources to go where they are most productive (e.g. SOEs versus POEs)

Key points:

- Per capita GDP will be higher in countries that have high quantities of capital per person and use that capital efficiently.
- Observed large differences in capital per person only explain about one-third of differences in GDP per capita
- TFP is very important in explaining cross-country differences in economic performance.

Are there limitations to using GDP to measure **welfare**? 

- Inequality: the distribution of income in an economy gives us more information about how the gains from economic growth are enjoyed by different citizens
- Consumption is what matters: for individuals consumption (not necessarily investment or government spending) is what allows us to enjoy a good or bad standard of living
- Leisure matters too: why not measure GDP per hour to give a measure for how efficiently workers can generate income?
- Non-economic measures: life expectancy, infant mortality rates, crime rates, access to medical care etc.
