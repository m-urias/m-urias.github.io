If we were just statisticians we could do the following:

- Take the time series for real GDP over a long time horizon
- Use a statistical method (e.g. log-quadratic filtering or HP filtering) to separate the trend and cyclical component of the time series

$$
Y_t = Y_t^{\text{trend}}+Y_t^{\text{cycle}}
$$

The trend component reflects the long-term trajectory of the economy while the cycle component reflects short term deviations around the trend.

If we were economists then we could distinguish between the long-run and short-run in the following way:

- Long-run GDP, called **potential GDP**, represents the level of production at which resources are being fully and sustainably utilized.
- Short-run GDP, called **actual GDP**, is the realized level of output taking into account any shocks that are hitting the economy.

When the economy is in a long-run equilibrium

$$
\text{Actual GDP}=\text{Potential GDP}
$$

When the economy is pushed off it's long-run equilibrium by shocks

$$
\text{Actual GDP} \neq \text{Potential GDP}
$$


Our theory of long-run GDP comes from Solow/Romer/Equilibrium Unemployment. Our theory of long-run inflation comes from the Quantity Theory of Money and the Fisher equation. Now we want to develop a theory to explain short-run fluctuations of the economy away from its long-run trajectory.

In our model, we will describe short-run output in terms of percentage deviations of actual output from potential output

$$
\tilde{Y}_t = \frac{Y_t-\bar{Y}_t}{\bar{Y}_t}
$$

where $Y_t$ is actual output and $\bar{Y}_t$ is the long-run level of output.  Notice the bar notation indicating that potential output is determined outside this model. It is determined by our theories of the long-run!
(Note: The Congressional Budget Office of the US uses a version of the Solow model to construct its measure of potential GDP).

Some data facts (based on the previous pictures) that our short-run model should match:

- Short-run output follows a boom-bust cyclical pattern
- High levels of unemployment accompany low short-run output
- Low short-run output (recession) usually causes a fall in inflation


Two key ideas to explain why the economy fluctuates away from it's long-run equilibrium.

- The economy is being constantly hit by shocks.  These **economic shocks** include changes in commodity prices, disruptions in financial markets, changes in military spending, natural disasters, pandemics etc.
- Monetary and fiscal policy are used to affect output.  The government (fiscal authority + monetary authority) has tools to respond to shocks.  For example,  central bank may lower interest rates to stimulate the economy after a bad recession.  Or the government may send checks in the mail to everyone to stimulate spending during the pandemic.
