# Central Bank Implementation of Monetary Policy

## The Market for Reserves

Central banks manipulate overnight interest rates by intervening in the market for bank reserves.

Banks demand reserves for two reasons:

- Regulatory requirements: Many central banks require member commercial banks to hold a certain fraction of their liabilities in the form of reserves.  This is called the **required reserve ratio**.  In China it is usually around 12\%.
- Payments:  Banks need to hold reserves to settle payment requests that come through the payments network. 

Banks must hold the required **reserve amount** stipulated by law, but will also choose to hold some **excess reserves** as insurance against unanticipated payment requests.



On any given day some banks will have more reserves than they want, and some banks will have less reserves than they want.


The interbank money markets are where banks borrow and lend reserves to each other using short-term financing arrangements. 


In the US the interbank money market is called the **Fed Funds Market** and the equilibrium interest rate that banks charge each other for reserves is called the **Fed Funds Rate**.



The Fed funds rate (or any interbank lending rate) reflects the opportunity cost of holding excess reserves:

- When the Fed funds rate is high, it very costly to hold excess reserves because those reserves could have been lent out to earn a high rate of return
- When the Fed funds rate is low, it is not so costly for banks to hold excess reserves

**There is a negative relationship between the total demand for bank reserves and the overnight lending rate.**



Central banks have monetary policy tools which help them keep the short-term interest rate inside a target range:

- Discount window: Central banks can lend reserves to commercial banks and decide the interest rate to charge for these reserves.  This ``discount rate" places a **ceiling** on the Fed funds rate since banks would never choose to borrow in the interbank market at a higher rate.
- Interest on excess reserves: Central banks can pay interest to commercial banks on their excess reserve balances.  This places a **floor** on the Fed funds rate because banks would never lend out reserves at a rate below the interest rate they could earn by simply holding reserve balances.  The opportunity cost of holding excess reserves is gone.

The central bank completely controls the amount of reserves in the banking system.


The primary tool central banks use to adjust the supply of reserves is **open market operations**.

- The central bank can sell government bonds from its balance sheet in order to remove reserves from the banking system.
- The central bank can purchase government bonds from commercial banks in order to add reserves to the banking system.

**The central bank completely controls the supply of reserves and by adjusting this supply can target overnight interest rates in the market for reserves**.

- Prior to the financial crisis the Federal Reserve did not pay interest on excess reserves (the floor on the Fed funds rate was at 0), but in October 2008 the Fed decided to pay a positive interest rate on excess reserves thereby raising the floor.
- In September 2013 the Fed began a program of taking
in money from a wide range of non-bank financial institutions and paying interest. The technical name for this programme is the Overnight Reverse Repurchase Agreement Facility (ON RRP). Now the ON RRP rate is the effective floor.


The Fed says, Any counterparty that can use the ON RRP facility should be unwilling to invest funds overnight with another counterparty at a rate below the ON RRP rate, just as any depository institution eligible to earn interest on reserves should be unwilling to invest funds overnight with another counterparty at a rate below the interest rate on excess reserves. 

Since the financial crisis, the Federal Reserve has been providing **ample reserves** which implies that the equilibrium overnight interest rate is near its floor---interest on excess reserves.


The ON RRP facility provides extra control over the Fed funds rate in such an ample reserve system.


Since the financial crisis, choosing the administrative rates---interest on excess reserves (IOER) and ON RRP---are the most important tools of the Federal Reserve.

{Monetary Policy in China}
China's implementation of monetary policy works more like the US prior to the financial crisis.

- There are scarce reserves
- The central bank's short-term lending facility provides a ceiling on the overnight interest rate
- Interest on excess reserves provides a floor on the overnight interest rate
- The equilibrium interest rate is determined by adjusting the supply of reserves through the repo market

In addition China makes use of it's medium term lending facility to adjust 1-year rates.

To summarize, there are two basic environments under which monetary policy manipulates short-term interest rates:

- Scarce reserves (China): Open market operations change the supply of reserves in the banking system thereby adjusting the equilibrium interbank interest rate.
- Ample reserves (US post 2008): Reserves are ample so that open market operations don't move interest rates much.  Instead, administrative rates IOER and ON RRP are used to adjust the equilibrium interbank interest rate.

Now we have a basic idea of how central banks manipulate short-term interest rates, but how do short-term interest rates translate into long term interest rates which are more relevant for savings and investment decisions?

## Term Structure of Interest Rates

What is the Term Structure of Interest Rates?

- Often in economic models we assume that there exists one interest rate that measures the real return to capital investments.
- In reality, there exist many different interest rates among a class of assets that reflect differences in the **risk, liquidity, and maturity** on heterogeneous capital projects.
- We will focus on the class of assets called bonds, and identify why bonds with the same term to maturity have different interest rates.  
- Then we look at why the bond's term to maturity also affects its interest rate, which is called the **term structure of interest rates**.

{Maturity, Default, and Liquidity Risk}
There are three main factors that influence the interest rate on a particular type of debt instrument:

- **Risk**: 

- Some types of debt (e.g. government bonds for the US) have very little risk of default.  Other types of debt (e.g. credit cards, corporate bonds, mortgages) differ in the probability of default and thus how much the lender will lose if there is a default.
- Default occurs when the issuer of a bond is unable or unwilling to make interest payments when the promised or pay off the face value when the bond matures.
- The spread between interest rates on bonds with default risk and interest rate on default-free bonds (both of same maturity) is called the **risk premium**.

Liquidity: 

- Some types of debt can be easily traded (e.g. government bonds for the US) and/or
can be quickly and cheaply converted into cash (or reserves).  Other debt varies in its ease of finding a trade (e.g. corporate bonds) so it can be costly to sell these bonds in an emergency.
- Some types of debt are readily accepted as **collateral** in financial contracts.  The degree to which an asset is collateralizable largely depends on how quickly one party can liquidate it in the event that the counter-party breaks the contract.
- Other things equal, a decrease in liquidity of a bond will reduce the demand for that bond and lower its equilibrium price.


{Maturity, Default, and Liquidity Risk}
Maturity: 

- Different types of debt have different maturities---how long the borrower has to pay back.
- Some debt is overnight (e.g. repos, interbank loans) while other debt can be many decades (e.g. long-term government debt).
- The relationship between the yield on bonds with differing terms to maturity but the same risk and liquidity characteristics is called a yield curve.

A plot of the yields on bonds with different maturities but the same risk and liquidity characteristics is called a yield curve.

- Interest rates on bonds of different maturities move together over time.
- Yield curves can slope upward, downward, or be flat (but most often are upward sloping)
- When short-term interest rates are low (high), the yield curve tends to slope upward (downward)

{Explaining the yield curve}
We will describe three theories to explain the term structure of interest rates:

- expectations theory
- segmented markets
- liquidity premium theory

**Expectations theory** says that the interest rate on a long-term bond will equal the average of the short-term interest rates that people expect to occur over the life of the long-term bond.

- it assumes bonds of different maturities are *perfect substitutes*
- if they are perfect substitutes, then the must give the same rate of return
- this is a simple application of the *rate of return equality principle* we discussed already 


Consider investing \$1 of wealth in either

- a two-period bond
- two one-period bonds

Define the following 

- $i_t = $ today's interest rate on a one-period bond
- $i^e_{t+1} = $ interest rate on a one-period bond expected for tomorrow
- $i_{2t} = $ today's interest rate on a two-period bond


The expected return to holding a two-period bond is 

$$
(1+i_{2t})(1+i_{2t})-1 = 2 i_{2t}+(i_{2t})^2
$$

The expected return to holding two one-period bonds is

$$
(1+i_t)(1+i^e_{t+1})-1 = i_t + i^e_{t+1}+i_t(i^e_{t+1})
$$

Both bonds will be held only if these expected returns are equal.  Approximately we have that

$$
2 i_{2t} = i_t + i^e_{t+1} \quad \text{or} \quad i_{2t} = \frac{i_t + i^e_{t+1}}{2}
$$

We have shown that the two-period rate must equal the average of the two one-period rates.  Of course we could generalize this to an $n$-period bond and say that

$$
i_{nt} = \frac{i_t+i^e_{t+1}+i^e_{t+2}+...+i^e_{t+(n-1)}}{n}
$$

- When the yield curve is upward sloping, short-term interest rates are expected to rise
- When the yield curve is inverted, short-term interest rates are expected to fall


Expectations theory does a good job explaining:

- interest rates on bonds with different maturities move together
- yield curve tend to slope upward when short-term rates are low, and be inverted with short-term rates are higher
- short-term rates are more volatile than long-term rates

but does a poor job explaining:

- why they yield curve usually slopes upward
- the theory would imply that short-term rates are normally expected to rise, but in practice they are just as likely to fall as to rise

**Segmented markets theory** views different-maturity bonds as completely separate and segmented.  The interest rate on a bond is determined by the supply and demand for that bond, and is not affected by the expected returns on other bonds with different maturities.

- bonds with different maturities are not substitutes
- investors want to match the maturity of the bond to the desired holding period (for discount bonds this means investors face no risk)

Segmented markets theory can explain:

- why the yield curve usually slopes upward: risk averse investors prefer a short holding period with a short-term bond that has less interest rate risk so the demand for short-term bonds is relatively more than long-term bonds

Segmented markets theory cannot explain:

- why interest rates on bonds with different maturities move together: if the bond markets are segmented then there is no reason why changes in demand/supply of one bond maturities should affect others
- why yield curves tend to slope upward (downward) when short-term rates are low (high): there is no clear theory for how the level of short-term rates affect the demand/supply of short- versus long-term bonds.


The **liquidity premium theory** says that the interest rate on a long-term bond will equal an average of short-term rates expected to occur over the life of the bond plus a liquidity premium (also called a term premium) that responds to supply and demand conditions for that bond.

- bonds with different maturities are substitutes, but not perfect substitutes.
- investors tend to hold short-term bonds because they are subject to less liquidity risk
- investors must be compensated a positive liquidity premium to hold long-term bonds


The liquidity premium theory is written as

$$
i_{nt} = \frac{i_t+i^e_{t+1}+i^e_{t+2}+...+i^e_{t+(n-1)}}{n} + l_{nt}
$$
where $l_{nt}$ is the liquidity (term) premium for the $n$-period bond at time $t$, which is positive and rises with the term to maturity of the bond, $n$.

Liquidity premium theory does a good job at explaining:

- interest rates on bonds with different maturities move together (same logic as the expectations theory)
- why yield curves tend to slope upward when short-term rates are low: investors expect short-term rates to rise when they are relatively low, and the liquidity premium says that the yield curve will be even more steep than under expectations theory
- yield curves usually slope upward: the liquidity premium says that even if short-term rates are expected to stay the same, the long-term interest rates will be higher than short-term interest rates.



- a steep upward sloping yield curve says that investors expect short-term rates to rise in the future
- a moderately upward sloping yield curve says that short-term rates are not expected to rise or fall much
- a flat yield curve says that short-term rates are expected to fall moderately in the future
- a steep inverted yield curve says that short-term rates are expected to fall sharply in the future

Empirical evidence suggests that

- the spread between long-term rates and short-term rates contains information about predicting very short-term interest rate movements (over a few months) and very long-run interest rate movements, but little information about intermediate-term movements.
- the yield curve also has predictive power in regard to business cycles and inflation
- an upward sloping yield curve usually indicates loose monetary policy, and an inverted yield curve tight monetary policy


Central Bank Policy and Interest Rates

- Rates on longer-term government bonds have a key influence on mortgage rates, corporate debt, etc. For instance, the ten-year Treasury rate is used as a benchmark for US fixed-rate mortgages.
- Central banks control overnight interest rates. 
- Liquidity premium theory of interest rates explains how the central bank's control over short-term interest rates translate into long-term interest rates
- The theory also explains why **forward guidance**---central banks giving the public guidance on what they plan to do in the future---is an important part of monetary policy.



Summary of Key Ideas
In the long-run:

- Central banks can use growth in money supply to control inflation
- Central banks cannot use growth in money supply to stimulate GDP growth
- The Fisher equation tells us nominal interest rates should be rising when inflation is rising

In the short-run:

- The central bank intervenes in the banking sector to manipulate short-term interest rates
- Short-term interest rates translate into longer-term interest rates through the yield curve
- The central bank may be able to influence real interest rates temporarily which in turn can affect real economic decisions by households, firms, and the government.
