# Money, Inflation, and Interest Rates


Economist’s meaning of money:

- Anything that is generally accepted in payment for goods and services
- Not the same as wealth or income

Functions of money:

- Medium of exchange
- Unit of account
- Store of value


Evolution of Payments System:

- Precious metals like gold and silver and some other commodities
- Paper currency (fiat money): instrinsically worthless and inconvertible
- Checks
- Electronic means of payment
- Electronic money: Debit cards, Stored-value cards, Smart cards, E-cash, cryptos


Construct **monetary aggregates** using the concept of liquidity:

- M0 = currency held by nonbank public 
- MB = M0 + reserves
- M1 = M0 + traveler’s checks + demand deposits + other checkable deposits
- M2 = M1 + time deposits + money market funds + savings deposits


What measure of money should policymakers and researchers pay attention to?

- if different money aggregates are highly correlated, then it may not matter too much which one we choose
- if different money aggregates are not synchronized, then one measure of money may give a very different picture of the money supply than another
- if there is a link between the real economy and money supply, then policymakers must decide which money aggregate is most important

The measurement of monetary aggregates $M0,M1,M2$ just uses a simple sum across different assets.  

- All assets are treated as perfect substitutes even when clearly are not
- Example: Cash and demand deposits are used for transaction purposes whereas many interest bearing accounts are used for savings
- It is better to weight different types of money based on their liquidity in order to extract an aggregate measure of money that reflects transaction purposes

The Divisia index assigns weights to different types of money based on their transaction services


The **Divisia index** measures the transaction services of assets based on the opportunity cost of holding that asset compared to a benchmark interest bearing asset that is not used for transaction purposes

- Let $r_i$ be the interest rate on monetary asset $i$
- Let $R$ be the benchmark interest rate (e.g. 3 month time deposit rate) not used for transaction purposes
- The opportunity cost of holding monetary asset $i$ is

$$
p_i = R-r_i
$$

which measures of the cost of holding asset $i$ for its liquidity services

The actual weight assigned to asset $i$ is the opportunity cost $p_i$ multiplied by the amount of asset $i$ held as a proportion of the cost of all monetary assets:

$$
s_{it}=\frac{p_{it} M_{it}}{\sum_{i=1}^N p_{it}M_{it}}
$$

## Quantity Theory of Money
A useful framework for thinking about the role of money supply in the economy is the \textbf{Quantity Theory of Money}:

$$
MV = PY
$$


- $M$ is money supply
- $V$ is the velocity of money
- $P$ is the price level
- $Y$ is real GDP

The velocity of money measures how many times one unit of money changes hands over some period of time.

$$
V = \frac{PY}{M}
$$

If the velocity of money is constant, then the quantity theory suggests that nominal GDP is proportional to money supply

$$
PY = \bar{V}M
$$

Furthermore, if you believe that real output does not respond to changes is money supply, then the theory suggests a one-for-one relationship between money and prices

$$
P \bar{Y}=\bar{V}M
$$

**Changes in prices (inflation) are due to changes in money supply**.

Why do we believe that real output does not respond to changes in money supply?  Real output is determined by the factors of production (technology, labor, capital, land, human capital etc.).  This was what our theories about the long-run were saying.  Simply printing more bills does not change the amount of real resources an economy has, and therefore will not impact real production.  If money supply does not affect real variables in the economy, then we have something called **money neutrality**.


Why do we take velocity to be constant? This is more controversial, but during the time period under which this theory was developed velocity of measured to be quite stable.  Although that is not the case since.

Of course during periods of hyperinflation velocity should increase as people try to quickly get rid of bills deteriorating in value.  But for relatively low inflation rates we think that velocity reflects liquidity preferences among households and firms that does not depend on the total quantity of money or prices.

The velocity of money will fall when agents in the economy decide to hoard their money

- Households save more
- Firms increase cash holdings
- Banks hold more reserves

Notice however, that \textit{strictly speaking}, banks' desire to hold more reserves is not related to money supply.  It reflects banks' preferences/objectives.  There is another state of the world where the injection of money is not held by banks but rather lent out (probably at lower rates).


For our purposes, we will continue to assume that real output and velocity are determined by factors unrelated to prices of money supply.
\vspace{5mm}

We can write the quantity theory either in levels or growth rates:
\begin{align*}
M\bar{V}=P\bar{Y} \\
g_M+\bar{g}_V=g_P+\bar{g}_Y
\end{align*}
where I use the bar notation to remind you that these variables are determined by factors outside of the quantity theory (real gdp determined by our long-run model and velocity determined by liquidity preferences).



The quantity theory in growth rates is useful because it gives us a way of predicting inflation:

$$
g_P = g_M+\bar{g}_V-\bar{g}_Y
$$


A natural question is how well does money supply growth *actually* explain inflation in the data.  Let's summarize the results of some empirical studies addressing this question.



## Some Monetary Facts

McCandless and Weber (1995) is a highly cited study that measures the long-run correlation between different monetary aggregates, inflation, and real GDP.

- Time series data for 110 countries covering 1960 to 1990
- For each country, compute the long-run geometric average rate of growth for real GDP, inflation (using consumer prices), and three monetary aggregates (M0,M1,M2).

$$
\text{(Reminder about geometric growth)}:y_t=y_0(1+g)^t
$$

- Compute the coefficient of correlation between money growth, inflation, and real GDP

$$
\text{(Reminder about correlation coefficient)}:r = \frac{\text{cov}(x,y)}{\sigma_x \sigma_y}
$$


You can access the same data at the following places

- International Monetary Funds \textit{International Financial Statistics}: https://data.imf.org/
- Federal Reserve Economic Data: https://fred.stlouisfed.org/
- Peoples Bank of China: http://www.pbc.gov.cn/



Fact 1: In the long-run, there is a high (almost unity) correlation between the rate of growth of the money supply and the rate of inflation.  This holds across three definitions of money and across the full sample of countries and two subsamples.

Gertler and Hofmann (2016) wrote a paper called *Monetary Facts Revisited* using more current data and fancier econometrics to re-evaluate the long-run relationship between money growth and inflation (and also credit growth and financial crises).
\vspace{5mm}

To allow for a dynamic lead-lag relationship between money growth and inflation, they use an error-correction model:'

$$
\Delta \pi_{i,t}=\alpha_i+\phi_i(\pi_{i,t-1}-\theta(m-y)_{i,t-1})+\sum_{j=0}^4
\delta_{i,j}\Delta(m-y)_{i,t-j}+\epsilon_{i,t}
$$

where $\pi$ is the inflation rate, $(m-y)$ is the growth rate of money supply in excess of real GDP, and $\alpha_i$ are country fixed effects.

It is important realize that the interest rates you see in the real world are all in units of money.  But of course, inflation erodes the the purchasing power of your money and so nominal interest rates do not tell you how much purchasing power you expect to receive from some investment:

- You lend someone \$100 today at an interest rate of 6\%
- In 12 months you receive \$106
- During the last 12 months prices rose by 6\%
- You earned nothing

What really matters for economic decisions is the \textbf{real interest rate}: the rate of return measured in units of goods. 

The Fisher equation says that in equilibrium the gross real rates of return on real and nominal bonds must be the same.
\vspace{5mm}

Plan A: Invest in \$1 nominal bond today

- Let $i_{t+1}$ be the one period nominal interest rate
- The amount you have tomorrow is $(1+i_{t+1})$

Plan B: Take your \$1 and convert it into real goods and invest it 

- Let $r_{t+1}$ be the real rate of return
- The amount you have tomorrow is

$$
\frac{1}{P_t}(1+r_{t+1})
$$

and in terms of money that is

$$
\frac{P^e_{t+1}}{P_t}(1+r_{t+1})
$$





## Fisher Equation

In equilibrium (no arbitrage) it must be the case that 

$$
(1+i_{t+1})=\frac{P^e_{t+1}}{P_t}(1+r_{t+1})
$$

Using the notation $\pi^e_{t+1}$ to denote the expected net inflation rate we have that

$$
(1+i_{t+1})=(1+\pi^e_{t+1})(1+r_{t+1})
$$

Often you will see the Fisher equation written in it's approximate form

$$
i_{t+1}=r_{t+1}+\pi^e_{t+1}
$$

which uses the fact that $\pi^e_{t+1}r_{t+1} \approx 0$.



In this equation we take the real interest rate as exogenous.  Why?  
\vspace{5mm}

The real interest rate is determined by the real rate of return on capital projects

$$
r=MPK
$$

and this, in turn, depends on real factors.  Again, think of our long-run model.
\vspace{5mm}

We are consistently assuming that real variables affect real variables and nominal variables affects nominal variables: the **classical dichotomy**.
\vspace{5mm}

Of course this is not realistic, but we have to start simple.

Taking the real interest rate as exogenous, the Fisher equation predicts a one-for-one relationship between inflation and nominal interest rates:

$$
i_t = \bar{r}_t+\pi^e_t
$$

Another way to use the Fisher equation is to compute the real interest rate, which is of course unobservable!

- We can observe nominal interest rates
- We can observe inflation
- We can use the Fisher equation to measure the real interest rate

Notice that if the real interest rate where always equal to MPK, then periods of where $r<0$ would be puzzling.


It is possible for the real interest rate to temporarily deviate from its equilibrium value MPK.


**Central bank interventions in nominal interest rates can influence real interest rates**.
