# Melitz Model of Intra-Industry Trade


{Motivation}
International trade effects the types of firms that operate in the economy.

- Trade enhances grwoth opportunities of some firms, while simultaneously forcing out other firms
- Profits are reallocated toward more productive firms
- Protectionist policies are often intended to shelter less efficient firms

Bernard and Jensen (1999) show that market share reallocations towards more productive exporting plants accounts for 20\% of U.S. manufacturing productivity growth


{Roadmap}
To capture these stylized facts, we need a model that allows for firm heterogeneity

- Introduce firms with different producitivities into Krugman's model 
- Endogenize the distribution of firms under monopolistic competition
- Incoporate forward looking entry decision of firms facing sunk entry costs

SEVERAL trade papers use this framework.


{Environment - Preferences}
Agents have identical preferences over different varieties of goods and are CES:
$$
U = \left[ \int_{\omega \in \Omega} q(\omega)^{\frac{\sigma-1}{\sigma}}d\omega\right]^{\frac{\sigma}{\sigma-1}}
$$

where $\sigma >1$ is the elasticity of substitution (and let $\rho = (\sigma-1)/\sigma$).


{Environment - Preferences}
Consumer maximize their utility subject to the budget constraint,
$$
\int_{\omega \in \Omega} p(\omega)q(\omega)d\omega \leq I
$$



{Let's derive demand function together...}
We will consider the dual problem of minimizing the expenditure required to obtain $U$ utility.
$$
\min \int_{\omega \in \Omega} p(\omega)q(\omega)d\omega \quad \text{s.t.} \quad U = \left[ \int_{\omega \in \Omega} q(\omega)^{\rho}d\omega\right]^{\frac{1}{\rho}}
$$



{Let's derive demand function together...}
The first-order condition for the expenditure minimization problem gives equality of marginal rates of substitution to price ratios,
$$
\frac{q(i)^{\rho-1}}{q(j)^{\rho-1}}=\frac{p(i)}{p(j)}
$$

for any pair $i,j$.


{Let's derive demand function together...}
Substitute the first-order condition into the contraint to obtain,
$$
q(j) = \frac{p(j)^{\frac{1}{\rho-1}}}{\left[ \int_{\omega \in \Omega} p(i)^{\frac{\rho}{\rho-1}}\right]^{1/\rho}}U
$$

which is the compensated demand function for the $j$th variety.


{Let's derive demand function together...}
We can derive an expression for the minimum cost of attaining one unit of a bundel of goods $Q=U$.
$$
\int_{\omega \in \Omega} p(j) q(j) dj = \underbrace{\left[ \int_{\omega \in \Omega} p(i)^{\rho/(\rho-1)} di\right]^{(\rho-1)/\rho}}_{\text{price index}\hspace{2mm} =P}U
$$

The price index, $P$, measures the minimum cost of purchasing a bundle of consumption that provides one unit of utility.


{Let's derive demand function together...}
Demand for $q(\omega)$ can now be written more compactly,
$$
q(\omega) = \left( \frac{p(\omega)}{P} \right)^{-\sigma}Q.
$$

Ta da!


{Expenditure}
The implied spending on each good is therefore,
$$
r(\omega) = p(\omega)q(\omega) = R\left(\frac{p(\omega)}{P} \right)^{1-\sigma}
$$

where 
$$
P = \left[ \int_{\omega \in \Omega} p(\omega)^{\rho/(\rho-1)} d\omega\right]^{(\rho-1)/\rho}; \quad R = \int_{\omega \in \Omega} r(\omega)d\omega; \quad Q = R/P
$$



{Welfare}
We can now compute the welfare of agents by substituting out the demand function into utility,
\begin{align*}
U &= \left[ \int_{\omega \in \Omega} \left( \frac{p(\omega)}{P} \right)^{1-\sigma}Q^{\frac{\sigma-1}{\sigma}} \right]^{\frac{\sigma}{\sigma-1}} \\
&= QP^{\sigma} \left[ \int_{\omega \in \Omega} p(\omega)^{1-\sigma} \right]^{\frac{\sigma}{\sigma-1}} \\
&=QP P^{\sigma-1} P^{-\sigma} \\
&= \frac{I}{P} \\
&=\frac{1}{P}
\end{align*}
Welfare can be completely summarized by the price index.


{Production}
Labor is the only factor of production

- Set the total amount of labor to $L$ and choose wage as the numeraire: $w=1$
- Each variety is produced by a separate firm and there is free entry and exit
- Each firm draws its productivity $\varphi$ from a distribution $G(\varphi)$ after paying a sunk cost of entry $f_e$.
- Each firm faces an exogenous death probabiliy $\delta$



{Production}
Production function exhibits increasing returns to scale:
$$
TC(\varphi) = f + \frac{q(\varphi)}{\varphi}
$$

Higher producitivity firms have lower marginal costs. (Notice that $w=1$ so the above is labor demand, as in Krugman, \textit{and} total cost function).


{Let's derive the optimal price together...}
Firm profits are given by
\begin{align*}
\pi(\varphi) &= r(\varphi) - TC(\varphi) \\
&=p(\varphi) q(\varphi) - \frac{q(\varphi)}{\varphi}-f \\
&= \frac{p(\varphi)^{1-\sigma}}{P^{-\sigma}} Q - \left( \frac{p(\omega)}{P} \right)^{-\sigma}Q\frac{1}{\varphi}-f
\end{align*}
where the last equality uses the demand function for any variety.


{Let's derive the optimal price together...}
Taking first-order condition gives us the optimal price
$$
p(\varphi) = \frac{\sigma}{1-\sigma}\frac{1}{\varphi} = \frac{1}{\rho \varphi}
$$



{Firm production, revenue, and profit}
We therefore have that
\begin{align*}
q(\varphi) &= \frac{R}{P}\left( \frac{p(\varphi)}{P} \right)^{-\sigma} = RP^{\sigma-1}(\rho \varphi)^{\sigma} \\
r(\varphi) &= p(\varphi) q(\varphi) = R (P\rho \varphi)^{\sigma-1} \\
\pi(\varphi) &= \frac{r(\varphi)}{\sigma}-f
\end{align*}
Notice that $r'(\varphi)>0$ and $\pi'(\varphi) > 0$.


{Characterizing the equilibrium}
After entry, a firm has the following expected discounted value of profits
$$
\nu(\varphi) = \max\{0, \sum_{t=s}^{\infty} (1-\delta)^{t-s}\pi(\varphi) \} = \max \{0, \frac{\pi(\varphi)}{\delta}\}
$$

Note that since $\pi'(\varphi) >0$ there exists a $\varphi^*$ such that $\nu(\varphi) > 0$ iff $\varphi > \varphi^*$.  This productivity cutoff is characterized by
$$
\pi(\varphi^*) = 0 \quad \text{or} \quad r(\varphi^*) = \sigma f
$$



{Characterizing the equilibrium}
Define the distribution of mass of active firms as
$$
\mu(\varphi) = \left\{ \begin{aligned}
&\frac{g(\varphi)}{1-G(\varphi^*)} \hspace{2mm} &&\text{if} \hspace{2mm} \varphi \geq \varphi^* \\
&0 \hspace{2mm} &&\text{if} \hspace{2mm} \varphi < \varphi^*
\end{aligned}
\right. $$

This is the ex-post producitivty distribution conditional on successful entry.


{Characterizing the equilibrium}
Melitz solves the model using a method of aggregation that has become very popular in the literature.


Define $\tilde{\varphi}$ as the weighted average producitivity:
$$
\tilde{\varphi} = \left[ \int_{0}^{\infty} \varphi^{\sigma-1}\mu(\varphi) d\varphi \right]^{\frac{1}{\sigma-1}}
$$



{Characterizing the equilibrium}
Using the distribution of active firms, the average producitivity is written as,
$$
\tilde{\varphi}(\varphi^*) = \left[ \frac{1}{1-G(\varphi^*)}\int_{\varphi^*}^{\infty}\varphi^{\sigma-1}g(\varphi) d\varphi \right]^{\frac{1}{\sigma-1}}
$$

Notice that $\tilde{\varphi}$ is uniquely pinned down by $\varphi^*$ and the exogenous distribution $G(\varphi)$. 


{Characterizing the equilibrium}
Let the equilibrium mass of firms (varieties) be defined as $M$.  The price index for the economy is therefore,
$$
P = \left[ \int_{\varphi^*}^{\infty} p(\varphi)^{1-\sigma}M\mu(\varphi)d\varphi \right]^{\frac{1}{1-\sigma}}
$$

Using the monopoly pricing equation $p(\varphi) = 1/\rho \varphi$ we can re-write the above,
$$
P = \frac{M^{\frac{1}{1-\sigma}}}{\rho}\left[ \int_{\varphi^*}^{\infty} \varphi^{\sigma-1}\mu(\varphi) d\varphi \right]^{\frac{1}{1-\sigma}}=\frac{M^{\frac{1}{1-\sigma}}}{\rho \tilde{\varphi}} 
$$



{Characterizing the equilibrium}
Using the same aggregation we have that
\begin{align*}
R &= Mr(\tilde{\varphi}) \\
\Pi &= M \pi(\tilde{\varphi}) \\
Q &= M^{\frac{\sigma}{\sigma-1}}q(\tilde{\varphi})
\end{align*}


{Characterizing the equilibrium}
Next, denote average profit by
$$
\bar{\pi} = \frac{r(\tilde{\varphi})}{\sigma}-f
$$

Note that,
$$
r(\tilde{\varphi}) = R(P\rho \tilde{\varphi})^{\sigma-1}; \quad r(\varphi^*)=R(P\rho \varphi^*)^{\sigma-1}
$$

therefore,
$$
r(\tilde{\varphi}) = r(\varphi^*)\left(\frac{\tilde{\varphi}}{\varphi^*} \right)^{\sigma-1}
$$



{Characterizing the equilibrium}
Since $r(\varphi^*) = \sigma f$ (zero cutoff profit condition) we have that
$$
\bar{\pi} = \frac{r(\tilde{\varphi})}{\sigma}-f = f\left( \left( \frac{\tilde{\varphi}(\varphi^*)}{\varphi^*} \right)^{\sigma-1}-1  \right)
$$

The above is a key equation in solving the model.  It is called ZCP: zero cutoff profit condition.


{A comparative static}
The relationship between $\bar{\pi}$ and $\varphi^*$ is ambiguous.
$$
\tilde{\varphi}'(\varphi^*) = \frac{g(\varphi^*)\tilde{\varphi}^{2-\sigma}}{(\sigma-1)^{2}}\int_{\varphi^*}^{\infty}(\varphi^{\sigma-1}-\varphi^{*\sigma-1})g(\varphi) d\varphi > 0
$$

Therefore the numerator and denominator in $\bar{\pi}$ are both increasing in $\varphi^*$.  


{A comparative static}
Because profits tend to increase with firm producitivity, an increase in $\varphi^*$ will have a direct effect on profits $\bar{\pi}$.  But becasue firm profits are decreasing in the producitivity of rivals, there is also an addition effect that goes in the opposite direction.  


If the distribution $G(\varphi)$ has a fat enough right tail, the latter effect will dominate and the ZCP will be downward slopiong.  In the case of Pareto distribtution, the two effects exactly offset each otehr in which case the ZCP is a flat line, but more generally it is downward sloping.


{Characterizing the equilibrium}
The other key equation comes from the free entry condition.  
$$
\underbrace{(1-G(\varphi^*))\frac{\bar{\pi}}{\rho}}_{\text{expected profit from entry}} = \underbrace{f_e}_{\text{cost of entry}}
$$

The above is called the free entry condition FE.  Intuitively, for a given cost of entry, if firms are less likely to survive they need to run higher average profits.  Therefore, $\bar{\pi}$ and $\varphi^*$ are positively related.

{Characterizing the equilibrium}
Having determined the equilibrium values of $\bar{\pi}$ and $\varphi^*$, the other variables of interest can be easily determined.


From market clearing we have that $L = R = r(\tilde{\varphi})M$, where it follows that 
$$
M = \frac{L}{r(\tilde{\varphi}} = \frac{L}{\sigma(\bar{\pi}+f)}
$$

Since $\bar{\pi}$ is independent of $L$, an increase in $L$ implies an increase in the mass of varieties.


{Characterizing the equilibrium}
The welfare of a representative consumer is given by,
$$
U = \frac{1}{P} = M^{\frac{1}{\sigma-1}}\rho \tilde{\varphi}
$$

An increase in $L$ implies an increase in welfare as in the standard Krugman model.  


Once $\tilde{\varphi}$ and $\bar{\pi}$ are determined, the aggregate outcome predicted by this model is identical to one generated by representative dfirms who share the same productivity $\tilde{\varphi}$ and profit level $\bar{\pi}$.


{Impact of trade}
In the absence of trade costs, integration of the two economies will simply lead to an increase in the number of varieties.  There is not going to be any change in $\tilde{\varphi}$ because it is independent of $L$.  


Introduce trade costs:
\begin{enumerate}
- Iceberg trade costs: in order to sell 1 unit abroad firms need to ship $\tau \geq 1$ units
- Fixed exporting costs: in order to export firms must incur some startup/fixed cost $f_{ex}$ (information, distrubiton or regulation costs) after learning their producitivity.
\end{enumerate}


{Impact of trade}
We will assume $c=0,1,...,n$ countries are symmetric so that $w_c = 1$ in all countries.


Monopoly pricing now implies:
$$
p_d(\varphi) = \frac{1}{\rho \varphi}; \quad p_x(\varphi) = \frac{\tau}{\rho \varphi}
$$

and revenues in domestic and export market are:
$$
r_d(\varphi) = R_d (P_d \rho \varphi)^{\sigma-1}; \quad r_x(\varphi) = \tau^{1-\sigma}R_x(P_x \rho \varphi)^{\sigma-1}
$$

Note that by symmetry we must have: $P_d = P_x = P$ and $R_d = R_x = R$.


{Impact of trade}
Given that the export decision is made after firms know their productivity, we do not need to distinguish between the one-time investment cost $f_{ex}$ and the amortized per-period portion of this cost $f_x = \delta f_{ex}$.


Profits in the domestic and export markets are:
$$
\pi_d(\varphi) = \frac{r_d(\varphi)}{\sigma}-f; \quad \pi_x(\varphi) = \frac{r_x(\varphi)}{\sigma}-f_x
$$



{Impact of trade}
Expected value of a firm with productivity $\varphi$ is the same as before:
$$
\nu(\varphi) = \max\{0, \sum_{t=s}^{\infty} (1-\delta)^{t-s}\pi(\varphi) \} = \max \{0, \frac{\pi(\varphi)}{\delta}\}
$$

But total profits are now given by:
$$
\pi(\varphi) = \pi_d(x) + n \max\{0, \frac{\pi(\varphi)}{\delta}\}
$$

Similar to the closed economy case, we now have two cutoffs:
\begin{align*}
\varphi^* \hspace{2mm} \text{s.t.} \hspace{2mm} \pi_d(\varphi^*)=0 \\
\varphi_x^* \hspace{2mm} \text{s.t.} \hspace{2mm} \pi_x(\varphi_x^*)=0
\end{align*}


{Impact of trade}
In order to have both exporters and non-exporters in equilibrium (i.e. $\varphi_x^* > \varphi^*$) it is assumed that
$$
\tau^{\sigma-1}f_x > f
$$

Therefore, more productive firms select into exports.


The profit from exporting is smaller than the profit from domestic sales because of trade costs and higher exporting fixed cost.


{Impact of trade}
In the open economy case, the average productivity is defined as
$$
\tilde{\varphi}_T = \left[ \frac{1}{M_T}\left(M \tilde{\varphi}^{\sigma-1}+nM_x\left( \frac{\tilde{\varphi}_x}{\tau} \right)^{\sigma-1} \right)  \right]^{\frac{1}{\sigma-1}}
$$

where
$$
M_T = M + nM_x
$$

is the total mass of firms selling in a country ($M$ Home firms and $nM_x$ Foreign firms).


{Impact of trade}
Average productivity across all Home firms is 
$$
\tilde{\varphi} = \left[\frac{1}{1-G(\varphi^*)}\int_{\varphi^*}^{\infty}\varphi^{\sigma-1}g(\varphi) d\varphi  \right]^{\frac{1}{\sigma-1}}
$$

Average productivity across all Home exporters is
$$
\tilde{\varphi_x} = \left[\frac{1}{1-G(\varphi_x^*)}\int_{\varphi_x^*}^{\infty}\varphi^{\sigma-1}g(\varphi) d\varphi  \right]^{\frac{1}{\sigma-1}}
$$



{Impact of trade}
Once we know $\tilde{\varphi}_T$ the aggregate variables of interest are given by,

$$
\begin{align*}
&P = \frac{M_T^{\frac{1}{1-\sigma}}}{\rho \tilde{\varphi}_T} \\
&R = M_t r(\tilde{\varphi}_t) \\
&\Pi = M_T \pi(\varphi_T) \\
&Q = M_T^{\frac{\sigma}{\sigma-1}}q(\tilde{\varphi}_T)
\end{align*}
$$


{Impact of trade}
The equilibrium can be obtained using modified versions of ZCP and FE.  The FE condition is given by
$$
\bar{\pi} = \frac{\delta f_e}{(1-G(\varphi^*))}
$$

where
$$\bar{\pi}=\pi_d(\tilde{\varphi})+n p_x \pi_x(\tilde{\varphi_x})$$

and 
$$
p_x = \frac{1-G(\varphi_x^*)}{1-G(\varphi^*)}
$$

is the probability of exporting conditional on successful entry.


{Impact of trade}
ZCP condition is obtained as follows.  Recall that zero cutoff profit conditions for domestic and export markets:

$$
\begin{align*}
&\pi_d(\varphi^*) = 0 \quad \text{or} \quad \frac{r_d(\varphi^*)}{\sigma} = f \\
&\pi_x(\varphi_x^*) = 0 \quad \text{or} \quad \frac{r_x(\varphi_x^*)}{\sigma} = f_x
\end{align*}
$$

Therefore,
$$
\frac{r_x(\varphi_x^*)}{r_d(\varphi^*)} = \frac{f_x}{f} \quad \text{or} \quad \varphi_x^* = \varphi^* \tau \left( \frac{f_x}{f}\right)^{\frac{1}{\sigma-1}}
$$

The above also verifies that $\varphi_x^* > \varphi^*$ iff $\tau^{\sigma-1}f_x > f$, a condition that was imposed earlier.


{Impact of trade}
Next, using the definition of $\pi_d(\tilde{\varphi})$ and $\pi_x(\tilde{\varphi}_x)$ obtain
$$
\bar{\pi}= \frac{r_d(\tilde{\varphi})}{\sigma}-f+np_x\left( \frac{r_x(\tilde{\varphi}_x)}{\sigma}-f_x \right)
$$

Finally, making the same substitutions as in the closed economy case obtain,
$$
\bar{\pi} = f\left( \left(\frac{\tilde{\varphi}(\varphi^*)}{\varphi^*} \right)^{\sigma-1} -1 \right) + np_x f_x \left( \left(\frac{\tilde{\varphi}_x(\varphi^*)}{\tilde{\varphi}^*_x(\varphi^*)} \right)^{\sigma-1} -1 \right)
$$

The above is the ZCP in the open economy case.

{Impact of trade}
We can measure welfare from the price index
$$
P^{1-\sigma} = M_T (\rho \tilde{\varphi}_T)^{\sigma-1}
$$

As long as $\tau$ is not too high, $M_T > M_a$, and since $\tilde{\varphi}_T > \tilde{\varphi}_a$ (obvious from previous picture), $P$ is lower in the open economy.


Melitz shows that even if $M_T < M_a$, welfare in the open economy is still higher than wlefare in autarky because of productivity effect ($\tilde{\varphi}_T > \tilde{\varphi}_a$) dominates the variety effect.


{Conclusion}

- Empirical evidence shows that firms (even exporters) are very heterogenous and that trade affects the types of firms who survive foreign competition.
- We introduce firm heterogeneity into the Krugman model by assuming firms have different productivities
- Trade causes a selection effect: only the most productive firms export
- There are intra-industry reallocations where the most productive firms survive while the least productive firms shut-down
- Welfare is improved by trade, even when the number of varieties fall, because of the productivity effect
