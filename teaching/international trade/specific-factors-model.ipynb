# Spcific Factors Model

The Heckscher-Ohlin model is useful in identifying how international trade may alter the distribution of income within a country \textit{permanently} (long-run effect).


There are reasons to suspect that the affects on income could be different in the short-run.


In economics, we often define the short-run as being a period of time in which factors are not easily shifted among production activities.


If resources cannot move immediately or costlessly from one industry to another, then the effects of trade on incomes should be different from what we observe in the long-run.


{Specific Factors Model}
The \textbf{specific factors model} is similar to HO, except that we want to remove the assumption that all factors are freely mobile between sectors.


We can think about this as a short-run version of HO theory.


{Environment - Countries, Goods Factors}
$2 \times 2 \times 3$ model:

- Two countries: Home and Foreign
- Two goods: $i=1,2$
- Three factors: labor ($L,L^*)$ capital, $(K,K^*)$, and land $(T,T^*)$.



We assume that labor is a mobile factor that can move between sectors, but capital and land are \textbf{specific factors} that are used only in the production of one good


{Environment - Technology and Market Structure}
We will assume that good 1 required capital and labor, while good 2 requires land and labor.

$$
\begin{align*}
y_1 &= f_1(K,L_1) \\
y_2 &= f_2(T,L_2)
\end{align*}
$$

where $L_1+L_2=L$.


We continue to assume that all market are competitive.

Why is the production possibilities frontier curved?  Diminishing returns to labor in each sector cause the opportunity cost to rise when an economy produces more of a good.

The opportunity cost of producing one unit of good 1 is

$$
\frac{MPL_2}{MPL_1}
$$


How much labor is actually employed in each sector?


Firms demand labor up until the point at which the value of an extra worker is exactly equal to the wage they must be paid:

$$
\begin{align*}
MPL_1 \times p_1 = w \\
MPL_2 \times p_2 = w
\end{align*}
$$

Remember that the supply of labor is fixed at $L$.

So long as both goods are produced we must have that

$$
MPL_1 \times p_1 = MPL_2 \times p_2
$$

so that the economy must be producing at a point such that

$$
\frac{MPL_2}{MPL_1} = - \frac{p_1}{p_2}
$$

We'll perform two experiments:
- an equal proportional change in prices
- a change in relative prices

When both prices change in the same proportion, no real change occurs.
- the wage rate ($w$) rises in the same proportion as prices
- the \textit{real} wage ($w/p_1,w/p_2$) doesn't change
- the real incomes of capital owners and landowners also remain the same

When the relative price of good 1 rises, what is the impact on the distribution of income?
- Owners of capital (specific to good 1) are better off, because the extra labor used in good 1 sector raises the marginal product of capital
- Landowners (specific to good 2) are worse off, because the loss of labor in good 2 sector decreases the marginal product of land
- We cannot say whether workers are better or worse off, because it depends on the relative importance of cloth and food in workers' consumption



Example: Prices, Wages, and Labor
Suppose that the production technology for both sectors is Cobb Douglas:

$$
\begin{align*}
y_1 &= A_1K^{\alpha_1}L^{1-\alpha_1} \\
y_2 &= A_2T^{\alpha_2}L^{1-\alpha_2}
\end{align*}
$$

The marginal products are therefore

$$
\begin{align*}
MPL_1 &= (1-\alpha_1)A_1(K/L_1)^{\alpha_1} \\
MPL_2 &= (1-\alpha_2)A_2(T/L_2)^{\alpha_2} \\
MPK &= \alpha_1 A_1 (L_1/K)^{1-\alpha_1} \\
MPT &= \alpha_2 A_2 (L_2/T)^{1-\alpha_2}
\end{align*}
$$


If labor is reallocated away from good 2 and toward good 1 then we have that $MPL_1 \downarrow, MPL_2 \uparrow, MPK \uparrow, MPT \downarrow$.  We can say that if the relative price of good 1 goes up, then the real income to capital owners increases and the real income to land owners decreases.  We can't say anything definitive about labor because the *real* wage is measured in terms of both good 1 and good 2

$$
\frac{w}{p_1}, \quad \frac{w}{p_2}
$$

so whether or not labor is better off or not depends on the relative importance of good 1 and good 2 in their consumption basket.

Now that we understand how changes in goods prices affects the distribution of income, we are prepared to make a statement about how international trade affects the distribution of income.  We know that the world relative price is determined by the intersection of world relative demand and world relative supply...

Opening up to trade increases the relative supply of good $i$ in an economy whose relatie supply of good $i$ is larger than for the world as a whole.

```{admonition} Specific factors model prediction for the distribution of income

:class: definition

International trade benefits the factor that is specific to the export sector of each country, but hurts the factor that is specific to the import-competing sectors.  Trade has ambiguous effects on mobile factors.

```


Trade benefits the country as a whole because it certainly allows the economy to consume more than it would under autarky. In theory, it possible to redistribute income so that everyone gains from trade.  In reality redistribution is hard to implement, and so there are usually winners and loser from trade. Our short-run model of trade tells us that those who stand to lose from trade are the immobile factors in the import-competing sector.  In reality this includes not only capital, but also a portion of the labor force in those import-competing sectors. Given enough time, those workers can transition from eh import-competing sector to the export-sector, but in the short-run they are likely to suffer. For example, production workers in the apparel sector in the US in 2015 received an average wage about 30 percent lower than the average wage for all production workers. Most economists still favor free trade despite the short-run losses it can impose on import-competing sectors. It is better to allow trade and compensate those who are hurt by it than to prohibit trade. There exist ``safety net" policies like unemployment benefits, and subsidized retraining and relocation programs that can help.


Example: Trade and the Distribution of Income
The sugar industry in the US is very organized and was able to acquire protections against import-competition. Over the past 30 years, the average price of sugar in the US has been about twice the average price on the world market. A 2015 study by the US General Accounting Office estimated that this distorted price distortion cost \$3.5 billion to consumers, or about \$30 per household. The benefits of import restriction is highly concentrated in a small number of producers: seventeen sugar cane farms generate more than half of the profits for the whole sugar cane industry. Those producers are organized in trade associations that actively lobby on their members' behalf, and make large campaign donations.  Consumers, on the other hand, are generally unaware of the cost and \$30 is not enough to incentivize them to organize. Sugar import restriction do save jobs in the sugar industry, but the cost is extremely high: over \$3 million per job saved. Sugar import restrictions also reduce employment in other sectors that rely on large quantities of sugar in their production processes. In response to high sugar prices in the US, many candy-making firms have shifted their production sites to Canada.  On net, the sugar restrictions generate employment *losses* for US workers.


<figure>
    <img src="_static/figs_ho/ho_20.png"
         alt="US employment to population ratio"
         width="600"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
There is no obvious correlation between unemployment and imports for the U.S.
    </figcaption>
</figure>


What about the impact of trade on unemployment in the manufacturing sector specifically?  Many studies have cited China as the major import competitor to US manufacturing firms:
- U.S. industries that compete most heavily with Chinese imports tend to suffer the most employment losses
- high cost of employment losses, especially for workers with relatively lower education and skills
- long unemployment spells and large drops in wages when they find alternative employment
- prolonged negative consequences for regions that rely on manufacturing firms

The job losses are mitigated by increases in employment in the export oriented sectors and by other employers who benefit from cheaper imported intermediate goods. Can we claim that on net Chinese imports hurt American workers?

<figure>
    <img src="_static/figs_ho/ho_21.png"
         alt="US employment to population ratio"
         width="600"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
There is no clear evidence that import competition from China hurt manufacturing employment on net in the US.
    </figcaption>
</figure>


This doesn't minimize the hurt felt by workers, particularly lower income and skill, in import competing firms. One example of a redistribution program is the U.S. Trade Adjustment Assistance program which provides extended unemployment coverage (1 extra year) and tuition reimabursement (for new job skill acquisitions) to workers who are displaced by a plant closure due to import competition.

Two critiques:

- Relatively few workers are able to qualify
- The program discriminates against workers who are displaced due to economic forces other than trade



{Summary}
Key lessons:

- In the long-run trade tends to benefit the factor which is abundant, and hurt the factor that is scarce
- In the short-run trade tends to benefit the factor that is specific to the export sector, and hurt the fact that is specific to the import-competing sector
- Free trade still benefits both countries in aggregate, even though there are affects on the distribution of income within a country
- The groups that are hurt by trade typically are organized and attempt to gain political influence to implement protectionist policies
- Economists tend to favor free trade and support programs that help those groups that were hurt
