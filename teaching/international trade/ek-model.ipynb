# Eaton and Kortum Model of Trade


{Motivation}
Eaton and Kortum (2002) identify four stylized facts that theories of international trade cannot adequately explain:
\begin{enumerate}
- trade diminishes dramatically with distance
- prices vary across locations
- factor reward are far from equal across countries
- countries' relative productivities vary substantially across industries
\end{enumerate}
Facts (1) and (2) suggest that geography matters for predicting the value of trade.  Facts (3) and (4) suggest that technologies differ across countries.


{Motivation}
Also, Ricardian models before EK were not well suited for empirical analysis.

    - Determinisitic comparative advantage: productivities are known and patterns of specialization are rigid; no intensive margin of trade
    - We don't observe the full distribution of productivities across all goods in each country, so we can't get model-implied trade shares to compare with data
    - No heterogeniety within sectors: each good was homogenous which ignored firm-level heterogeneity

One main innovation of EK is that it can generate closed-form expressions for bilateral trade shares based on Ricardian forces.


{Motivation}
What do Eaton and Kortum do?
\vspace{5mm}

``\textit{We develop a Ricardian trade model that incorporates realistic geographic features into general equilibrium.  It delivers simple structural equations for bilateral trade with parameters relating to absolute advantage, to comparative advantage, and to geographic barriers.}


{Gravity}
The key paper which is builds on is Dornbusch, Fischer, and Samuelson (1977), which was the continuum of goods Ricardian model we studies in class.
\vspace{5mm}

Eaton and Kortum add:

- Multiple countries
- Technological heterogeniety
- Geographic barriers



{Gravity}
The reason that we think geography is important, is because there is a long-standing robust empirical finding that predicts the value of trade between two countries based on their size and distance
\[
T_{ij} = \frac{ A \times Y_i \times Y_j}{D_{ij}}
\]
is called a \textbf{gravity equation}.  It says the the value of trade between country $i$ and $j$ is proportional to the product of the two countries' GDPs and diminishes with the distance between the two countries.

{Motivation}
One of the contributions of EK is that gravity equations arise structurally
\vspace{5mm}

Structural gravity equations can also be derived in other classes of models; including classical trade models (Heckscher-Ohlin + Ricardo) and new trade models like Chaney (2008) and Melitz.


{Environment}

- Many countries $i=1,...,N$
- Continuum of goods $j \in [0,1]$
- Countries have access to different technologies so that the productivity of country $i$ in producing good $j$ is given by $z_i(j)$
- Good are produced with a bundle of inputs whose price is homogeneous within countries $c_i$
- There exist iceberg trade costs $d_{ni} \geq 1$ which capture geographic barriers



{Environment}
Consumers choose to purchase individual goods in amounts $Q(j)$ with the aim of maximizing CES utility
\[
U = \left( \int_0^1 Q(j)^{(\sigma-1)/\sigma}dj\right)^{\sigma/(\sigma-1)}
\]


{Environment}
Just as in Ricardo, the cost of producing good $j$ is $c_i/z_i(j)$, where $z$ is the inverse of unit labor requirement.
\vspace{5mm}

With geographic barriers, the cost of delivering a unit of good $j$ produced in country $i$ to country $n$ costs
\[
p_{ni}(j) = \left( \frac{c_i}{z_i(j)}\right)d_{ni}
\]  
where we continue to assume perfect competition so that this is the price a buyer in country $n$ could pay for good $j$ imported from country $i$.


{Environment}
Since there are multiple countries, buyers will purchase good $j$ from the country that produces it at the lowest cost.  So the price a buyer actually pays for good $j$ is
\[
p_n(j) = \min \{p_{ni}(j); i=1,...,N \}
\]


{Production Technology}
EK have a fancy way of modeling technology that allow them to generate structural equations that relate trade values to model parameters for an arbitrary number of countries.
\vspace{5mm}

Stochastic technology where $z_i(j)$ is draw from a country specific distribution $F_i(z)$ which is independent across goods and countries.
\vspace{5mm}

By law of large numbers, $F_i(z)$ is also the fraction of goods for which country $i$'s productivity is below $z$.


{Production Technology}
Since productivity $z_i(j)$ is a r.v., so too is the cost of purchasing good $j$ from country $i$
\[
p_{ni}(j) = \frac{c_i d_{ni}}{Z_i(j)}
\]
and therefore the price that the buyer will actually choose is also a r.v.
\[
P_n(j) = \min \{p_{ni}(j); i=1,...,N \}
\]
The liklihood that country $i$ supplies a particular good $j$ to country $n$ is the probability $\pi_{ni}(j)$ that $i$'s price turns out to be the lowest.


{Production Technology}
EK assume that $F_i(z)$ is Frechet distributed so that the expression for $\pi_{ni}$ is simple and the distribution of prices can be computed.
\[
F_i(z) = e^{-T_iz^{-\theta}}, \quad T_i>0,\theta>1
\]
where $T_i$ is a country specific parameter that governs location and $\theta$ is a common parameter which governs dispersion.

- Bigger $T_i$ ('technology'/absolute advantage) implies higher productivity draws are more likely
- Bigger $\theta$ ('heterogeneity'/comparative advantage) implies less variability


%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
{Production Technology}
For clarity, let's understand this modeling choice of stochastic productivity within the continuum of goods model we studied earlier.

    - We had technology $A(z) = a^*(z)/a(z)$
    - You can show, using the ratio of Frechet random variables, that in EK 
    $$
    A(z)=\frac{T_1}{T_2}^{1/\theta}\left(\frac{1-z}{z}\right)^{1/\theta}
    $$
    - Changes in parameter $T$ shift the $A(Z)$ curve while changes in $\theta$ change the slope



{Production Technology}
We can compute the distribution of prices now:
\begin{align*}
G_{ni}(p)=Pr[p_{ni}(j)<p] &= Pr\left[\frac{c_i d_{ni}}{Z_i(j)}\right] \\
&=Pr\left[\frac{c_i d_{ni}}{p} < Z_i(j)\right] \\
&=1-F_i\left(\frac{c_i d_{ni}}{p}\right) \\
&= 1-e^{-[T_i(c_i d_{ni})^{-\theta}]p^{\theta}}
\end{align*}


{Production Technology}
And we can compute the distribution of prices for what country $n$ actually buys.  That is, distribution for 
\[
p_n(j) = \min \{p_{ni}(j); i=1,...,N \}
\]
Intuition: The probability that at least one $p_{ni}$ is smaller than $p$ is equal to the complement probability that all $p_{ni}$ are greater than $p$.  So we have that
\begin{align*}
G_n(p) &= 1-\prod_i [1-G_{ni}(p)] \\
&=1-e^{-\Phi_n p^\theta}
\end{align*}
where $\Phi_n =\sum_{i=1}^N T_i (c_i d_{ni})^{-\theta}$ is a parameter that summarizes
\begin{enumerate}
- states of technology around the world
- input costs around the world
- geographic barriers around the work
\end{enumerate}


{Production Technology}
There are some general lessons from this price distribution:

- Prices tend to be lower the better is technology $\uparrow T_i$
- Prices tend to be lower the lower are input prices $\downarrow c_i$
- Prices tend to be lower the less are geographic barriers $\downarrow d_{in}$



{Production Technology}

- In the extreme case of free trade ($d_{in} \rightarrow 1$), price parameter $\Phi$ is the same everywhere and the law of one price holds
- In the extreme case of autarky ($d_{in}\rightarrow \infty$), we have $\Phi_n=T_n c_n^{\theta}$: country $n$'s state of technology weighted by its input cost



{Production Technology}
EK can now compute the probability that country $i$ provides the good at the lowest price in country $n$
\begin{align*}
\pi_{ni} &= Pr[P_{ni}(j) \leq \min \{P_{ns}(j); s \neq i \}] \\
&= \int_0^{\infty} \prod_{s \neq i}[1-G_{ns}(p)]dG_{ni}(p) \\
 & =\frac{T_i (c_i d_{ni})^{-\theta}}{\Phi_n}
\end{align*}
by the law of large numbers this is also the fraction of goods that country $n$ buys from country $i$


{Trade Flows, and Gravity}
The fraction of goods that country $n$ buys from country $i$, given by $\pi_{ni}$ is also the fraction of its expenditure on goods from country $i$
\[
\frac{X_{ni}}{X_n} = \frac{T_i (c_i d_{ni})^{-\theta}}{\Phi_n} = \frac{T_i (c_i d_{ni})^{-\theta}}{\sum_{i=1}^N T_i (c_i d_{ni})^{-\theta}}
\]
which we can use to write the exporter's total sales $Q_i$
\[
Q_i = \sum_{m=1}^{N}X_{mi} = T_i c_i^{-\theta}\sum_{m=1}^N \frac{d_{mi}^{-\theta }X_{m}}{\Phi_m}
\] 


{Trade Flows, and Gravity}
You can show that the price index for the CES aggregator is given by
\[
p_n = \gamma \Phi_n^{-1/\theta}
\]
which is used to write the following expression for country $n$ imports from country $i$
\[
X_{ni}=\frac{\left( d_{ni}/p_n\right)^{-\theta}X_n}{\sum_{m=1}^{N}\left( d_{mi}/p_m\right)^{-\theta}X_m}Q_i
\]
which is a standard gravity equation where both exporter's total sales $Q_i$ and importer's total purchases $X_n$ enter with unit elasticity.  This linearity crucially depends on the functional forms of technology.


{Trade Flows, and Gravity}
\[
X_{ni}=\frac{\left( d_{ni}/p_n\right)^{-\theta}X_n}{\sum_{m=1}^{N}\left( d_{mi}/p_m\right)^{-\theta}X_m}Q_i
\]

- Geographic barrier $d_{mi}$ is deflated by the importer's price level $p_m$
- Tougher competition in country $m$ reduces $p_m$, reducing importer $i's$ access in the same way as a higher geographic barrier
- We can interpret the term $(d_{mi}/p_m)^{-\theta}X_m$ as the market size of destination $m$ as perceived by country $i$.
- The denominator is therefore the total world market from $i's$ perspective



{Trade Flows, and Gravity}
There are some key differences in terms of how EK's bilateral trade volumes behave differently from monopolistically competitive environments:

- Trade costs affect bilateral trade through the technology parameter $\theta$ rather than a preference parameter $\sigma$
- Bilateral trade responds at the extensive margin rather than the intensive margin: as a source becomes more expensive it exports a narrower range of goods; different from monopolistic competition where the range of goods is unaffected by less is traded.



{Trade Flows, and Gravity}
After normalizing by exporters share at home we have
\[
\frac{X_{ni}/X_n}{X_{ii}/X_i}=\frac{\Phi_i}{\Phi_n}d_{ni}^{-\theta}=\left( \frac{P_i d_{ni}}{P_n}\right)^{-\theta}
\]

- the left side can be constructed using trade data and gross output data (EK use a sample of 19 OECD countries)
- The right side can use data on geographic distance and retail prices, and then estimate the impact of comparative advantage through $\theta$



{Trade Flows, and Gravity}
\[
\frac{X_{ni}/X_n}{X_{ii}/X_i}=\frac{\Phi_i}{\Phi_n}d_{ni}^{-\theta}=\left( \frac{p_i d_{ni}}{p_n}\right)^{-\theta}
\]

- As prices in market $n$ fall relative to prices in market $i$, $i$'s share in $n$ declines
- As $n$ becomes more isolated from $i$, $i$'s share in $n$ declines
- As comparative advantage weakens (higher $\theta$), import shares become more elastic

{Trade Flows and Distance}
A simple negative correlation between trade and distance is provocative, but subject to major limitations

- Ignores the price indices in the gravity equation
- Omits information on comparative advantage $\theta$
- Distance may not be a good proxy for trade costs

The observed correlation could arise under strong comparative advantage (low $\theta$) if trade costs rise rapidly with distance; or under high $\theta$ where comparative advantage is weak.
\vspace{5mm}

Need to identify $\theta$ for cleaner insights.


{Trade Flows and Prices}
To identify $\theta$, use price data to measure the term
\[
\frac{p_i d_{ni}}{p_n}
\]
The exact measurements of relative prices used in the paper can be found on page 1753, but for our purposes it's enough to just think of the righthandside of the normalized gravity equation.

{Trade Flows and Prices}

- Correlation coefficient of -0.40 confirms the negative relationship between prices and trade predicted by the theory
- Since the model implies no intercept term, the parameter $\theta$ is simply the slope coefficient of a log linear regression of the lefthandside on the righthandside gravity equation
\[
\hat{\theta}=-8.03
\]
which implies a standard deviation in efficiency (for a given $T$) of 15\%.




{Identifying Comparative Advantage $\theta$}
At this point, we have a gravity equation that can quantitatively show the impact of comparative advantage versus geographic barriers in impacting trade.
\vspace{5mm}

EK go further and close the model by solving for factor prices and then estimate the resulting trade flows for manufacturing sector trying to find an estimate for $\theta$


{Identifying Comparative Advantage $\theta$}
I omit the details (pagers 1755-1759), but include the updated gravity equation for reference:
\[
\frac{X_{ni}}{X_n}=T_i \left( \frac{\gamma d_{ni}w_i^{\beta}p_i^{1-\beta}}{p_n}\right)^{-\theta}
\]
where $\gamma$ is the parameter from CES preferences, $\beta$ is labor share from a Cobb-Douglas production function, $w$ is the wage rate.


{Identifying Comparative Advantage $\theta$}
EK then incorporate the following information:

- Use $\beta=.21$ which is the average labor share in manufacturing in their sample
- Estimate $d_{ni}$ using a linear regression on distance, border effects, shared language, belonging to a trading area, and country fixed effects (See equation (29) in the paper for the exact regression specification)
- Capture $T_i$ and $w_i$ as source-country dummy variables $S_i$ which is a measure of a country's competitiveness
- Allow for asymmetric trade costs between country pairs (see equation 30 for details)

{Conclusion}

- A clever and parsimonious way of introducing Ricardian forces into a multi-country model
- Delivers a structural gravity equation which can be used to estimate different countries comparative advantage and the role that geographic barriers play
- Analytical solutions rely on some key assumptions: Frechet distribution, variance of productivity is the same across industries
- Estimation results are sensitive to the estimation strategy
