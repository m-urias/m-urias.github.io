# The Heckscher-Ohlin Model of Trade


The Ricardian model assumes the source of comparative advantage is differences in labor productivity.  In the real world, however, trade also reflects differences in a countries' resources. For example, Canada exports forest products to the United States not because its lumberjacks are more productive relative to the US, but because Canada has more forested land per capital than the United States.


Key Questions:
- How can countries with identical technology and preferences can gain from trade?
- Who are the winners and losers from international trade?
- What predictions of the model are consistent with empirical evidence?


To explain the role of resources in international trade, we develop an economic model that shows that comparative advantage is influenced by the interaction of
- the relative \textbf{abundance} of resources and
- how \textbf{intensively} a country's technology uses resources to produce goods


We will observe that trade effects the \textbf{distribution of income} within a country: some people are made better off from trade, and some are made worse off.


## The Heckscher-Ohlin Model

2 x 2 x 2 Model:
- Two countries: Home and Foreign
- Two goods: $i = 1,2$
- Two factors: labor ($L, L^*$) and capital ($K,K^*$)


**Production function** describes how factors are transformed into goods

$$
y_i = f_i(L_i,K_i)
$$

The production functions are assumed to be increasing, concave, and homogeneous of degree one.

Assume that labor and capital are fully mobile between the two industries, but immobile across countries.

Let's introduce some extra notation that is useful:
- $a_{K1} \equiv$ capital used to produce one unit of good 1
- $a_{L1} \equiv$ labor used to produce one unit of good 1
- $a_{K2} \equiv$ capital used to produce one unit of good 2
- $a_{L2} \equiv$ labor used to produce one unit of good 2

Although these are similar to unit input requirements defined in the Ricardian model (for labor at least), there is an important difference: we are talking about the quantity of capital or labor \textit{used} in production, not the amount of capital or labor \textit{required}.

Remember that when there are two inputs, producers have some choice about what combination to use them in to create goods.  The profit maximizing choice $(a_{Ki},a_{Li})$ will depend on the input price $(w,r)$.

Assume that preferences are identical and homothetic across countries so that relative demand is a function of relative price only,

$$
\left( \frac{y_1}{y_2} \right)^d = f\left(\frac{p_1}{p_2} \right) \quad \text{where} \quad f'(\cdot) <0
$$

Goods market and labor market are perfectly competitive.

- Workers will be paid a wage rate equal to its marginal product of labor, and capital owners will be paid a rental rate equal to its marginal product of capital.

$$
w = MPL \quad \text{and} \quad r = MPK
$$

- Firms set marginal cost equal to price which equals average cost so that economic profits are zero.

$$
p = MC = AC
$$


An Example:
Let's consider an economy where the technology is such that there is \textit{no possibility of substituting labor for capital or vice versa}.

- Two goods: Cloth (C) and Food (F)
- Production of C required 2 work-hours and 2 machine-hours
- Production of F required 1 work-hour and 3 machine hours
- Let the economy have 3,000 machine hours and 2,000 work hours

We write:

$$
a_{KC}=2, \quad a_{LC}=2, \quad a_{KF}=3, \quad a_{LF}=1
$$

and 

$$
L=2000, \quad K=3000
$$


In this special case of no factor substitution, the economy's PPF is derived from the constraints on capital and labor

$$
\begin{align*}
\text{Capital:}& \quad a_{KC} y_C + a_{KF} y_F \leq K \quad \Rightarrow 2y_C + 3y_F \leq 3000 \\
\text{Labor:}& \quad a_{LC} y_C + a_{LF} y_F \leq L  \quad \Rightarrow 2y_C + y_F \leq 2000
\end{align*}
$$

The economy is must obey **both** constraints.

When the economy is producing mostly food, the opportunity cost of cloth is $2/3$).  When the economy is producing mostly cloth, the opportunity cost of cloth is $2$.  The opportunity cost of producing cloth changes when the economy's mix of production shifts.


{Production Possibilities Frontier}
Now let's make the model more realistic and allow the possibility of substituting capital for labor and vice versa in production.

Labor and capital can be used in any combination across industries so long as the resource constraint is satisfied

$$
\begin{align*}
&L_1 + L_2 \leq L \\
&K_1 + K_2 \leq K
\end{align*}
$$

Possible production plans are given by the following program

$$
\begin{align*}
\max y_2 = f_2(L_2,K_2) \qquad \text{s.t.} \quad y_1 = f_1(L_1,K_1) 
\end{align*}
$$

Where on the PPF does the economy actually produce? It depends on **goods prices**: the economy will always produce at the point that maximizes the value of production.
- If the price of good 1 is relatively high, the economy will produce more of good 1 and less of good 2
- If the price of good 2 is relatively high, the economy will produce more of good 2 and less of good 1

The value of production (GDP) is given by

$$
Y = p_1 y_2 + p_2 y_2
$$



Formally, the gross domestic production (GDP) for the economy is given by

$$
G(p_1,p_2,L,K) = \max_{y_1,y_2} \hspace{2mm} p_1 y_1 + p_2 y_t \qquad \text{s.t.} \quad y_2 = h(y_1,L,K)
$$

The first order condition for this problem is,

$$
p = \frac{p_1}{p_2} = -\frac{\partial h}{\partial y_1} = -\frac{\partial y_2}{\partial y_1}
$$

(We will show that this must be the case for a competitive market in a worksheet.)


Given goods prices, the equilibrium production plan will maximize total output.

Given a level of production, firms still have a choice about what combination of capital ($K$) and labor ($L$) they want to use to produce each good.  That is, they still need to choose $(a_{1L},a_{2L},a_{1K},a_{2K}$).


What combination of inputs do produces actually use?  It depends on **factor prices**: producers will minimize cost by using more of the factor that is relatively cheaper.
- If capital rental rates ($r$) are high and wages ($w$) low, producers will choose to use relatively little capital and a lot of labor
- If wages are low and rental rates are high, producers will choose to use relatively little labor and a lot of capital

It will be convenient to work with the \textbf{unit cost functions} which are dual to the production functions,

$$
c_i(w,r) = \min_{L_i,K_i \geq 0} \hspace{2mm} \{wL_i + rK_i \vert f_i(L_i,K_i) \geq 1\}
$$

Tells us the minimum cost to produce one unit of output.

We will write the *solution* to the minimization problem above as,

$$
c_i(w,r) = w a_{iL} + r a_{iK}
$$

where $a_{iL}$ is the optimal choice for $L_i$ and $a_{iK}$ is the optimal choice for $K_i$.


For any given factor prices $(w,r)$, we can find the cost minimizing combination of labor and capital $(a_{iL},a_{iK})$.

As drawn in the previous example, the input choice curve for good 1 is shifted out compare to good 2.  This means that for any relative factor prices good 1 requires relatively more labor than capital compared to good 2.

- Good 1 is **labor-intensive**
- Good 2 is **capital-intensive**

Mathematically we recognize that this implies

$$
\frac{a_{1L}}{a_{1K}}>\frac{a_{2L}}{a_{2K}}
$$

$\Rightarrow$ a good cannot be both capital- and labor-intensive.


Notice that our definition of **intensity** relies on the ratio of labor to capital used in production, not the ratio of labor or capital to output.

```{admonition} Definition: Intensity

:class: definition

Good 1 is said to be labor intensive if 

$$
\frac{a_{1L}}{a_{1K}}>\frac{a_{2L}}{a_{2K}}
$$

it must then be that Good 2 is capital intensive since

$$
\frac{a_{2K}}{a_{2L}}>\frac{a_{1K}}{a_{1L}}
$$

```
There is a connection between the prices of goods and the prices of inputs.  Due to free entry under perfect competition equilibrium profits must equal zero,

$$
\begin{align}
p_1 &= c_1(w,r) \\
p_2 &= c_2(w,r)
\end{align}
$$
There must be full employment of factors in equilibrium,

$$
\begin{align}
&\underbrace{a_{1L}y_1}_{L_1} + \underbrace{a_{2L}y_2}_{L_2} = L \\
&\underbrace{a_{1K}y_1}_{K_1} + \underbrace{a_{2K}y_2}_{K_2} = K
\end{align}
$$


```{admonition} Definition: Intensity

:class: definition

An equilibrium is a set of goods prices $(p_1,p_2)$, factor prices $(w,r)$ and outputs $(y_1,y_2)$ such that profits are zero and factors are fully employed.

```
**Factor Price Insensitivity**: So long as both goods are produced, and factor intensity reversals (FIRs) do not occur, then each price vector $(p_1,p_2)$ corresponds to unique factor prices $(w,r)$.  If FIRs do occur, then an equilibrium can only be determined by also using the full employment conditions.

We have the following logic to connect goods prices and factor prices:
- Suppose the relative goods price is $p_1/p_2$
- From the $FG$ curve we know what is the competitive wage to rental rate $w/r$
- Given $w/r$, from the $FI_i$ curves we know what the input choices of producers are going to be $\frac{a_{2K}}{a_{2L}}>\frac{a_{1K}}{a_{1L}}$



(Note that this logic employs two standard economic theories: perfect competition guarantees zero economic profits, and cost minimization guarantees produces pick the least expensive mix of inputs.)

We already have an insight about how goods prices affect the distribution of income in the economy.

```{admonition} Stolper-Samuelson Theorem

:class: definition

An increase in the relative price of a good will increase the real return to the factor used intensively in that good, and reduce the real return to the other factor.

```

If we think that trade will affect relative goods prices, then we should suspect that trade will affect the distribution of income too...

Begin with zero profit conditions,

$$
p_i = w a_{iL} + r a_{iK}
$$

Totally differentiate to obtain,

$$
dp_i = a_{iL}dw + a_{iK} dr.
$$

Judiciously multiply by one and divide by the price

$$
\frac{dp_i}{p_i}=\frac{w a_{iL}}{c_i}\frac{dw}{w}+\frac{r a_{iK}}{c_i}\frac{dr}{r}
$$

which is useful because it allows us to express variables in terms of percentage changes ($d ln x = dx/x$) and cost shares

- $\theta_{iL}=\frac{w a_{iL}}{c_i}$ is the cost share of labor in industry $i$
- $\theta_{iK}=\frac{r a_{iK}}{c_i}$ is the cost share of capital in industry $i$

We have that

$$
\hat{p}_i=\theta_{iL}\hat{w}+\theta_{iK}\hat{r}
$$

which is a two-by-two system of linear equations which can be written as

$$
\begin{pmatrix}
\hat{p}_1 \\
\hat{p}_2
\end{pmatrix}
= \begin{pmatrix}
\theta_{1L} & \theta_{1K} \\
\theta_{2L} & \theta_{2k}
\end{pmatrix}
\begin{pmatrix}
\hat{w} \\
\hat{r}
\end{pmatrix}
$$

using Cramer's rule we have that,
\begin{align*}
\hat{w} &= \frac{\theta_{2K}\hat{p}_1-\theta_{1K}\hat{p}_2}{\lvert \Theta \rvert} = \frac{(\theta_{2K}-\theta_{1K})\hat{p}_1+\theta_{1K}(\hat{p}_1-\hat{p}_2)}{\theta_{2K}-\theta_{1K}} \\
\hat{r} &= \frac{\theta_{1L}\hat{p}_2-\theta_{2L}\hat{p_1}}{\lvert \Theta \rvert} = \frac{(\theta_{1L}-\theta_{2L})\hat{p}_2-\theta_{2L}(\hat{p}_1-\hat{p}_2)}{\theta_{1L}-\theta_{2L}}
\end{align*}


Suppose that sector 1 is labor intensive

$$
\theta_{1L}>\theta_{2L}
$$

so that $\lvert \Theta \rvert>0$; and suppose that the relative price of good 1 increases

$$
\hat{p_1}>\hat{p}_2
$$

Then we have that

$$
\begin{align*}
\hat{w} > \hat{p}_1 \\
\hat{r} < \hat{p}_2
\end{align*}
$$

which shows that wages increase by more than the price of good 1, so we can say that the \textit{real wage} has indeed increased.  Conversely, the \textit{real return to capital} has fallen.

The last ingredient of our model is to understand the relationship between a countries resources ($L$ and $K$) and the output it produces.

We already established the following:

- At a given relative goods price $p_1/p_2$, we know there is an equilibrium wage to rental rate $w/r$
- At a given wage to rental rate, there is an optimal mix of labor to capital employed in each sector $(a_{iL}/a_{iK})$.

Now let's assume that the labor force grows $\uparrow L$.  How can the economy accommodate the extra labor and maintain the optimal mix of inputs $(a_{iL}/a_{iK})$?  Let's continue with the example where good 1 is labor intensive.  Then an increase in the workforce $\uparrow L$ can be accommodated by allocating more labor *and* capital to the production of good 1. This way, the producer's mix of labor to capital remains unchanged, even though the economy now has a greater supply of labor relative to capital than before.  As labor and capital move toward good 1 sector, the economy produces more of good 1 and less of good 2.


<figure>
    <img src="_static/figs_ho/fig15.png"
         alt="US employment to population ratio"
         width="600"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    </figcaption>
</figure>


```{admonition} Rybzynski Theorem

:class: definition

If relative goods prices are constant and if both commodities continue to be produced, an increase in a factor endowment will increase the output of the industry using it intensively, and decrease the output of the other industry.

```

Mathematical Proof
Begin with full employment conditions,

$$
\begin{align*}
a_{1L} y_1 + a_{2L} y_2 = L \\
a_{1K} y_1 + a_{2K} y_2 = K
\end{align*}
$$

Totally differentiate to obtain,

$$
\begin{align*}
a_{1L}dy_1 + a_{2L} dy_2 = dL \\
a_{1K}dy_1 + a_{2K} dy_2 = dK
\end{align*}
$$

We now have the following system,

$$
\begin{pmatrix}
dL \\
dK
\end{pmatrix}
= \begin{pmatrix}
a_{1L} & a_{2L} \\
a_{1K} & a_{2K}
\end{pmatrix}
\begin{pmatrix}
dy_1 \\
dy_2
\end{pmatrix}
$$

using Cramer's rule we have that,

$$
\begin{align*}
\frac{dy_1}{dL} &= \frac{a_{2K}}{a_{1L}a_{2K}-a_{2L}a_{1K}} \\
\frac{dy_2}{dL} & = \frac{-a_{1K}}{a_{1L}a_{2K}-a_{2L}a_{1K}}
\end{align*}
$$

Without loss of generality assume that industry 1 is labor intesntive

$$
\frac{a_{1L}}{a_{1K}} > \frac{a_{2L}}{a_{2K}}
$$

then we have that

$$
\frac{dy_1}{dL} > 0, \quad \frac{dy_2}{dL} <0
$$


Generally, an economy will tend to be relatively effective at producing goods that are intensive in the factors with which the country is relatively well endowed.


Finally we can look at the effects of international trade on this economy.  We assume that Home and Foreign has the same preferences and the same technology.  We allow for Home and Foreign to have *different resources*.  Let's assume that Home has a higher ratio of labor to capital than Foreign does.
- Home is **labor abundant**
- Foreign is **capital abundant**

and we'll continue to assume that good 1 is **labor intensive** and good 2 is **capital intensive**.

Heckscher-Ohlin Theorem: Home becomes an exporter of good 1 because it is labor abundant and because the production of good 1 is labor intensive.  Foreign becomes an exporter of good 2 because it is capital abundant and because the production of good 2 is capital intensive.

```{admonition} Heckscher-Ohlin Theorem

:class: definition

Each country will export the good that uses its abundant factor intensively.

```

In the more realistic case with multiple countries, factors of production, and numbers of goods, we can say something more general: *Countries tend to export goods whose production is intensive in factors with which the countries are abundantly endowed.*


Trade and the Distribution of Income
We just saw how trade will cause relative prices to converge, so that Home (who has a comparative advantage in good 1) will realize in increase in the relative price of good 1, while Foreign (who has a comparative advantage in good 2) while realize a decrease in the relative price of good 1.


A rise in the price of good 1...

- ...raises the purchasing power of labor in terms of both goods
- ...decreases the purchasing power of capital in terms of both goods

In Home, laborers benefit from trade while capital owners are hurt.  In Foreign, capital owners benefit from trade while laborers are hurt.


Trade has important consequences for the distribution of income within a country.



```{admonition} Hecksher-Ohlin's prediction on the distribution of income

:class: definition

Owners of a country's abundant factors gain from trade, but owners of a country's scarce factors lose.

```
Opening to trade still expands an economy's consumption possibilities, representing aggregate gains from trade, but it's not obvious how to practically distribute these gains to *both* factors.


Recall that trade tensions between the US and China largely arose from claims that manufacturing workers in the US were hurt by import competition from China.  Compared with the rest of the world, the US is abundantly endowed with high-skilled labor while low-skilled labor is relatively scarce.  This means that international trade has the potential to make low-skilled workers in the US worse off---not just temporarily, but permanently.  Clearly the manufacturing sector in the US has an incentive to protect its industry from import competition, but this would come at the expense of aggregate gains from trade for the country as a whole.


Factor Price Equalization
We just saw how trade impacts the distribution of income *within* a country.  What about *between* countries?


In the absence of trade, labor-abundant Home would have a power relative price of good 1 than capital-abundant Foreign, and the difference in relative prices of *goods* would be reflected in  a difference in the relative price of *factors*.

When Home and Foreign trade, the relative price of goods converges.  Therefore, the relative factor prices converge too!

```{admonition} Factor Price Equalization Theorem

:class: definition

Free trade tends to equalize factor prices between countries.

```
Although Home has a higher ratio of labor to capital than Foreign, once they trade with each other, the wage and rental rate are the same in both countries.

When Home and Foreign trade, more is happening than a simple exchange of goods.

- In an indirect way, the two countries are effectively trading factors of production
- Home is letting Foreign use some of its abundant labor, not by selling labor, but by trading goods produced with a high ratio of labor to capital for goods produced with a low labor-capital ratio.
- Labor is \textbf{embodied} in Home exports more than its imports
- Capital is embodied in Foreign exports more than its imports

In reality, factor price equalization does not occur.  Why? The model makes some stark assumptions:
- technologies are the same in each country
- transportation of goods is costless so that goods prices are the same in each country
- both countries produce both goods



In reality, none of these assumption are likely to hold.


The Hechscher-Ohlin model reveals an important source of mutually beneficial trade: differences in factor abundance.  Even if we don't observe factor price equalization in reality, it doesn't mean that differences in resources don't matter for trade.  It could simply mean that a better model of trade has to relax some of the strong assumption Heckscher and Ohlin made.


We can still try to use data on international trade to test whether or not the mechanism Hecksher and Ohlin is observed in reality.


## Empirical Tests of the Heckscher-Ohlin Theory

{Empirical Tests of HO Theory}
Until recently, the US had a reputation for possessing much more capital per person than other countries.


According to the HO theory, the US should export goods that are relatively capital intensive and import goods that are relatively labor-intensive.


If we had data that showed the relationship the amount of capital and labor embodied in US imports and exports, then we could test the prediction of the theory.


{Empirical Tests of HO Theory}
Economists Wassily Leontif was the first to conduct such a study (published in 1953) using data on the 25 years following World War II.


Leontif used input-output tables for the US to compute the amount of capital and labor embodied in the exports and imports of goods...


{Empirical Tests of HO Theory}
\begin{figure}
\centering
\includegraphics[scale=.45]{ho_12}
\end{figure}


{Empirical Tests of HO Theory}
Leontif found that the ratio of capital to labor embodied in US exports was actually less than imports.


This is the opposite of what HO would predict!


This famous result is known as the \textbf{Leontif paradox}


{Empirical Tests of HO Theory}
There are several reasons why the prediction of HO model could fail:

- technologies are not the same in the US and elsewhere
- Leontif ignores other factors like human capital and land
- the US was not engaged if \textit{free} trade


More importantly, Leontif was actually using the incorrect test.


{Empirical Tests of HO Theory}
Leamer (1980) formalized the correct test of the Heckscher-Ohlin model.  
We want to develop a test for the HO model.

- Many countries $i=1,...,C$
- Many industries $j=1,...,N$
- Many factors $k = 1,...,M$

Let the $(M \times N)$ matrix $A = [a_{jk}]'$ denote the amounts of factor $k$ required to produce one unit of good $j$.


Example:

$$
\begin{pmatrix}
a_{1L} & a_{2L} \\
a_{1K} & a_{2K}
\end{pmatrix}
$$



{Heckscher-Ohlin-Vaneck}
Let the $(N \times 1)$ matrix $Y^i$ denote the amount of ouptut in each industry for each country $i$.


Let the $ (N \times 1)$ matrix $D^i$ denote the amount demanded for each good.


The \textbf{factor content of trade} is given by,

$$
F^i \equiv A\underbrace{(Y^i-D^i)}_{\text{net exports}}
$$

An individual component of this matrix is $F_k^i$.

- If $F_k^i >0$ then the factor is exported
- If $F_k^i <0$ then the factor is imported



{Heckscher-Ohlin-Vaneck}
We want to relate the factor content of trade to endowments.


Full-employment condition says that all endowments of a country must be used in the production of goods:

$$
\text{endowment vector:}\hspace{5mm} V^i = AY^i
$$

Identical and homothetic preferences with balanced trade imply that consumption in any country is proportional to its share of world GDP:

$$
\text{consumption vector:}\hspace{5mm} AD^i = s^i AD^w = s^iAY^w = s^i V^w
$$



{Heckscher-Ohlin-Vaneck}
\begin{tcolorbox}
Heckscher-Ohline-Vaneck (HOV) theorem: The factor content of trade in factor $k$ should be positive ($F^i_k > 0$) if that factor is abundant ($V_k^i/V_k^w > s^i$), and conversely negative ($F^i_k < 0$) if that factor is scarce ($V_k^i/V_k^w < s^i$).
\end{tcolorbox}


{Resolving the Leontif Paradox}
Let's revisit Leontif's test where there are only two factors:
\begin{align*}
&F_k^i = K^i - s^iK^w \\
&F_l^i = L^i - s^iL^w
\end{align*}
If the US is capital abundant, relative to labor, then the HOV theorem predicts:

$$
\frac{K^{US}}{L^{US}} > \frac{K^{US}-F_{k}^{US}}{L^{US}-F_{L}^{US}}
$$

The capital/labor ratio embodied in production for the US should exceed the capital/labor ratio embodied in consumption.


{Heckscher-Ohlin-Vaneck}
PROOF:
From (i) our definition of the factor content of trade; (ii) identical homothetic preferences; and (iii) market clearing we have that

$$
K^w = \frac{K^i-F_k^i}{s^i} \quad \text{and} \quad L^w = \frac{L^i-F_l^i}{s^i}
$$

It follows that 

$$
\frac{K^i}{K^w} = s^i \frac{K^i}{K^i-F_k^i} \quad \text{and} \quad \frac{L^i}{L^w} = s^i  \frac{L^i}{L^i-F^i_l}
$$

If $K^i/K^w > L^i/L^w$ then it must be that

$$
\frac{K^i}{K^i-F_k^i} > \frac{L^i}{L^i-F^i_l}
$$



{Partial Tests of HOV}
The HOV theorem tells us how to do a test of the theory

$$
F^i \equiv A T^i = V^i-s^i V^W
$$


- compute the left-hand side using data on trade $T^i$ and technology $A$
- compute the right-hand side using data on endowments $V^i$ and $V^w$
- compare how well the two sides match up



{Partial Tests of HOV}
There are two ways to test how well the HOV theory performs:

- Sign test: $\text{sign}(F_k^i) = \text{sign}(V^i_k-s^i V_k^w)$
- Rank test: $F_k^i > F_l^i \Leftrightarrow (V_k^i - s^i V_k^w) > (V^i_l-s^i V_l^w)$

for $\quad i=1,...,C; k=1,...,M$


Of course, you can perform these tests by running the regression

$$
F_k^i = \alpha_k + \beta_k(V_k^i-s^i V_f^w)+\varepsilon_k^i
$$

and observing the sign and magnitude of $\beta_k$ coefficients.


{Why Leontif's test was wrong}
Consider this example (courtesy of ChatGPT):


- \textbf{Two goods:} Machinery (M), Textiles (T) 
- \textbf{Two factors:} Capital (K), Labor (L)
- \textbf{Country:} U.S. (capital-abundant)


\medskip
\textbf{Input coefficients (units per \$1M output):}
\begin{center}
\begin{tabular}{lccc}
\toprule
Good & Capital & Labor & K/L Ratio \\
\midrule
Machinery (Exported) & 300 & 100 & 3.0 \\
Textiles (Imported) & 600 & 100 & 6.0 \\
\bottomrule
\end{tabular}
\end{center}

{Leontief's Method: Per-unit Comparison}

    - Compare capital-labor ratio of export vs. import goods
    - Machinery: $K/L = 3.0$
    - Textiles: $K/L = 6.0$


\begin{block}{Conclusion}
U.S. appears to import more capital-intensive goods than it exports.\\
$\Rightarrow$ Leontief Paradox
\end{block}


{The HOV Approach: Factor Content of Trade}
Suppose:

    - U.S. exports \$2M of Machinery
    - U.S. imports \$1M of Textiles



$$
T = 
\begin{bmatrix}
2 \\
-1
\end{bmatrix}, \quad
A = 
\begin{bmatrix}
300 & 600 \\
100 & 100
\end{bmatrix}
$$



$$
A T =
\begin{bmatrix}
(2)(300) + (-1)(600) \\
(2)(100) + (-1)(100)
\end{bmatrix}
=
\begin{bmatrix}
0 \\
100
\end{bmatrix}
$$




{HOV Interpretation}
\textbf{Net factor exports:}

    - Capital: 0
    - Labor: 100 units


\begin{block}{Finding}
The U.S. is effectively exporting labor, not capital — consistent with Leontief’s finding.
\end{block}



{New Scenario: Larger Export Volume}
Suppose:

    - Export \$10M of Machinery
    - Import \$2M of Textiles



$$
T = 
\begin{bmatrix}
10 \\
-2
\end{bmatrix}
\quad \Rightarrow \quad
A T = 
\begin{bmatrix}
(10)(300) - (2)(600) \\
(10)(100) - (2)(100)
\end{bmatrix}
=
\begin{bmatrix}
1800 \\
800
\end{bmatrix}
$$

\begin{block}{Finding}
So the U.S. is exporting both capital and labor — but more capital than before. The overall export bundle is capital-abundant, despite the lower capital-labor ratio per good.
\end{block}



{Why HOV is Better}
Leontief's method misses the scale of trade and makes inferences based solely on intensity, not on quantities. The HOV approach corrects this by looking at the total factor content of trade, which is what the H-O theory actually predicts.



{Empirical Tests of HO Theory}
A study by Bowen, Leamer, and Sveikauskas extended Leontif's test to 27 countries and 12 factors of production.


For each country and factor they tested whether or not the factor content of trade matched the HO prediction.


The prediction was correct in only 61 percent of cases.  The Leontif paradox is still there.


{Empirical Tests of HO Theory}

Trefler (1995) used 33 countries and 9 factors to test both the direction of factor content of trade and the magnitude

- the sign test was consistent with HO 50 percent of cases
- the rank test was consistent with HO 60 percent of cases

Trefler performs several diagnostic tests on the data and determines that the assumption of equal technologies across countries is especially bad.


{Empirical Tests of HO Theory}
Davis and Weinstein (2001) show that if you relax the assumption of common technology, as well as countries producing the same set of goods and costless trade, then the predictions of the model are much better:
\begin{figure}
\centering
\includegraphics[scale=.50]{ho_13}
\end{figure}


{Empirical Tests of HO Theory}
We can see how the predictive failure of the HO model rested on particular assumptions built into the purest version of the model.


When those assumptions are dropped, the predictive power of the model is vastly improved.


You can decide for yourself whether it is really the HO model at that point.
