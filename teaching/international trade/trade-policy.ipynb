# International Trade Policy


{Trade Restrictions}
Many countries impose some type of trade restriction, usually under pressure from special interest groups that may be hurt by trade, or to exert political pressure.


We will evaluate the costs and benefits, welfare effects, and the winners and losers of tariffs.


{Trade Restrictions}
There are several types of trade protections that countries use:
- \textbf{tariffs}: a tax on imports or exports- \textbf{export subsidies}: government pays exporters- \textbf{voluntary export restraints}: exporting country agrees to limit exports (usually under pressure)- \textbf{import quotas}: limits the quantity of a specific import- \textbf{regulations/customs/licensing}: importer must obtain licenses to buy certain products/safety regulations on imports/bureaucratic rules



{Tariffs}
Tariffs are the most common form of trade restriction.
- An import tariff is a tax on the imported commodity- An export tariff is a tax on the exported commodity


Many developing countries rely on export tariffs (e.g. Ghana on its cocoa and Brazil on its coffee) to get better prices and raise revenues.


Some developed countries do not allow the use of export tariffs but do use import tariffs (e.g. the US Constitution prohibits export tariffs).


{Tariffs}
Tariffs can be ad valorem, specific, or compound.
- \textbf{Ad valorem tariffs} are taxes levied as a fraction of the value of the imported good (e.g. 40 percent Chinese tariff on imported cigarettes)- \textbf{Specific tariffs} are levied as a fixed amount for each unit of goods imported (e.g. \$505 per tonne of aluminum Chinese tariff)- \textbf{Compound tariff} is a combination of an ad valorem and a specific tariff. 



{Types of trade protections}
China categorizes import tariffs into six groups:
- \textbf{general rates}: tariff rates applied to imports from countries that do not have a trade agreement or MFN status with the importing country- \textbf{most-favored-mation (MFN) rates}: standard, non-discriminatory tariffs applied to all WTO members. Under the WTO principle of non-discrimination, members must extend the same MFN rate to all other members.- \textbf{agreement rates}: negotiated tariff rates that apply to imports under specific trade agreements, such as Free Trade Agreements (FTAs) or Customs Unions. They are often lower than MFN rates.



{Types of trade protections}
- \textbf{preferential rates}: reduced or zero tariffs offered unilaterally or under special schemes to certain countries, especially developing or least-developed countries. They are not reciprocal and aim to promote trade and development.- \textbf{tariff rate quota rates}:  two-tiered tariff system with a lower (in-quota) tariff rate is applied to a set quantity (quota) of imports and a higher (out-of-quota) tariff rate applies to imports exceeding the quota limit.- \textbf{provisional rates}: temporary tariff rates, often introduced while final rates are being negotiated or while a formal review or investigation is ongoing (e.g., an anti-dumping or safeguard investigation).



{Types of trade protections}
The dutiable value of an imported good is its cost, insurance, and freight (CIF) price, which includes the normal transaction price of the good, plus the cost of packing, freight, insurance, and seller's commission. China Customs is tasked with assigning a fair value to all imports.


This is different from the free on board (FOB) price which is includes the cost of goods at the point of shipment; including cost of goods sold, loading charges, and export documentation


Buyers either agree to FOB Origin or FOB Destination pricing: FOB Origin means the sellers liability ends as soon the goods are loaded onto the transportation where FOB Destination means that sellers are liable until the goods are delivered.

By July 1, 2023, China's official average import tariff rate will be 7.3 percent.

{Tariffs}
When we do a theoretical analysis of tariffs on an economy, we will be doing \textbf{partial equilibrium} analysis.  That is, assume that trade policies toward one sector can be understood well enough without going into detail about the policies effects on the rest of the economy.


If we think that the effects on the entire economy are critical to understand a trade policy, then we need to use \textbf{general equilibrium} analysis.  


All of our models of trade (Ricardo, Heckscher-Ohlin, New Trade) were general equilibrium since investigated how impacts to one industry affected all other industries too.

{Tariffs}
If Home imposes a specific tariff of \$2 on every bushel of wheat imported, shippers will be unwilling to move the wheat unless the price differences between the two markets is at least \$2.


If no wheat is being shipped, however, there will be an excess demand for wheat in Home and excess supply in Foreign. Therefore, the price in Home will rise and the price in Foreign will fall.


The price in Home will rise and the price in Foreign will fall until the price difference is equal to the tariff $\tau$.


{Tariffs}
A specific import tariff $\tau$ imposed by Home drives a wedge between the prices in the two markets:
- a tariff raises the price in Home from $P_W$ to $P_T$- a tariff lowers the price in Foreign to $P_T-\tau$

Home producers supply more at a higher price, while consumers demand less, so fewer imports are demanded.  Foreign producers supply less at a lower price, while consumers demand more, so there is a smaller export supply.


{Tariffs}
The tariff results in
- lower volume of wheat traded $Q_{\tau}<Q_W$- a wedge in between the Home and Foreign price $P_T-P^F = \tau$



This analysis assumes that Home is an important part of the wheat industry, so it's decision to impose a tariff impacted Foreign's price.  This is why the rise in Home's price is not equal to the tariff $\tau$.


If Home were a small player in the wheat industry, then it's tariff will have little (approximately zero) effect on on Foreign's export price.  In this case, Home's price rise is equal to the tariff $\tau$.



{Effective Rate of Protection}
The goal of a tariff is often to protect an import-competing industry, resulting in an increase in the domestic price of that good.


In practice, we want to measure the amount of protection a tariff affords an industry.  One useful metric is \textit{the percentage of the price that would prevail under free trade.}


Example: An import tariff on sugar raised the domestic market price of sugar by 35 percent compared to the free trade price.
- If the tariff is ad valorem, the tariff rate itself measures the amount of protection.- If the tariff is specific, dividing the tariff by the price net of the tariff gives an ad valorem equivalent



{Effective Rate of Protection}
There are two problems with trying to calculate the rate of protection this way:
\begin{enumerate}- The small-country assumption may not be a good approximation of reality, so part of the effect of a tariff will be to lower foreign export rather than raise domestic import prices.- Tariffs may have very different effects on different stages of production of a good
\end{enumerate}


{Effective Rate of Protection}
Consider the following example to see why tariffs may have different effects on different stages of production of a good:
- Imagine that a car sells for \$8,000 in the world market, and the parts out of which the car is made sell for \$6,000- To encourage domestic car industry, Home levies a 25\% tariff, allowing domestic car assemblers to charge \$10,000 instead of \$8,000



It is incorrect to say that car assemblers receive 25 percent protection.
- Before the tariff, domestic assembly takes place only if it can be done for \$2,000 (the different between \$8,000 of a completed car and \$6,000 cost of parts)- After the tariff, domestic assembly takes place even it costs as much as \$4,000- The tariff provides assemblers with an effective rate of protection of 100 percent!



{Effective Rate of Protection}
- Now suppose that Home imposes a 10\% tariff on imported parts, raising the cost of parts for domestic assemblers from \$6,000 to \$6,600- Before the tariff, it would have been worth assembling a car locally if it could be done for \$2,000- After the tariff, local assembly takes place only if it can be done for \$1,400- The tariff on parts, provides protection to parts manufacturers, but -30\% (600/2,000) protection to car assemblers 



{Effective Rate of Protection}
Similar reasoning leads economists to make complex calculations to measure the degree of effective protection provided by trade protection policies.


The effective rate of protection for an industry is formally

$$
\frac{V_T-V_W}{V_W}
$$

where $V_W$ is the value added to the industry at world prices; $V_T$ is the value added in the presence of trade policies.


In our example of Home's 25\% import tariff on cars we have

$$
\frac{4,000-2,000}{2,000} = 100\%
$$



{Effective Rate of Protection}
Many industrial nations have a ``cascading" tariff structure with very low or zero tariffs on raw materials and higher and higher rates the greater is the degree of processing.
- By making finished imports more expensive, countries encourage firms to process raw materials locally rather than importing finished products- Finished goods often have higher prices, so taxing them at higher rates generates more tariff revenue for the government- The rate of effective protection on a final commodity with imported inputs much greater than the nominal tariff rate would indicate

- Consumers lose in the importing country and gain in the exporting country- Producers gain in the importing country and lose in the exporting country- The government imposing the tariff gains revenue



{Costs and Benefits of a Tariff}
In order to compare these costs and benefits, we need to quantify them using two additional concepts in economics:
\begin{enumerate}- \textbf{Consumer surplus}: the amount a consumer gains from being able to purchase a good at a lower price than he would have been willing to pay- \textbf{Producer surplus}: the amount a producer gains from being able to sell its good at a higher price than he would have been willing to sell it for
\end{enumerate}

{Trade Warfare}
Retaliation against protectionist trade policies is a real concern.  


Example: After Trump imposed punitive tariffs on China, China responded with more tariffs of its own.


Trade wars ruin so many gains achieved through free trade.  So why do they happen?


{Trade Warfare}
Suppose there are two countries: the United States and Japan.


Suppose that there are two policy choices: free trade or protection.


We can represent the strategic interaction of these countries in a game matrix.
\begin{figure}
\centering
\includegraphics[scale=.6]{tp_17}
\end{figure}


{Trade Warfare}
The payoffs represent two assumptions:
- Each country's government would choose protection if it could take the other country's policy as given (e.g. terms of trade argument, political interests, supporting domestic industry)- Both countries would be better off if the both engaged in free trade (there exist gains from trade)



This is an application of the \textbf{Prisoner's dilemma}.  Each country has a dominant strategy to protect itself from import competition, and the resulting Nash equilibrium is one where each country is worse off.


This is the rationale for why \textbf{trade agreements} (and the WTO) can be useful.


{Trade Warfare}
Given our example of trade negotiations, it may seem puzzling why  countries would ever agree to free trade at all!


We can justify this in the same game theoretic setting by appealing to the idea of repeated interactions: games are played several times and players understand this.


Let's extend the Prisoner's Dilemma to multiple periods and evaluate when free trade can be sustained as a subgame perfect Nash equilibrium.

- Both countries have the same discount factor $\beta \in (0,1)$
- Both countries adopt a \textbf{grim trigger strategy}: cooperation ends permanently if there is any defection



{Trade Warfare}

Payoff from always cooperating:

$$
V_C = \sum_{t=0}^\infty \beta^t \cdot 10 = \frac{10}{1 - \beta}
$$


Payoff from defecting once and triggering permanent retaliation:

$$
V_D = 20 + \sum_{t=1}^\infty \beta^t \cdot (-5) = 20 + \frac{-5\beta}{1 - \beta}
$$


Cooperation is sustainable if:

$$
\frac{10}{1 - \beta} \geq 20 - \frac{5\beta}{1 - \beta}
$$


Which simplifies to:

$$
\beta \geq \frac{2}{5}
$$



{Trade Warfare}
If both countries care enough about the future, they will avoid a trade war and sustain free trade.


A grim trigger strategy may to to harsh and unrealistic.  Instead, we could suppose the countries adopt a \textbf{forgiving trigger strategy}: punish defection with Protection for a fixed number of round $T$, then return to Free Trade.


{Trade Warfare}
Payoff from defecting once, facing $T$ rounds of retaliation, then returning to cooperation:
\begin{align*}
V_D &= 20 + \sum_{t=1}^{T} \beta^t \cdot (-5) + \sum_{t=T+1}^\infty \beta^t \cdot 10 \\
&=20 - 5 \left( \frac{1 - \beta^T}{1 - \beta} \right) + 10 \left( \frac{\beta^{T+1}}{1 - \beta} \right)
\end{align*}

Cooperation is sustainable if:

$$
\frac{10}{1 - \beta} \geq 20 - 5 \left( \frac{1 - \beta^T}{1 - \beta} \right) + 10 \left( \frac{\beta{^T+1}}{1 - \beta} \right)
$$



{Trade Warfare}

- Longer punishment period $T$ makes cooperation easier to sustain
- Higher discount factor $\beta$ also support cooperation
- More realistic than grim trigger: After a violation (like a tariff hike), the injured party responds with temporary retaliatory tariffs, possibly escalating to the WTO or ending in negotiation. Once resolved, countries usually restore normal trade rather than ending cooperation forever.



{Trade Warfare}
There are many other strategies you can consider:

- Tit-for-Tat: Begin with cooperation, then do whatever your opponent did in the previous rounds
- Tit-for-$T$-Tat: Begin with cooperation, then retaliate if your opponent defects $T$ times in a row
- Win-Stay, Lose-Shift: If you received a good payoff last round, repeat your action.  Otherwise, switch your action.
- Mixed strategy trigger: After defection, punish with some probability.

Which strategy is actually employed depends on many factors including: patience, risk of noise, reputation, communication, enforcement, number of players etc.


{Trade Warfare}
Trade agreements and institutions (like the WTO) act as formal mechanisms to facilitate cooperation and reduce the need for harsh unilateral punishment.


{Measuring the cost of tariffs}
The partial equilibrium analysis we investigated is also used by policy makers and academics to measure the real-world cost of tariffs.


{Measuring the cost of tariffs}
World Integrated Trade Solution (WITS) provides a simulation tool called SMART which performs partial equilibrium analysis of trade policy changes at the industry level.  (Have a look: https://wits.worldbank.org/wits/wits/witshelp/Welcome.htm and click on "SMART").  You can specify a tariff (addition or reduction) for a particular industry and it will estimate:

- Trade creation effect
- Trade diversion effect
- Price effect effect
- Tariff revenue
- Welfare change

The software defaults to an infinitely elastic export supply curve, but you can change this assumption.


{Measuring the cost of tariffs}
Amiti, Redding, and Weinstein (2019) use partial equilibrium analysis to measure the impact of the 2018 US trade war.


\textit{The deleterious impacts of the tariffs have been largely in line with what one might have predicted based on a simple supply and demand framework. We estimate the likely impact on U.S. consumers and find that by the end of 2018, import tariffs were costing U.S. consumers and the firms that import foreign goods an additional \$3 billion per month in added tax costs and another \$1.4 billion dollars per month in deadweight welfare (efficiency) losses. Tariffs have also changed the pricing behavior of U.S. producers by protecting them from foreign competition and enabling them to raise prices and markups.}


{Measuring the cost of tariffs}
ARW (2019) first compute the 1 year change of a price index for a bundle of imports affected by the Trump 2018 tariffs
$$
\hat{P}_t=\prod_{\omega \in \Omega}\hat{p}_{ijt}(\omega)^{s_{ij}(\omega)}
$$
where $\hat{p}_t=p_t/p_{t-12}$ and $s_{ij}$ is the logarithmic mean of import shares  from country $j$ in sector $i$ among all HTS10 imports in categories affected by tariff wave $\omega$.


(Note: This is the exact price index for CES demand system)


They also compute the same index but use total value of imports rather than unit values.

{Trade policy uncertainty}
One potentially important cost of tariffs, which is ignored in this study, is the effect of trade policy uncertainty on the decision of exporters to enter the world market.


Handley and Limao (2017) develop a dynamic model where firms must pay of a sunk entry cost and there exists an option value of waiting due to trade policy uncertainty.  They show that, indeed, trade policy uncertainty deters investment by exporting firms and entry of firms into the export sector.

Autor, Dorn, and Hanson (2013) measure how exposed regional labor markets in the US were to import competition from China

$$
\Delta IPW_{it}=\sum_{j}\frac{L_{ij,1990}}{L_{j,1990}}\frac{\Delta M_{jt}}{L_{it}}
$$

where 

- $(L_{ij}/L_{j})$ is the share of region $i$ in the US employment in industry $j$
- $\Delta M_{jt}$ is the observed change in US imports from China in industry $j$ between the start and end of the period
- $L_{it}$ is the start of period employment (year $t$) in region $i$

The formula shows that a region is more exposed to Chinese import growth if it specializes (has a higher-than-average employment share) in industries where U.S. imports from China are rising. The import growth is scaled by the size of the region.

In order to isolate the supply-driven China shock (control for U.S. domestic shocks), ADH (2013) use the following instrument for change in imports: Chinese exports to high-income countries (excluding the US).


The instrumental variable approach aims to identify the component of US import growth that is due to Chinese productivity and trade costs.  The identifying assumption underlying this strategy is that the common within-industry component of rising Chinese imports to the United States and other high-income countries is due to China’s rising comparative advantage and falling trade costs.

{The China Shock}
Main findings:

- Rising Chinese imports negatively impacted workers who were exposed the most to import competition
- Manufacturing workers suffered higher unemployment and discouraged workers
- Non-manufacturing workers realized lower wages suggesting general equilibrium effects within a region
- The results, taken together, suggest that US labor markets are not so flexible to absorb trade induced sectoral reallocation
- At the national level, gains from trade are almost surely positive; however, localized costs are large and persistent
