# Ricardian Model of International Trade

The Ricardian model is our first attempt at explaining why countries trade with each other.  We use it to identify the source of mutually beneficial trade, and also predict which type of goods will be exported and imported by different countries.   Although a simple model, it provides some powerful insights that are still relevant in policy debates today. Is foreign competition unfair when it is based on low wages? Is free trade only beneficial if your country is strong enough to stand up to foreign competition? Does trade exploit a country and make it worse off when its workers receive much lower wages than workers in other nations?  

Even after hundreds of years of trade theory, and a near universal consensus among economists about the merits of free trade, policy makers continue to make speeches denouncing the growing imports of manufactured goods from developing countries, claiming they are displacing domestic workers from jobs.  Policymakers in the US, for example, cite the persistent current account deficit between the US and China as a major problem and seek to implement trade protections (e.g. tariffs) to reduce this trade imbalance. The claim is that the US is ``getting a bad deal" and losing wealth since it is importing more than it is exporting.  This rhetoric sounds eerily similar to the neo-mercantilists hundreds of years ago.  Take Thomas Munn, England's Treasurer in 1682: *The ordinary means therefore to increase our wealth and treasure is by Foreign Trade, wherein we must ever observe this rule; to sell more to strangers yearly than we consume of theirs in value. For . . . that part of our stock exports which is not returned to us in wares imports must necessarily be brought home in treasure bullion...* 

Even with a simple model, we can begin to scrutinize such claims and develop arguments for or against them.  By the end of this chapter you should:
- Explain how the Ricardian model of trade works and how it illustrates the principle of **comparative advantage**.
- Demonstrate gains from trade and refute common fallacies about international trade
- Describe empirical evidence that wages reflect productivity and that trade patterns reflect relative productivity.

After discussing the Ricardian model, we then turn to another classical model developed by Heckscher and Ohlin.  With both models in hand, we can add nuance to our arguments for and against free trade by discussing who are the winners from free trade and trade protection policies.  



## Motivating Numerical Example

The overarching lesson from classical trade theories is the following: Nations (and individuals) can benefit from their differences by reaching an arrangement in which each does the things it does relatively well.  The greater are the differences, the greater benefit.

Before we rigorously demonstrate this by means of a theoretical model, let's look at a numerical example to illustrate the main lesson.

Adam Smith first proposed the theory of **absolute advantage** as a response to the mercantilist idea that trade is a zero-sum game.

Wheat (bushels/hour) & 6 & 1  \\ \hline
Cloth (yards/hour) & 4 &  5 \\ \hline
\end{tabular}
\end{table}
The US has an absolute advantage in the production of Wheat, and Mexico has an absolute advantage in the production of Cloth.

## The Ricardian Model

Now we develop the theoretical model to generalize the results from our numerical example.

We assume that there are
- Two countries: Home and Foreign
- Two goods: $i=1,2$
- One factor: labor at Home is $L$ and in Foreign $L^*$


Production technology is described by \textbf{unit labor requirements}:

- $a_i \equiv$ labor need per unit of good $i$ at Home
- $a_i^* \equiv$ labor need per unit of good $i$ at Foreign

We assume that labor is perfectly mobile between industries, but immobile across countries.


Assume that preferences are identical and homothetic across countries.
- Homotheticity guarantees that ratios of goods demanded only depends on relative prices and not income or scale (CES preferences for example)
- Identical preferences are assumed in order to show that technological differences can be a source of mutually beneficial trade



Because each country has a limited number of workers, there are limits to the amount of goods that can be produced.  The feasible production plans are represented by a **production possibilities set**:

$$
\begin{align*}
\text{Home:} \quad &a_1 y_1 + a_2 y_2 \leq L \\
\text{Foreign:} \quad &a_1^* y_1^* + a_2^* y_2^* \leq L^*
\end{align*}
$$

The **opportunity cost** of good 1 is how many units of good 2 must stop being produced in order to make one more unit of good 1.  


Producing one additional unit of good 1 requires $a_1$ hours of labor.  Each hour devoted to the production of good 1 could have been used instead to produce an amount of good 2 equal to $1/a_2$.


The opportunity cost of good 1 in terms of good 2 is 

$$
\frac{a_1}{a_2}.
$$



Since labor is the only factor of production, the relative supply of goods is determined by whether workers choose to produce good 1 or good 2.  Workers will choose to work in the industry with higher wages.

- The hourly wage of workers producing good 1 is $p_1/a_1$.
- The hourly wage of workers producing good 2 is $p_2/a_2$.



Both goods are produced if and only if 

$$
p_1/a_1 = p_2/a_2.
$$


A reminder about equilibrium wages.  Consider the profit of a firm:
 
$$
\pi_i = p_i q_i - w a_{iL}q_i
$$

and recognize that if $p_i/a_{iL} < w$ then the price is too low for firms to survive so no amount of good $i$ is produced. Conversely if $p_i/a_{iL} > w$ then firms will want to produce good $i$ and competition will force the price down until $p_i/a_{iL}=w$.  The lesson is that if good $i$ is produced, then it must be the case that $w/p_i = 1/a_{iL}$.



In the absence of international trade, the relative prices of goods are equal to their relative unit labor requirements.

$$
\frac{p_1}{p_2} = \frac{a_1}{a_2}
$$

We may say that the \textbf{relative price} of Good 1 in terms of Good 2 is equal the opportunity cost of Good 1 in terms of Good 2.

{Example}
Suppose each country produces Cloth (C) and Wine (W).
\begin{table}[h!]

\begin{tabular}{|l|l|}
\hline
$a_{C} = 3 \hspace{2mm} \text{hours}$ & $a_{C}^* = 5 \hspace{2mm} \text{hours}$  \\ \hline
$a_{W} = 2 \hspace{2mm} \text{hours}$ &  $a_{W}^* = 1 \hspace{2mm} \text{hour}$ \\ \hline
\end{tabular}
\end{table}

- The relative price of C in terms of W at Home is $a_C/a_W = 3/2$
- The relative price of C in terms of W at Foreign is $a_C^*/a_W^* = 5$
- The relative price of W in terms of C at Home is $a_W/a_C = 2/3$
- The relative price of W in terms of C at Foreign is $a_W^*/a_C^* = 1/5$


{Example cont.}
Suppose that the price of Cloth is \$4 per pound and price of Wine is \$7 per bottle.

- Wage paid to a Home worker producing Cloth is $p_C/a_c = \$4/3 = \$1.33$
- Wage paid to a Home worker producing Wine is $p_W/a_w = \$7/2 = \$3.5$

Home workers are willing to make only Wine because the relative price of Wine 7/4 exceeds the opportunity cost of producing Wine 3/2.  In order for both Wine and Cloth to be produced, the price of Wine must drop to \$2.67 or the price of Cloth must rise to \$10.5.


```{admonition} Definition: Comparative Advantage

:class: definition

A country has a **comparative advantage** in the production of good i if its **opportunity cost** of producing good i is smaller than the other country.
- If $a_1/a_2 < a_1^*/a_2^*$ then Home (Foreign) has a comparative advantage in the production of good 1 (good 2).
- If $a_1/a_2 > a_1^*/a_2^*$ then Foreign (Home) has a comparative advantage in the production of good 1 (good 2).

A country has a **comparative advantage** in the production of good i if the **autarky relative price** of good i is less than the other country.
- If $p_1/p_2 < p_1^*/p_2^*$ then Home (Foreign) has a comparative advantage in the production of good 1 (good 2).
- If $p_1/p_2 > p_1^*/p_2^*$ then Foreign (Home) has a comparative advantage in the production of good 1 (good 2).

```

Suppose that Home is more efficient in the production of both goods

$$
a_1 < a_1^* \qquad \text{and} \qquad a_2 < a_2^*
$$

then Home is said to have an **absolute advantage** in the production of both goods.


A country can be more efficient producing both goods, but it will have a comparative advantage in only one good.  Even if a country is the most (or least) efficient producer of all goods, it still can benefit from trade.


Let's allow both countries to trade with each other.  Each country will benefit by completely specializing in the production of the good for which it has a comparative advantage and trading.

<figure>
    <img src="_static/figs_ric/fig2.png"
         alt="Gains from trade"
         width="600"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    An illustration of the gains from trade.  Points $(A,A^*)$ are the autarky points, points $(B,B^*)$ are the production points under free trade, and $(C,C^*)$ are the consumption points under free trade.
    </figcaption>
</figure>

To determine the equilibrium world price, let's begin by defining the **world relative supply curve**.  Assume that Home has a comparative advantage in the production of good 1.
- For any world relative price $p^W < p $ and $p^W < p^* $ both countries completely specialize in good 2 so the world relative supply of good 1 is zero.
- For $p < p^W < p^*$ Home is specialized in good 1 and Foreign is specialized in good 2 so the world relative supply is $\frac{L/a_1}{L^*/a_2^*}$.
- For $p^W > p$ and $p^W > p^*$ both countries are specialized in good 1.

<figure>
    <img src="_static/figs_ric/fig3.png"
         alt="World relative supply curve"
         width="600"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    The world relative supply curve.
    </figcaption>
</figure>

<figure>
    <img src="_static/figs_ric/fig4.png"
         alt="World equilibrium"
         width="600"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    The world equilibrium.
    </figcaption>
</figure>


Each country benefits since it can import the good at a lower price compared to autarky.

If demand is particularly strong for one of the goods, the country that does not have a comparative advantage in that good will not gain from trade, but does not lose from trade either.

Smaller countries are more likely to gain from trade.  All else equal, smaller countries...

- ...have better terms of trade
- ...are more resilient to shocks



## Ricardian Model with Many Goods

To move a bit closer to reality, let's extend the model to include multiple goods $i=1,2,...N$.  For each good, we can calculate the ratio of Home's unit labor requirement to Foreign's and order them,

$$
a_1/a_1^* < a_2/a_2^* < ... < a_N/a_N^*.
$$

Goods will always be produced where it cheapest to make them.  It will be cheaper to produce the good in Home if 

$$
w a_i < w^* a_i^*
$$

where $(w,w^*)$ are the wages at Home and Foreign.  This implies that any good which satsifies

$$
a_i^*/a_i > w/w^*
$$

will be produced at Home.  We have already ordered the goods in terms of the ratio of Home's unit labor requirement to Foreign's.  Therefore, all the goods to the left of $w/w^*$ will be produced at Home and the remaining goods will be produced at Foreign.

<figure>
    <img src="_static/figs_ric/fig5.png"
         alt="Unit labor requirements"
         width="600"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    </figcaption>
</figure>

<figure>
    <img src="_static/figs_ric/fig6.png"
         alt="Equilibrium relative wage"
         width="600"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    The equilibrium relative wage.
    </figcaption>
</figure>

We can extend the model further to allow for a continuum of goods $z \in [0,1]$.  Let $a(z)$ and $a^*(z)$ be the unit labor requirements for good $z$ in Home and Foreign respectively.  We can always order the good such that there exists a function $A(z) \equiv a^*(z)/a(z)$ such that $A'(z) <0$.

There will be a cutoff value $\tilde{z}$ for any given relative wage which determines the pattern of specialization.
- if $A(z) > w/w^*$ then Home specializes  in $z$
- if $A(z) < w/w^*$ then Foreign specialized in $z$

<figure>
    <img src="_static/figs_ric/fig7.png"
         alt="World relative supply curve"
         width="600"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    Pattern of specialization for a given relative wage
    </figcaption>
</figure>

Let preferences be given by 

$$
U = \int_0^1 c(z)^{b(z)} dz
$$

where $b(z) = \frac{p(z)c(z)}{Y}$ expenditure share on good $z$.  Of course we must have that $\int_0^1 b(z) dz = 1$.  Suppose that $\tilde{z}$ is the cutoff good.  Then define

$$
\phi(\tilde{z}) = \int_0^{\tilde{z}} b(z) dz
$$

to be the share of expenditure on goods produced domestically.  To determine the equilibrium relative wage we use the fact that total income equals total expenditure in equilibrium,

$$
\begin{align*}
wL &= \phi(\tilde{z})(wL + w^* L^*) \\
\omega L &= \phi(\tilde{z})(\omega L + L^*) \\
\omega &= \frac{\phi(\tilde{z})}{1-\phi(\tilde{z})}\frac{L^*}{L} \\
\omega &= f(\tilde{z},L^*/L)
\end{align*}
$$

where $\omega=w/w^*$.  The $f$ schedule shows that if the range of domestically produced goods were increased at constant relative wages, demand for domestic labor (goods) would increase and demand for foreign labor (goods) would decline.  A rise in the domestic relative wage would then be required to equate the demand for domestic labor to the existing supply.  


An alternative interpretation of the $f$ schedule is that it shows the trade balance equilibria

$$
(1-\phi(\tilde{z}))wL = \phi(\tilde{z})w^* L^*
$$

An increase in the range of goods produced domestically at constant relative wages lowers home imports and raises home exports. The resulting trade imbalance would have to be corrected by an increase in home's relative wage that would raise import demand and reduce exports, and thus restore balance.

<figure>
    <img src="_static/figs_ric/fig8.png"
         alt="World relative supply curve"
         width="600"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    Free trade equilibrium.
    </figcaption>
</figure>

What are equilibrium prices? We know that,

$$ p(z)/p(z')=wa(z)/wa(z')=a(z)/a(z') \quad \text{for} \quad z,z' < \tilde{z}$$


$$ p(z)/p(z'')=wa(z)/w^*a^*(z'')=\omega a(z)/a^*(z'')  \quad \text{for} \quad z < \tilde{z}, z'' >\tilde{z}$$

so that relative goods prices are determined once we have equilibrium $(\omega,\tilde{z})$.

What is the effect of an increase in the relative size of Foreign?
<figure>
    <img src="_static/figs_ric/fig9.png"
         alt="World relative supply curve"
         width="600"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    Increase in the size of Foreign's labor force.
    </figcaption>
</figure>

What is the effect of an increase in the relative size of Foreign?
- At the initial equilibrium, the increase in Foreign labor would create an excess supply of labor abroad and an excess demand for labor at home: a trade surplus.
- Domestic relative wages must increase to eliminate the trade surplus while at the same time raising relative unit labor costs at Home.
- Higher labor costs at Home implies a loss of comparative advantage and thus fewer commodities are produced domestically.

What is the effect of technological improvements in Foreign?
<figure>
    <img src="_static/figs_ric/fig8.png"
         alt="World relative supply curve"
         width="600"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    Increase in Foreign's productivity.
    </figcaption>
</figure>

What is the effect of technological improvements in Foreign?
- At the initial equilibrium, a reduction in Foreign labor costs implies fewer commodities produced at Home and a corresponding trade deficit.
- Domestic relative wages must decrease to eliminate the trade deficit and offsets parts of the decline in comparative advantage.


What is the effect of a demand shift away from high $z$ goods to low $z$ goods?
<figure>
    <img src="_static/figs_ric/fig9.png"
         alt="World relative supply curve"
         width="600"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    Demand shock toward low $z$ goods.
    </figcaption>
</figure>

- Equilibrium relative wage rises
- Range of goods produced at home shrinks
- Although home income certainly increased, we cannot say anything about welfare since preferences have changed

The Ricardian model proposes the following features of international trade:

- Gains from trade arise from technological differences.
- Each country exports the good for which it has a comparative advantage.
- The world equilibrium price must lie in between autarky prices.
- The smaller country stands to gain more from trade.
- More productive countries are rewarded with higher wages.


{Extensions}
The model can be extended to include:

- Non-traded goods
- Transport costs
- Tariffs
- Multiple countries (a bit difficult but we will see later in the course...)

### Empirical Evidence for the Ricardian Model

Few economists believe that the Ricardian model is a sufficient description of the causes and consequences of world trade, but it's two principles are powerful ideas that are still relevant to the real world:

- Productivity differences matter for the pattern of trade
- It is comparative advantage rather than absolute advantage that matters


One difficulty with testing the predictions of Ricardo is that the growth of world trade and resulting specialization means that we do not get a chance to see what countries do badly---we can't measure the productivity in sectors that don't exist.

Ricardian theory of comparative advantage can explain the emergence of countries with very low overall productivity as export powerhouses in some industries.

- Bangladesh is a very poor country, with extremely low overall productivity, but it's exports of clothing was booming and catching up to China (the dominant supplier at the time)
- Even though Bangladesh was less productive than China in every industry, its  disadvantage was much smaller in clothing

<figure>
    <img src="_static/figs_ric/Emp_Ricardo.png"
         alt="World relative supply curve"
         width="600"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    The world relative supply curve.
    </figcaption>
</figure>

There was a set of studies conducted comparing America and Britain in the post-World War II period.

- British labor productivity was lower than American productivity in almost every sector (US had an absolute advantage in nearly everything)
- British exports were about as large as American exports
- There must have been some sectors in which Britain had a comparative advantage
- The Ricardian model predicts that these would be sectors in which America's productivity advantage was smaller

<figure>
    <img src="_static/figs_ric/Emp_Ricardo3.png"
         alt="World relative supply curve"
         width="600"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    MacDougall (1951). *British and American Exports: A Study Suggested by the Theory of Comparative Costs.*
    </figcaption>
</figure>

Golub (1995) looked at relative unit labor costs (the ratio of wages to unit labor productivity) and the exports of the US relative to those of UK, Japan, Germany, Canada, Australia.

<figure>
    <img src="_static/figs_ric/Emp_Ricardo4.png"
         alt="World relative supply curve"
         width="600"
         style="display: block; margin-left: auto; margin-right: auto;">
    <figcaption style="font-style: italic; text-align: left;">
    MacDougall (1951). *British and American Exports: A Study Suggested by the Theory of Comparative Costs.*
    </figcaption>
</figure>

These results were confirmed in a follow up study by Golub and Hsieh (2000).

{Some caveats to keep in mind}
In reality there may exist several frictions that are not accounted for by our simple model:

    - Goods markets are not competitive
    - Wages are not flexible
    - Labor is imperfectly mobile across industries

However, if you want to take empirical facts seriously, you need to move beyond the simple Ricardian model anyway.
