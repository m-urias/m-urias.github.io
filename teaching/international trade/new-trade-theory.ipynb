# New Trade Theory


{Introduction}
The classical models of trade tell us that countries specialize in the production of the good for which they have a comparative advantage.

- Differences in technology (Ricardo)
- Differences in resource endowments (Heckscher-Ohlin)
- Differences in preferences (no model, just obvious)


Comparative advantage is all about taking advantages of differences between countries, and letting each country produce the good at the cheapest cost.



{Introduction}
According to classical theories:

- Different countries should trade more with each other
- Countries should specialize in the production of different goods
In reality, we see that:

- Most trade occurs between similar countries
- Countries tend to trade similar goods

Roughly half of all world trade involves shipping goods between the fairly similar high-income economies.

{Gravity Equation}
Economists have found that the following \textbf{gravity equation} does a good job of predicting the volume of trade between any two countries:

$$
T_{ij}=\frac{A \times Y_i^a \times Y_j^b}{D_{ij}^c}
$$

where $Y_i, Y_j$ are are two countries' GDPs and $D_{ij}$ is the distance between countries.


Why does the equation work?

- Larger economies tend to spend more an imports because they have large incomes
- Large countries tend to attract other countries' spending because they tend to produce a wider range of products.

{``New" Trade Theory}
``New" trade theory argues that economies of scale can explain these empirical facts.


Economies of scale (or increasing returns to scale) result when large firms have a cost advantage over small firms---the average cost of production decreases the more output is produced.


Economies of scale can emerge either within a company (internal) or within an industry (external).

How might economies of scale provide an incentive for trade?




{An example of economies of scale}
Plan A:

- Suppose each country produces 10 cars, so total care production is 20
- Each country requires 15 labor, so total labor use is 30


Plan B:

- The U.S. employs 30 labor to produce 25 cars
- The world economy can use the same amount of labor (30) and produce 5 more cars


{An example of economies of scale}

- In order for the U.S. to recruit more labor to the car industry, it must decrease or abandon the production of other goods
- These other goods could be produced in Britain, using labor that was formerly employed in the production of cars that now the U.S. makes
- If U.S. and Britain each specialize in producing certain varieties of cars, then cars can be produced at a much larger scale!


{An example of economies of scale}
Why would consumers be willing to import different varieties from other countries?


Because consumers enjoy having options.  Part of our theory will rely on the assumption that consumers have \textbf{love for variety}.  


Mutually beneficial trade occurs because each country can produce goods at a larger scale (lower average cost) and consumers still have access to the full range of goods available through trade.


{Internal versus External Economies of Scale}
Economies of scale can emerge either because of industry-level effects, or firm-level effects:

- \textbf{Internal economies of scale} occur when the cost per unit depends on the size of an individual company, but not necessarily on the size of the industry
- \textbf{External economies of scale} occurs when the cost per unit depends on the size of the industry, but not necessarily the size of the firm


{Internal versus External Economies of Scale}
An industry where economies of scale are purely external will typically consist of many small firms and be perfectly competitive.


An industry where economies of scale are purely internal will typically give rise to large firms and results in imperfectly competitive markets.


Because these different types of economies of scale have different implications for the market structure, it's difficult analyze both in a single model.  Let's investigate \textit{external} economies of scale first.


{Internal versus External Economies of Scale}
Since external economies of scale rely on cost advantages at the industry level, often these types of industries tend to be geographically concentrated:

- Semiconductor industry in California's Silicon Valley
- Investment banking in New York
- Financial analysis in Shenzhen and Shanghai
- Entertainment industry in Hollywood
- Button and zipper manufacturing in Qiaotou
- Cigarette lighters in Zhang Qi
- Information services in India


{External Economies of Scale}
There are three main reasons why a cluster of firms may be more efficient than an individual firm in isolation:

- Specialized suppliers
- Labor market pooling
- Knowledge spillovers



{Specialized Suppliers}
Production of goods, and the development of new products, often requires the use of specialized equipment or services.


Often an individual business does not provide enough demand to support the business of these specialized suppliers, but a group of firms does.



{Specialized Suppliers}
Consider Silicon Valley's semiconductor industry as an example:

- Many engineers left established semiconductor companies to start businesses in Silicon Valley that manufactured specialized capital goods like diffusion ovens, step-and-repeat cameras, photomasks, testing jigs, and specialized chemicals
- This specialized equipment sector promoted the formation of semiconductor companies by freeing individual businesses from the expense of developing capital equipment internally
- A company that tried to enter the semiconductor industry not in Silicon Valley would be at a disadvantage, since it would lack access to Silicon Valley's suppliers.


{Labor Pooling}
A cluster of firms can create a pooled market for workers with highly specialized skills.  Businesses benefit because they can always find qualified employees, and workers benefit because they can always find a job.


Consider the movie industry (Hollywood, Bollywood, Nollywood etc).  Sometimes there are many movie projects that can be made, while other times there are few.  If actors, directors, producers, and staff all work in the same area, they can easily relocate their labor to the movie project that gets funding.


{Knowledge Spillovers}
An important source of knowledge/technological development is done by the information exchange of information and ideas that take place on a personal level.  This kind of knowledge diffusion seems to take place most effectively when an industry is concentrated in a small area.


Consider University Town.  It located three universities in the same geographic area.  Educational institution can observe each other, learn from each other, and professors and students can exchange information and ideas.

{External Economies of Scale and Trade}
If history and luck plays a role in deciding where production occurs, then it is very possible that industries aren't always located in the ``right" place.


Once a country has established an advantage in an industry, it may retain that advantage even if some other country could potentially produce the goods more cheaply.


If this is true, it's possible that trade based on external economies of scale may actually leave a country worse off than it would have been in the absence of trade.

\centering


{Internal Economies of Scale}
Now we will investigate the role of \textbf{internal economies of scale} in determining the pattern of international trade.


Internal economies of scale occur when the average cost of production decreases within a firm.  


Modeling internal economies of scale requires us to leave the world of perfect competition...


{Internal Economies of Scale}
Why do we have to abandon perfect competition?


Perfect competition drives the price of a good down to marginal cost

$$
p=AR=MC
$$

But economies of scale imply that marginal cost is less than average cost

$$
p = AR = MC < AC
$$

but this means that firms operate at a loss!


{Internal Economies of Scale}
If perfect competition drives the price down to marginal cost, firms would suffer losses because the would not be able to recover the higher costs incurred from producing the initial units of output.  As a result, these firms would be forced out of the market, and this process would continue until an equilibrium featuring imperfect competition is attained.


{Internal Economies of Scale: An Example}
Suppose that a firm's cost of producing $Q$ units of output is given by

$$
C(Q) = F+cQ
$$

where would say $F$ is a fixed cost of production (e.g. overhead) and $c$ is a variable cost of production (e.g. supplies, labor, capital).


We define the \textbf{average cost} and \textbf{marginal cost} of production to be

$$
\begin{align*}
AC &= \frac{C(Q)}{Q} = \frac{F}{Q}+c \\
MC &= \frac{\partial C(Q)}{\partial Q}=c
\end{align*}
$$

{Internal Economies of Scale}
We need to build a model of imperfect competition.


This will allow us to introduce two additional characteristics of firms that are prevalent in the real world:

- In most sectors, firms produce goods that are differentiated from one another
- Performance measures, such as size and profits, vary widely across firms.



First let's develop an economy featuring imperfect competition without performance differences, then we can add those in later.


{Krugman 1979}
\textit{``This paper develops a simple, general equilibrium model of noncomparative advantage of trade.   Trade is driven by economies of scale, which are internal to firms.  Because of the scale economies, market are imperfectly competitive.  Nonetheless, one can show that trade, and gains from trade, will occur, even between countries with identical tastes, technology, and factor endowments."}


{Environment: Imperfect Competition}
Imperfect competition describes a market environment with the following features:

- Monopoly power: each firm produces a differentiated \textbf{variety} of good and faces a downward sloping demand curve
- No strategic interaction: there are a large number of firms to each individual firm ignores its impact on the demand of other firms
- Free entry and exit: there are no barriers to entry or exit so economic profits are driven to zero in equilibrium 



{Imperfect Competition}
Monopoly power means that firms have to choose the price they sell their product at.  They are \textit{not price takers}.  


Boeing, the large aircraft manufacturer, competes with European firm Airbus.  Boeing knows that if it wants to sell more airplanes than it currently does, it has to reduce its price.


This is very different from \textit{perfect} competition, where firms have no control over the price because everyone produces exactly the same good, so if a firm charges a higher price that the market it sells nothing; and if it charges a lower price it loses money.


{Environment: Preferences}
There is a fixed number $L$ of consumers who have identical tastes over varieties $i=1,...,N.$

$$
U = \sum_{i=1}^{N} v(c_i), \quad v' > 0 \quad v'' <0
$$

Each consumer has one unit of labor which earns them a wage $w$ so that 

$$
\sum_{i=1}^{N} p_i c_i \leq w
$$



{Environment: Technology}
Labor is the only factor of production whose requirements are given by,

$$
l_i = f + \frac{q_i}{\psi}
$$

where $f$ is the fixed cost of production and $\psi$ is a productivity parameter common to all firms.


{Equilibrium condition 1: Agents' maximize utility}
Agents maximize utility:

$$
\begin{align*}
\max_{c_i} \sum_{i=1}^{N} v(c_i) \quad \text{s.t.} \quad \sum_{i=1}^{N} p_i c_i \leq w
\end{align*} 
$$

$$
\Rightarrow v'(c_i) = \lambda p_i
$$

where $\lambda$ is the Lagrange multiplier associated with the agents budget constraint.


{Equilibrium condition 2: Firms' maximize profits}
Firms maximize profits:

$$
\begin{align*}
\max_{q_i}\hspace{2mm} &p_i q_i - w (f + \frac{q_i}{\psi}) \\
\end{align*}
$$

$$
\Rightarrow p_i = \underbrace{\left( \frac{\eta_i}{\eta
_i -1}\right)}_{\text{markup}}\frac{w}{\psi}
$$

where $\eta_i \equiv -(v'/c_i v'') >0$ is the elasticity of demand for variety $i$.


{Equilibrium conditions 3: Free entry and exit + Market clearing}
Free entry and exit, so firms earn zero profits in the long-run ($p_i = AC_i$):

$$
p_i = \frac{f w}{q_i} + \frac{w}{\psi}
$$

Goods and labor market clear:

$$
\begin{align}
&q_i = L c_i \\
&L = \sum_{i=1}^n l_i = \sum_{i=1}^n  f + \frac{q_i}{\psi}
\end{align}
$$

{Equilibrium}
Symmetry $\Rightarrow p_i = p, q_i = q, c_i =c$ for all $i \in [0,n]$.


$(c,p/w)$ are jointly determined by the following two equations:

$$
\begin{align*}
&(PP): \quad \frac{p}{w} = \left( \frac{\eta_i}{\eta
_i -1}\right)\frac{1}{\psi} \\
&(FE): \quad \frac{p}{w} = \frac{f}{Lc} + \frac{1}{\psi} 
\end{align*}
$$

{Equilibrium}
The number of varieties produced in equilibrium is given by,

$$
N = \frac{1}{f/L + c/\psi}.
$$



{International Trade}
Suppose there exist two economies with identical tastes and technologies that open up to trade.  Symmetry ensures that the wage rates and price of any good will be the the same in both countries.  The effect will be the same as if each country experienced an increase in its labor force.

{International Trade}
The direction of trade is intdeterminate.  All we can say is that each good will be produced in one country.  Since all goods have the same price, expenditures on each country's goods will be proportional to the country's labor force.

$$
M = wL \cdot \frac{L^*}{L+L^*} = M^*
$$

Trade is balanced, and the volume of trade as a fraction of world income is maximized wehn the economies are of equal size.


{Key Point}
Economies of scale can generate gains from trade even when there are no international differences in tastes, technology, of factor endowments.


{Comparative statics}
There exist gains from trade because:

- more product variety
- higher real wages

The number of goods produced in each country is proportional to its labor force,

$$
N = \frac{L}{f+q/\psi}, \qquad N^* = \frac{L^*}{f+q/\psi}
$$

As firms move down their average cost curve, output $q$ increases which necessarily means each country has fewer firms operating after trade. 


{Effects on business from trade}
Krugman model posits two impacts of trade on the productivity of firms:

- Selection effect: some firms are forced to exit after trade
- Scale effect: surviving firms expand their output



{An Empirical Study}
Trefler (2001) studies the selection effect from the FTA.

- Short-run costs included a 15\% decline in employment and a 10\% decline in both output and the number of plants
- Long-run gains included productivity gains of about 17\% (or 1\% per year!).
- About half of the productivity gains come from favorable plant turnover (entry and exit)
Our model assumed symmetry---all firms were the same size and effiency; so entry and exit would not affect producitivity.  If we want to capture a meaningful selection effect, we need to allow for heterogenous firms.  Next time...


{Conclusion}

- We began with the observation that similar countries trade a lot, and often similar goods.
- We need a theory that generates gains from trade even when countries are very similar.
- Introduce economies of scale and (necessarily) monopolistic competition.
- Free trade benefits economies by lowering the average prices of goods (higher real wages) and introducing more product variety
- Next time, we want to all firms to be different because we see large selection effects in international trade where top firms seem to be exporting while low productivity firms do not.
