# Diamond-Dybvig Model of Banking


Why do financial intermediaries exist?  Household can lend to non-financial firms directly in stock or bond markets, yet still most securities are traded indirectly via intermediaries rather than directly. Financial intermediaries (banks) should then beat the market in some way.

Our focus is on banks as liquidity providers: they issue liquid assets (deposits) while financing illiquid ones (loans).  Banks can improve on the competitive market by providing better **risk sharing** among people with different liquidity needs. Banks hold a substantial amount of **liquidity risk** due to **maturity mismatch** between the liability and asset sides of their balance sheet.

<a href="https://www.bu.edu/econ/files/2012/01/DD83jpe.pdf" target="_blank">Diamond and Dybvig developed a model</a> which shows that banks can create deposit contracts that do better than competitive exchange markets, however are also subject to self-fulfilling panics resulting in bank runs.  Diamond and Dybvig, along with former Fed chair Ben Bernanke, were award the 2022 Nobel Prize in Economics for their work on this model.

## Portfolio Choice under Uncertainty

Let's describe the environment of the model and discuss how agents make decisions about a portfolio of assets when there is uncertainty over future liquidity needs.  

### Timing and Types of Consumers

We consider a static three period model.  At time $t=0$, agents are born and invest their wealth in a portfolio of assets.  At time $t=1$ some agents are **impatient** and desire to use their assets to purchase consumption goods.  Other agents are **patient** and wait until time $t=2$ to liquidate their portfolio for consumption.  Agents do not know their type when born, but are impatient with probability $\theta$ and patient with probability $1-\theta$.  They learn which type they are at time $t=1$. This description is abstract enough to imagine that agents do not necessarily have different preferences for early or late consumption, but rather are subject to probabilistic liquidity shocks.  Some agents are hit by a liquidity shock (e.g. my car broke down and I need money to fix it), which requires them to liquidate their portfolio at time $t=1$, while other are not hit by the shock (e.g. my car is running just fine) and can wait to liquidate their portfolio until $t=2$.  


### Assets

All agents are endowed with 1 unit of wealth and make an investment choice at time $t=0$ over two assets:
- **Risky Asset:** yields $r_1$ in period $t=1$ or $r_2$ in period $t=2$ where $r_1<1<r_2$.
- **Safe Asset** yields $1$ in both periods $t=1,2$

Let $x$ denote the fraction of wealth invested in the risky asset, so that $(1-x)$ is invested in the safe asset.  The risky asset yields a higher payoff than the safe asset only if it is held until maturity at time $t=2$.  Otherwise, liquidating it early suffers a penalty resulting a rate of return lower than the safe asset.  In fact, $r_1/r_2$ may be interpreted as a measure of liquidity where lower $r_1/r_2$ implies less liquid asset.


### The Consumer's problem

Consider a consumer who has invested $x \in [0,1]$ in the risky asset.  With probability $\theta$ the consumer is impatient and wants to consume today

$$
c_1 = (1-x) + xr_1 = 1-x(1-r_1), \quad x
$$

while with probability $1-\theta$ the consumer is patient and wants to consume tomorrow

$$
c_2 = (1-x) + xr_2 = 1+x(r_2-1)
$$

The patient agent can realize the high rate of return $r_2$ on their risky asset, while the impatient agents suffers the low rate of return $r_1$.  The safe asset always gives a rate of return of 1.

The consumer decides how to allocate her portfolio between the two assets by maximizing expected utility subject to period budget constraints,

$$
\begin{align*}
\max_{c_1,c_2} \hspace{2mm} &\theta u(c_1) + (1-\theta) u(c_2) \\
\text{s.t.} \quad &c_1 = 1-x(1-r_1) \\
&c_2 = 1+x(r_2-1)
\end{align*}
$$

First order condition equates the marginal rate of substitution to the relative price of early-to-late consumption and gives us the Euler equation,

$$
\frac{u'(c_1)}{u'(c_2)} = \frac{1-\theta}{\theta}\frac{r_2-1}{1-r_1}
$$

Graphically, optimal consumption $(c_1^*,c_2^*)$ is located at the intersection of the above f.o.c. with the feasible frontier given by agents' lifetime budget constraint.  When plotting the lifetime budget constraint, however, we must be careful.  The lifetime budget line $c_2=1+\frac{r_2-1}{r_1-1}(c_1-1)$ defines an infinite line in the $(c_1,c_2)$ plane, but only a portion of that line is economically meaningful since $x \in [0,1]$.  Consider the limiting cases of investing in only one asset and the resulting consumption possibilities:

| Portfolio Allocation  | Investment Strategy | Resulting consumption pair | Interpretation               |
| ------------- | ----------------- | -------------------------- | -------------------------------------- |
| $x = 0$       | All safe asset    | $(c_1,c_2)=(1,1)$          | Fully liquid, low return               |
| $x = 1$       | All risky asset   | $(c_1,c_2)=(r_1,r_2)$      | Illiquid, high return                  |
| $x \in (0,1)$ | Mixed portfolio   | Between the endpoints      | Trade-off between liquidity and return |

Therefore, the appropriate lifetime budget constraint is the line segment connecting the points $(c_1,c_2)=(1,1)$ and $(c_1,c_2)=(r_1,r_2)$.  

Notice that for both assets to be held in the portfolio, the consumer's problem must have an interior solution.  That is, we must have $c_2>c_1$.  According to the Euler equation this implies that

$$
\frac{u'(c_1)}{u'(c_2)} = \frac{1-\theta}{\theta}\frac{r_2-1}{1-r_1} > 1
$$

which is re-arranged to say

$$
\theta r_1+(1-\theta)r_2 >1
$$

Said simply, the expected rate of return on the risky asset must be higher than the safe asset for both assets to be held in equilibrium.  This should not come as a surprise.  If the safe asset paid higher returns compared to the risky asset, then agents would only hold the safe asset as it gives certain payoffs that dominate the risky asset.  Therefore, the optimal portfolio lies somewhere on the budget line:

### Autarkic Equilibrium

Under autarky--a state of the world where agents are unable to trade assets among themselves---a risk-averse consumer either partially hedges herself against liquidity risk by holding the safe asset $(x<1)$ or does not invest in the risky asset at all $(x=0)$.  The consumer's portfolio choice is given by their Euler equation subject to their budget constraints, which is graphically the tangency of a consumer's indifference curve with the lifetime budget constraint:

Although not explicitly included in our environment yet, it is worth noting that if investments in physical capital were illiquid, then the economy operates along a steady state that is not Pareto optimal: several projects with positive NPV are discarded.  Notice that our setup resembles the three-period OLG model developed earlier.  In that environment, capital was similar to the risky asset in that it made a rate of return $r_2=f'(k)$ if held to maturity or $r_1=0$ if liquidated earlier.  Fiat money, on the other hand, was similar to the safe asset since it paid a constant rate of return $\nu_{t+1}/\nu_t=n/z$ in every period.  So long as $f'(k)>n/z$ there existed an equilibrium where both assets were held, similar to the condition that $\theta r_1+(1-\theta)r_2 >1$ in our Dimaond-Dybvig model. The extra element is uncertainty over whether or not consumption happens when middle aged ($t=1$) or old ($t=2$).  

### Planner's Problem

We now derive the optimal allocation by considering a central planner who can choose the mix of risky and safe assets to maximize expected utility of representative consumer.  There is a crucial difference between the planner and an individual consumer: **the planner does not face uncertainty over the aggregate liquidity needs**.  We assume that the liquidity shock (or type of consumer) is independent and identically distributed across agents.  Therefore, we can appeal to the law of large numbers to say that there must exist a fraction $\theta$ of the population which will desire to consumer early and the remaining fraction $1-\theta$ will want to consumer late.  So from the planner's perspective, it *knows for certain* how many people will desire to consume early versus late.  This idea is reflected in the **feasibility constraints** of the planner:

$$
\begin{align*}
\theta c_1 &\leq (1-x) \\
(1-\theta)c_2 &\leq (1-x)-\theta c_1+x r_2
\end{align*}
$$

The first constraint shows that the planner must invest enough in the safe asset to satisfy the liquidity needs of a fraction $\theta$ of the population.  Every early type, of which there are $\theta$ since we assume the population is of size one, receives consumption $c_1$.  And this consumption is financed from the returns of the safe asset which are $(1-x)$.  The second constraint shows that the total returns from the portfolio less what is paid out to early types can be used to finance late consumption.  In principle, the planner could liquidate some of the risky asset at time $t=1$ to finance a portion of early consumption, but this is not optimal since $r_1<1$.  The planner optimally chooses to only use the safe asset to finance early consumption.  This optimality condition is implicit in the feasibility constraints above since we do not include early liquidation of the safe asset.

Rearranging the second feasibility constraint gives $(1-\theta)c_2 \leq 1-x\theta c_1++x(r_2-1)$.  Since the risky asset gives a higher return than the safe asset, $r_2>1$, the planner will choose $x$ ar large as feasible--up to the bound $x \leq 1-\theta c_1$ to maximize the right hand side.  Plugging in $x=1-\theta c_1$ gives the lifetime feasibility constraint:

$$
\theta r_2 c_1 + (1-\theta) c_2 \leq r_2
$$

The planner's problem is thus given by

$$
\begin{align*}
\max_{c_1,c_2} \hspace{2mm} &\theta u(c_1)+(1-\theta)u(c_2) \\
\text{s.t.} \quad &\theta r_2 c_1 + (1-\theta) c_2 \leq r_2
\end{align*}
$$

and the optimal consumption plan is

$$
\frac{u'(c_1^{pp})}{u'(c_2^{pp})}=r_2
$$

which implies $c_1^{pp} < c_2^{pp}$.  Recall that the planner knows $\theta$ and can choose an aggregate $x$ which optimally trades off liquidity and return without premature liquidation.  Therefore the marginal trade-off between current and future consumption is the risky asset’s full maturity return r_2.  In contrast, the decentralized economy is subject to uncertainty from liquidity shocks so that the marginal trade-off between current and future consumption is $(r_2-1)/(r_1-1)$; where $r_1<1$ makes liquidity provision costly.

### Competitive Asset Market

We now allow agents to trade their assets, after learning their type, in a competitive spot market.  If a competitive asset market can achieve the optimal allocation $(c_1^{pp},c_2^{pp})$, then there is no need for any other institution or policy in this economy.  Simply let agents trade their assets and everything works out.  If, however, the competitive allocation falls short from the optimal one then there may be room for banks to improve things.

Formally, we introduce a competitive asset market that opens at time $t=1$ where agents buy and sell the risky asset in exchange for the safe asset.  An impatient consumer wants to the sell the risky asset, while a patient consumer want to the buy the risky asset.  We denote by $p$ the price of the risky asset which is determined by competitive market forces.  

Impatient consumers realize a liquidity shock and therefore want to liquidate their risky asset by selling it in the asset market.  Given the market price $p$ of the risky asset, the amount of consumption they can enjoy is

$$
c_1=(1-x) + px
$$

An impatient consumer could instead directly liquidity the risky asset for return $r_1$, and so would never accept a trade where $p<r_1$.  Therefore, we make the conjecture that the equilibrium will have $p \geq r_1$ and check whether or not this is true later.

Patient consumers want to purchase the risky asset in order to realize its high return.  Each unit of the safe asset they sell buys them $1/p$ units of the risky asset, so that consumption is

$$
c_2 = (\frac{(1-x)}{p}+x)r_2
$$

Patient consumers will only pay up to $p \leq r_2$ for the risky asset.  So participation constraints from the patient and impatient types suggest that $r_1 \leq p \leq r_2$.  We check if our equilibrium $p$ indeed satisfies these participation constraints.

The consumers's problem at time $t=0$ is

$$
\begin{align*}
\max_{c_1,c_2} \hspace{2mm} &\theta u(c_1) + (1-\theta)u(c_2) \\
\text{s.t.} \quad &c_2=c_1 \frac{r_2}{p}
\end{align*}
$$

What price $p$ is compatible with equilibrium in the asset market?  A competitive price $p$ must equalize the supply and demand for the risky asset.  The supply of the risky asset is the number of patient consumers times their holdings of the risky asset $\theta x$.  The demand for the risky asset is the number of impatient consumers times the value of their safe asset $(1-\theta)\frac{1-x}{p}$

Suppose that $p > 1$.  In this case, the return on the risky asset strictly dominates the safe asset and all consumers choose $x=1$ implying zero demand.  At $t=1$ impatient consumers want to sell the risky asset, but nobody holds the safe asset to exchange with.  With zero demand we must have $p=0$ which contradicts our assumption that $p > 1$.

Suppose instead that $p<1$. In this case, the return of the safe strictly dominates the risky and all consumers choose $x=0$ implying zero supply. At $t=1$ patient consumers want to buy risky asset, but nobody has any to sell so that $p=r_2$ which is also a contradiction.

For the asset market to be in equilibrium, it must be that $p=1$. Equilibrium consumption is therefore,

$$
\begin{align*}
&c_1=x+(1-x)=1 \\
&c_2 = (x+(1-x))r_2 = r_2
\end{align*}
$$

Notice that with a competitive asset market the portfolio choice is immaterial.  

How does this allocation compare to autarky?  The competitive asset market allows agents to trade after uncertainty about liquidity shocks has been resolved, and therefore gives consumption which dominate any allocation under autarky:

Does the asset market yield an efficient outcome? Recall the Pareto optimal allocation satisfies

$$
u'(c_1)=R u'(c_2)
$$

while they competitive asset market equilibrium has

$$
(c_1,c_2)=(1,R)
$$

Given agents are risk averse, we have that

$$
u'(1) > Ru'(R)
$$

So the competitive equilibrium is not efficient!


Proof:

$$
\begin{align*}
Ru'(R) &= u'(1) + \int_{1}^{R}\frac{\partial c u'(c)}{\partial c}dc \\
&=u'(1)+\int_{1}^{R}[cu''(c)+u'(c)]dc \\
&=u'(1)+\int_{1}^{R} u'(c) \left[\frac{cu''(c)}{u'(c)}+1 \right] \\
&< u'(1)
\end{align*}
$$

which uses the fact that risk averse agents must have $-\frac{cu''(c)}{u'(c)} >1$

### Banks

Surprisingly, the competitive asset market did not give the optimal allocation.  This opens up the possibility that an institution or policy could intervene and improve things.  Here we introduce a competitive banking sector offering the following deposit contract:
- Receive a 1 unit deposit at time $t=0$ and promise to pay $c_1^B$ if a consumer withdraws at time $t=1$ or pay $c_2^B$ if a consumer withdraws at time $t=2$.
- The bank collects the initial endowments and invests the resources in the economy in safe asset and risky asset appropriately
- Consumer then choose at time $t=1$ whether to "withdraw" or ``wait"

In a competitive banking equilibrium, banks compete for deposits and earn zero economic profits so the problem facing a bank is

$$
\begin{align*}
\max_{x,y} \hspace{2mm} &\theta u(c_1) + (1-\theta)u(c_2) \\
\text{s.t.} \quad &x+y=1 \\
&\theta c_1=y \\
&(1-\theta)c_2=xR
\end{align*}
$$

Which gives the optimal allocation,

$$
u'(c_1^B)=Ru'(c_2^B)
$$

We do not need either observability of types or a central planner to achieve the efficient allocation.  Just open banks. Banks provide optimal insurance against liquidity shocks. Banks are involved in an asset-liability management problem. The bank knows a fraction $\theta$ of consumers must be paid $c_1^B$ which it can honor if it ivests in the liquid asset a fraction $y=\theta c_1^B$. The remaining $(1-\theta c_1^B)$ can be invested in the long asset to return $R(1-\theta c_1^B)$ that can be distributed amongst a fraction $1-\theta$ depositors.

Moreover, we have that $c_1^B < c_2^B$ which implies that the allocation is incentive compatible: each depositor has the right incentive to report faithfully its type to the bank.  The deposit scheme and asset allocation are a Nash equilibrium.  But there is another bad Nash equilibrium...

### Bank Runs

If depositors believe that only impatient consumer withdraw at time $t=1$, the bank can satisfy all depositors. If depositors hold different beliefs, however, then all consumers may want to withdraw at time $t=1$ and the total amount the bank can pay back is

$$
r_1 x + y < x+y <c_1^B
$$

The bank becomes insolvent given that it can pay back just a fraction of what it promised to depositors. Since all assets must be liquidated in $t=1$, nothing remains at $t=2$ for an inattentive latecomer depositor.

A bank run is a coordination failure among patient types:

Bank runs are an equilibrium phenomena. The good equilibrium is fragile: the system can jump to a bad one due to an exogenous adverse random signal not necessarily linked to the fundamental of the bank. Trust in the bank is essential to preserve the good equilibrium. Even if banks cannot determine a person's true liquidity need, there are still several ways that can be protected from runs
- Interbank lending: if a bank can borrow enough to meet all withdrawals, it can avoid those losses that would result from the sale of its capital
- Suspension of withdrawals: a bank can temporarily close its doors when its reserves of the short-term asset have been used up
- Government deposit insurance: the government can establish regulations that guarantee patient-types will be paid even if the bank becomes insolvent

For example, China established its Deposit Insurance Regulation in May 2015:

**Article 2**: Commercial banks, rural cooperative banks, rural credit cooperatives, and other deposit-taking banking financial institutions formed within the territory of the People's Republic of China (hereinafter collectively referred to as “institutional policyholders”) shall buy deposit insurance under this Regulation.

**Article 3**: For the purpose of this Regulation, the term “deposit insurance” means the system under which institutional policyholders pay insurance premiums to the deposit insurance fund management institution, forming a deposit insurance fund, and the deposit insurance fund management institution pays the insured deposits to depositors and take necessary measures to maintain the safety of deposits and the deposit insurance fund.

If the government insures depositors against all losses, depositors will no longer care about their bank's exposure to risk.  Banks seeking depositors must therefore offer the highest rates of return they can find, which incentivizes them to hold a riskier portfolio of assets.

Deposit insurance will induce banks to take greater risks than they would if all the risks were borne by the shareholders and depositors of the bank---this is the **moral hazard problem** of insurance.

When banks attract investors as shareholders, depositors are more protected than losses even when bank assets are risky.]

If a bank suffers a sudden loss, it will be subtracted from the bank's net worth and not from deposits.  Only after net worth falls to zero will depositors lose anything.

We already saw that ROE is positively related to leverage

$$
\uparrow \frac{\text{assets}}{\text{bank capital}} \rightarrow \uparrow \text{ROE}
$$

so there is an incentive to boost leverage which puts more risk on depositors.


Bank regulators impose **capital requirements** to force banks to maintain a net worth no less than some fraction of their risk-weighted assets.  Basel III is the latest international standard for bank regulations that stipulate how banks should measure their risk, and what capital is required given those risks.


{Conclusion}

- The Diamond-Dybvig model shows that banks can accomplish perfect risk-sharing by transforming illiquid assets into liquid deposits
- This maturity mismatch on the bank balance sheet, however, can generate bank runs driven by self-fulfilling beliefs
- Bank regulations are largely motivated by preventing such runs and engendering trust in the financial system
