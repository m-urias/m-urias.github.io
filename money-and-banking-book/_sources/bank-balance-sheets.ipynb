# The Bare Bones of Bank Balance Sheets

To understand how banks operate--and how monetary policy works--we use T-accounts to show how common banking transactions are reflected on the balance sheet.  We build a hypothetical bank from scratch and observe how basic banking transactions affect its financial position.  We then aggregate across banks to show the balance sheet for the entire banking sector.  Finally, we discuss the balance sheet of the central bank and how it relates to private banks and creation of liquidity for the entire economy.

## Assets, Liabilities, and Bank Capital

Recall the fundamental accounting identity which underpins every company's balance sheet:

$$
\text{Assets} = \text{Liabilities} + \text{Owner's Equity}
$$

Assets are what the bank owns and represent future economic benefits. For a bank, the primary assets are:
- Reserves: Cash in the vault or deposits held at the Central Bank (e.g., the Federal Reserve in the U.S.). This is the most liquid asset.
- Loans: Funds extended to households and businesses. These are the bank's most profitable assets, but also carry the highest risk.
- Securities: Interest-bearing assets like government bonds or Treasury bills. These are held for liquidity and income.

Liabilities are what the bank owes to others. They represent the bank's main sources of funds:
- Deposits: Funds provided by customers (e.g., checking, savings, time deposits). These are a liability because the bank owes this money back to the depositor.
- Borrowing: Funds borrowed from other banks (the interbank market) or from the Central Bank (the discount window).

Owner's equity is the bank's net worth (often called bank capita)-â€”the funds contributed by its owners/investors. It acts as a buffer against unexpected losses.

## Common Bank Transactions

Recall that modern accounting systems obey the principle of double-entry bookkeeping: every financial transaction is recorded in at least two different accounts with equal and debit and credit entries.  This system is based on the fundamental accounting identity which demand that every transaction has must maintain the identity; hence there must be at least two accounts equally affected to maintain the balance between the left hand side and right hand side of the T-account.

### Starting a Bank and Raising Deposits

Let's build a hypothetical bank from scratch. Suppose that a group of investors pool their wealth to start First National Bank (FNB) and put up $\$1,000,000$ in capital. The bank holds this cash in its vault as reserves.  Recall that the real-world definition of reserves is vault cash plus deposits held with the central bank.  The balance sheet of the newly created FNB is

| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves      $ \$1,000,000$    | Bank Capital $ \$1,000,000$ |

FNB starts operating and successfully attracts $\$9,000,000$ in new customer deposits. This injection of $\$9,000,000$ increases reserves, but deposits also represent a liability to the bank because if depositors request a withdrawal then the bank needs to pay what is demanded from its reserves.  The bank therefore simultaneously debits reserves for $\$9,000,000$ and credit deposits for $\$9,000,000$.

| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$10,000,000$   | Bank Capital $ \$1,000,000$ |
|                             | Demand Deposits $ \$9,000,000$ |

The customer's cash deposit leads to an **increase in the bank's reserves equal to the increase in demand deposits**.

### Creating Loans

In order to make profits, FNB decides to extend a loan of $\$5,000,000$ to a local business.  The bank does not physically hand over cash or reserves when it gives the business loan. Instead, it creates the loan (an asset) and simultaneously credits the borrower's checking account (a demand deposit liability).  The resulting balance sheet is

| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$10,000,000$   | Bank Capital $ \$1,000,000$ |
| Loans $ \$5,000,000$          | Demand Deposits $ \$14,000,000$ |

*The act of lending creates new money (deposits) in the banking system.*  

The local business needs to use their newly created deposits to finance business operations (e.g. paying suppliers, workers, rent).  Let's suppose that the local business uses the full $\$5,000,000$ of their FNB deposits to pay a supplier, who happens to bank at Second National Bank (SNB). FNB must transfer the reserves to SNB to settle the transaction.  FNB credits its reserves $\$5,000,000$ and debits deposits $\$5,000,000$.  Similarly, SNB debits reserves $\$5,000,000$ and credits deposits $\$5,000,000$. 

First National Bank:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$5,000,000$   | Bank Capital $ \$1,000,000$ |
| Loans $ \$5,000,000$          | Demand Deposits $ \$9,000,000$ |


Second National Bank:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$5,000,000$   |  Demand Deposits $ \$5,000,000$ |


*When a bank receives additional deposits, it gains an equal amount of reserves.  When it loses deposits, it loses an equal amount of reserves.*  For the banking system as a whole, total loans increased by $\$5M$ and total deposits increased by $\$5M$. The system's total reserves are unchanged, they have only been redistributed from FNB to SNB.

### Investing in Securities

To generate additional interest income and hold a liquid, low-risk asset, FNB uses another $\$2,000,000$ of its remaining reserves to purchase government bonds (securities).  FNB credits reserves for $\$2,000,000$ and debits securities for $\$2,000,000$

First National Bank:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$3,000,000$   | Bank Capital $ \$1,000,000$ |
| Loans $ \$5,000,000$       | Demand Deposits $ \$9,000,000$ |
| Securities $ \$2,000,000$    |                                |

### Liquidity Management and Money Markets

Bank's make profits by transforming short-term liabilities (deposits) into longer-term assets (loans and securities).  Sometimes it is said that banks are in the business of "borrowing short and lending long" which means that there is always a maturity mismatch between assets and liabilities on its balance sheet.  As such, banks must hold enough reserves for two reasons: (1) there may be an unexpected deposit outflow requiring reserves or (2) there exist reserve requirements imposed by the central bank.  Reserve requirements stipulate the fraction of deposits that must be held as reserves.  A reserve ratio of 100\%, for example, would preclude the bank from issuing loans in excess of bank capital (owner's equity).  Conversely, a reserve ratio of 0\% allows the bank to have zero reserves on hand.  In practice, a bank would never do this because they still optimally hold reserves as insurance against deposit outflow risk.  A bank that is unable to pay out its depositors would surely fail.  Any reserve ratio $rr < 1$ allows the bank to create inside money when they issue new loans.  Such a system is called **fractional reserve banking**.  Reserve requirements are often used as a regulatory measure to guarantee that the banking sector has enough liquid assets to satisfy its liabilities, instill trust in the banking sector, and insure against liquidity shocks.  The quantity of reserves required to be held by law is called **required reserves** and additional reserves the bank voluntarily holds are called **excess reserves**.

Suppose that FNB is subject to a 10\% required reserve ratio and its balance sheet is

| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$3,000,000$   | Bank Capital $ \$1,000,000$ |
| Loans $ \$5,000,000$         | Demand Deposits $ \$9,000,000$ |
| Securities $ \$2,000,000$    |                                |

The bank is required to hold $ \$900,000$ in reserves, so it has $ \$2,100,000$ in excess reserves.  If a deposit outflow of $ \$2,500,000$ occurs, then the bank's balance sheet becomes

| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$500,000$   | Bank Capital $ \$1,000,000$ |
| Loans $ \$5,000,000$         | Demand Deposits $ \$6,500,000$ |
| Securities $ \$2,000,000$    |                                |

Now the bank is required to hold $ \$600,000$ in reserves, but only has $ \$500,000$!  This illustrates the importance of holding reserves as insurance against deposit outflow risk: if bank has ample excess reserves, a deposit outflow does not necessitate changes in other parts of its balance sheet.  The department within a bank which is responsible for managing an appropriate level of reserves is called **liquidity management**.

When a bank cannot meet deposit outflows from its reserves, it has four options:
- Borrow reserves from other banks
- Sell securities
- Borrow reserves from the central bank
- Reduce loans

Selling securities before their maturity date (or until a sufficient rate of return has been realized) is very costly.  Similarly, selling off loans at a fraction of their value or calling in loans early is bad for business.  Borrowing from central bank lending facilities is a viable option, but often their is a stigma associated with doing so--it signals the market that the balance sheet of the bank is in bad shape.  The lowest cost source of reserves is the **interbank market**.  The interbank market is a decentralized exchange where financial institutions, primarily commercial banks, borrow and lend unsecured reserves to one another.  These interbank loans are often very short term (overnight), occur at the **interbank rate**, and facilitate liquidity transfers between banks.  In the US the interbank market is called the **Fed Funds market** and the interbank interest rate is called the **Fed Funds rate**

Suppose FNB's reserves fall too low, and it needs $ \$100,000$ immediately. FNB goes to the interbank market and finds that SNB has excess reserves it is willing to lend at a particular interest rate.  When FNB borrows these reserves from SNB is debits reserves for $ \$100,000$ and credits interbank borrowing for $\$100,000$

First National Bank:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$600,000$      | Bank Capital $ \$1,000,000$ |
| Loans $\$5,000,000$          | Demand Deposits $ \$6,500,000$ |
| Securities $\$2,000,000$     | Interbank borrowing $\$100,000$ |

Similarly, SNB debits interbank loans for $ \$100,000$ and credit reserves for $ \$100,000$

Second National Bank:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$4,900,000$   |  Demand Deposits $ \$5,000,000$ |
| Interbank loans  $ \$100,000$   |                  |

Again, notice that the total amount of reserves in the banking system is unchanged.  Interbank activity has simply redistributed reserve holdings across banks.  These reserves are not free for FNB as they have to pay interest to SNB for the overnight loan.  This underscores the importance of holding excess reserves as insurance against the costs associated with unanticipated outflows.  *The interbank interest reflects is the price of acquiring liquidity in the banking system.* 

### Capital Adequacy Management

In addition to liability risk from unanticipated deposit outflows, banks assets are also subject to risk as loans may default or the market value of securities may fall.  Bank capital acts has a financial cushion to absorb such unexpected losses.  Bank capital also affects the rate of return equity holders realize.  It also provides protection to depositors in the case of bank failure and is therefore an important source of public confidence in the banking system.  Maintaining adequate capital levels can help prevent a bank from taking on too much risk by limiting excessive asset growth.  International standards, such as the Basel II framework, provide regulators guidelines for bank capital requirements. 

Let's consider two banks with different bank capital and see how they respond to negative news about their loan portfolio.  Suppose that High Capital Bank and Low Capital Bank have the following balance sheets:

High Capital Bank:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$1,000,000$   |  Bank Capital $ \$1,000,000$ |
| Loans  $ \$9,000,000$   |  Demand Deposits $ \$9,000,000$ |

Low Capital Bank:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$1,000,000$   |  Bank Capital $ \$400,000$ |
| Loans  $ \$9,000,000$   |  Demand Deposits $ \$9,600,000$ |

Suppose that both banks get caught up in the euphoria of the housing market, only to find later than $ \$500,000$ of their housing loans have become worthless.  The bank must write-off the non-performing loans by crediting loans for \$500,000$ and debiting bank capital for \$500,000$.

High Capital Bank:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$1,000,000$   |  Bank Capital $ \$500,000$ |
| Loans  $ \$8,500,000$   |  Demand Deposits $ \$9,000,000$ |

Low Capital Bank:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$1,000,000$   |  Bank Capital $ -\$100,000$ |
| Loans  $ \$8,500,000$   |  Demand Deposits $ \$9,600,000$ |


Low Capital Bank is in trouble.  The value of its assets have fallen below that of its liabilities, and it's net worth is now negative!  It is **insolvent**: it does not have enough assets to pay off all holders of it's liabilities.  It may the appear that having ample bank capital is good for the bank.  While this indeed provides more of a financial cushion to prevent default, it is actually bad for the return to equity holders in the bank.

The return on equity is defined by

$$
ROE = ROA \times EM
$$

where 

$$
\begin{align*}
\text{Return on Equity:}  ROE &= \frac{\text{net profit after taxes}}{\text{equity capital}} \\
\text{Return on Assets:}  ROA &= \frac{\text{net profit after taxes}}{\text{assets}} \\
\text{Equity Multiplier:}  EM &= \frac{\text{assets}}{\text{equity capital}}
\end{align*}
$$

*For a given rate of return on assets, the lower the bank capital, the higher the return for the owners of the bank.*

The temptation to reduce bank capital is one reason why regulators require capital adequacy standards. Capital requirement for banks take two forms:
- Leverage ratio: the amount of bank capital divided by bank's total assets
- Risk-based capital requirement: different types of assets are subject to different leverage ratios

According the Basel Accord, banks must hold at least 8 percent of their risk-weighted assets
- Reserves and government securities (OECD countries) carry zero weight
- Claims on banks in OECD countries carry 20 percent weight
- Municipal bonds and residential mortgages carry 50 percent weight
- Loans to consumers an corporations carry 100 percent weight
- Off-balance sheet activities---trading financial instruments and generating income from fees---are assigned a credit-equivalent percentage that converts them to on-balance-sheet items

### Central Bank Balance Sheet

The central bank is the bank for commercial banks.  Its balance sheet looks like

Central Bank:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Securities                     |  Currency in circulation  |
| Loans to financial institutions  |  Reserves              |

- Currency in circulation: currency notes are IOUs from the central bank to the bearer, but unlike most liabilities, they promise to pay back the bearer solely with central bank notes (they pay off IOUs with other IOUs).
- Reserves: most banks have an account with the central bank in which they hold deposits.  Reserves consist of those deposits and vault cash.
- Securities: mostly government securities, but also others.  Buying and selling securities is the main way the central bank influences the supply of reserves.
- Loans to financial institutions: central bank can make loans to commercial banks or other institutions.

The **monetary base**, or **high-powered money**, is defined as the sum of currency in circulation and reserves

$$
MB = C + R
$$

The central bank controls the monetary base through open market operations---the purchase and sale of securities through primary dealers.

Suppose that the balance sheet of the entire banking system and the central bank are given by:

Banking System:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$1,000,000$      | Bank Capital $ \$1,000,000$ |
| Loans $\$5,000,000$          | Demand Deposits $ \$10,000,000$ |
| Securities $\$5,000,000$     |                                 |

Central Bank:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Securities    $\$5,000,000$   |  Currency in circulation $\$2,000,000$   |
| Loans to financial institutions $\$5,000,000$  |  Reserves    $\$8,000,000$           |

Suppose that the central bank conducts an open market purchases of $\$1,000,000$ bonds from a primary dealer.  The primary deal sells the bonds to the central bank in exchange for reserves. The central bank debits securities for $\$1,000,000$ and credit reserves $\$1,000,000$ to reflect the acquisition of bonds from primary dealers and the commiserate injection of new reserves into the banking sector.

Banking System:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$2,000,000$      | Bank Capital $ \$1,000,000$ |
| Loans $\$5,000,000$          | Demand Deposits $ \$10,000,000$ |
| Securities $\$4,000,000$     |                                 |

Central Bank:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Securities    $\$6,000,000$   |  Currency in circulation $\$2,000,000$   |
| Loans to financial institutions $\$5,000,000$  |  Reserves    $\$9,000,000$           |

*The central bank's open market purchase expands reserves in the banking system by an equal amount.*  By definition, the monetary base also expands by an equal amount.


If instead the central bank conducts an open market sale of $\$1,000,000$ bonds, then we would have

Banking System:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$1,000,000$      | Bank Capital $ \$1,000,000$ |
| Loans $\$5,000,000$          | Demand Deposits $ \$10,000,000$ |
| Securities $\$5,000,000$     |                                 |

Central Bank:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Securities    $\$5,000,000$   |  Currency in circulation $\$2,000,000$   |
| Loans to financial institutions $\$5,000,000$  |  Reserves    $\$8,000,000$           |

*The central bank's open market sale reduces reserves in the banking system by an equal amount, and also the monetary base.*

The central bank may also affect the monetary base through its lending facilities.  Suppose that the central bank loans $\$1,000,000$ to FNB.  We then have

Banking System:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$2,000,000$      | Bank Capital $ \$1,000,000$ |
| Loans $\$5,000,000$          | Demand Deposits $ \$10,000,000$ |
| Securities $\$5,000,000$     | Borrowings from central bank   $\$1,000,000$   |

Central Bank:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Securities    $\$5,000,000$   |  Currency in circulation $\$2,000,000$   |
| Loans to financial institutions $\$6,000,000$  |  Reserves    $\$9,000,000$           |

The central banks lending facility increased the monetary base by $\$1,000,000$.  Of course, when the bank pays off the loan the transaction will be reversed and the monetary base contracts by an equal (up to the interest payment) amount.

Depositors may also affect the aggregate amount of reserves if they convert deposits into cash.  Suppose that during the shopping season the public wants to withdraw $\$500,000$ cash from the banking system.  The non-bank public realizes an increase in their cash holdings and an equal decrease in their deposits. The banking system and central bank balance sheets now appear as

Banking System:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$500,000$      | Bank Capital $ \$1,000,000$ |
| Loans $\$5,000,000$          | Demand Deposits $ \$9,500,000$ |
| Securities $\$5,000,000$     |                                 |

Central Bank:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Securities    $\$5,000,000$   |  Currency in circulation $\$2,500,000$   |
| Loans to financial institutions $\$5,000,000$  |  Reserves    $\$7,500,000$           |

Therefore, even if the central bank does not conduct open market operations, shifts from deposits into currency still affect reserves.  Although reserves are subject to random fluctuations, note that the net effect on the monetary base is zero.  Hence, the monetary base is under more control of the central bank.

## Deposit Creation in a Fractional Reserve Banking System

Although the central bank has control over open market operations, it does not control desired borrowing by banks.  We can classify the monetary base into three components: part of which the central bank has control over and the other it doesn't.

$$
MB = BR + NBR + C
$$

The monetary base consists of borrowed reserves (BR), non-borrowed reserves (NBR) and currency (C).

Consider the balance sheet of FNB:

First National Bank:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$1,000,000$   | Bank Capital $ \$1,000,000$ |
| Loans $ \$8,000,000$       | Demand Deposits $ \$10,000,000$ |
| Securities $ \$2,000,000$    |                                |

and suppose that the central bank conducts an open market purchase for $\$1,000,000$ with FNB:

First National Bank:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$2,000,000$   | Bank Capital $ \$1,000,000$ |
| Loans $ \$8,000,000$       | Demand Deposits $ \$10,000,000$ |
| Securities $ \$1,000,000$    |                                |

After the open market operation the bank finds it has excess reserves of $\$1,000,000$.  The bank decides to make a loan equal to excess reserves, thereby creating deposits and increasing M1.  Recall that the business who borrows from the bank will not leave the $\$1,000,000$ sitting idle in its account, but will purchase goods and services to operate its business.  When the borrower makes these purchases, the $\$1,000,000$ will leave FNB.  Therefore, *a bank cannot safely make a loan for an amount greater than the excess reserves that it has before it makes the loan.*


We now track the initial newly created $\$1,000,000$ deposits by FNB through a competitive banking system.  Let's assume that Bank A receives the $\$1,000,000$ deposits (the business who took out the loan paid a supplier who banks with Bank A).  Bank A realizes an increase in its reserves adn deposits:

Bank A:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$1,000,000$   | Demand Deposits $ \$1,000,000$ |

Assuming a 10 percent required reserve ratio, Bank A finds itself with $ \$900,000$ excess reserves it can use to create new loans. 

Bank A:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$1,000,000$   | Demand Deposits $ \$1,900,000$ |
| Loans  $ \$900,000$        |                                 |

Again, these newly created deposits do not remain in Bank A, but are instead end up in Bank B:

Bank B:
| **Assets**                  | **Liabilities**      |
|-----------------------------|----------------------|
| Reserves  $ \$900,000$   | Demand Deposits $ \$900,000$ |

Bank B finds itself with $ \$810,000$ in excess reserves, which it uses to create loans.  These newly created deposits end up at Bank C, and the process iterates.  Because we assume there exists a competitive banking sector, we iterate this deposit creation process infinitely to compute the aggregate amount of new deposits in the banking system:

Creation of Deposits in the Banking System (assuminig 10\% reserve requirement)
| **Bank**                    | **Increase in Deposits**      | **Increase in Loans**        | **Increase in Reserves**   |
|-----------------------------|-------------------------------| ---------------------------- | -------------------------- |
| First National   | 0  |  \$1,000,000$                  | 0      |
| A     |  \$1,000,000$  |  \$900,000$                  |  \$100,000$    |
| B     | \$900,000$  | \$810,000$                  | \$90,000$       |
| C     | \$810,000$   | \$729,000$                  | \$81,000$       |
| E     | \$729,000$   | \$656,100$                  | \$72,900$       |
| ...     | ...  | ...                  | ...      |
| ...     | ...  |...                  | ...      |
|-----------------------------|-------------------------------| ---------------------------- | -------------------------- |
| Total    | Demand Deposits $ \$10,000,000$ | $ \$10,000,000$                  | $ \$1,000,000$      |


The bank still has excess reserves, and so may desire to loan more.  However, the borrowers took out loans not to leave 100 million sitting idle in its account, but to purchase goods and services from other individuals and firms.  When the borrower makes these purchases, the 100 million of reserves will leave First National Bank.

The aggregate creation of deposits follows a geometric series

$$
\Delta D = \Delta R \times [1+(1-rr)+(1-rr)^2+(1-rr)^3+...]
$$

Which gives us a simple relationship between the supply of reserves and deposits :

$$
\Delta D = \frac{1}{rr} \times \Delta R
$$

We call $1/rr$ the **money multiplier**.


If the bank had instead used excess reserves to purchase securities (not issue loans) the result would be the same.  If the borrower holds cash, then the deposit creation process stops. If the bank holds excess reserves, then there is less deposit creation. Money supply M1 depends on depositors decisions (how much currency to hold) and banks decisions (how much excess reserves to hold) that are outside the control of the central bank.

We can construct a slightly more advanced relationship between M1 and high powered money.

Assume that desired holding of currency $C$ and excess reserves $ER$ grow proportional with checkable deposits $D$:

$$
\begin{align*}
c &= C/D = \text{currency ration} \\
e &= ER/D = \text{excess reserve ratio}
\end{align*}
$$

We know that total reserves consist of required reserves and excess reserves

$$
R = RR + ER = (rr \times D) + (e \times D)
$$

and the monetary base is currency plus reserves

$$
MB = R + C = (rr \times D) + (e \times D) + (c \times D)
$$

So that deposits are related to the monetary base by

$$
D = \frac{MB}{rr + e + c}
$$


Recall that M1 is defined as deposits plus currency in circulation:

$$
M1 = D + (c \times D) = (1+c) \times D
$$

Substituting out our relationship between deposits and high powered money gives us

$$
M1 = \frac{1+c}{rr+e+c}MB
$$

so the money multiplier is now

$$
m = \frac{1+c}{rr+e+c}
$$
